<div style="background-color: #f0f8ff; padding: 15px; border-radius: 10px; border-left: 5px solid #4682b4;">
<h4>🔑 Axın şifrələri (Stream Ciphers)</h4>
<p>Axın şifrələri simmetrik şifrələmənin bir növüdür. Onlar açardan istifadə edərək sonsuz uzunluqda <b>psevdo-təsadüfi açar axını (keystream)</b> yaradır və bu axını açıq mətnlə XOR-layaraq şifrəli mətn əldə edirlər:</p>

<p style="text-align: center; font-size: 1.2em;">$$c_i = p_i \oplus k_i$$</p>

<p>Burada $k_i$ açar axınının i-ci biti, $p_i$ açıq mətnin i-ci biti, $c_i$ isə şifrəli mətnin i-ci bitidir.</p>
</div>

**Əsas xüsusiyyətlər:**

| Xüsusiyyət | İzah |
|------------|------|
| **Sürət** | Blok şifrələrdən daha sürətli, xüsusilə hardware-da |
| **Yüngüllük** | Aşağı resurs tələb edir, embedded sistemlər üçün ideal |
| **Xəttilik** | LFSR xəttidir, təkbaşına təhlükəsiz deyil |
| **Period** | Generatorun təkrarlanma müddəti |

**LFSR (Xətti Əks-Əlaqəli Sürüşdürmə Registri)** axın şifrələrinin əsas komponentidir.

## 🔧 LFSR sinfinin implementasiyası (25 dəq)

### 5.1 Əsas LFSR sinfi

LFSR (Linear Feedback Shift Register) - Xətti Əks-Əlaqəli Sürüşdürmə Registri

````python
class LFSR:
    """
    Xətti Əks-Əlaqəli Sürüşdürmə Registri (Linear Feedback Shift Register)

    LFSR bit ardıcıllığı generasiya edən xətti rekursiv registrdir.
    Növbəti bit = XOR(tapmalardakı bitlər)
    """

    def __init__(self, init_state, taps):
        """
        Parametrlər:
            init_state: ilkin vəziyyət (bit siyahısı və ya tam ədəd)
            taps: əks-əlaqə tapmalarının indeksləri (məsələn, [3, 0] - 3 və 0-cı bitlər)
        """
        if isinstance(init_state, int):
            # Tam ədədi bit siyahısına çevir
            self.state = [int(b) for b in bin(init_state)[2:]]
        else:
            self.state = list(init_state)

        self.taps = taps
        self.length = len(self.state)
        self.initial_state = self.state.copy()

    def next_bit(self):
        """
        Növbəti çıxış bitini hesablayır və registri sürüşdürür

        Qaytarır:
            output_bit: çıxış biti (0 və ya 1)
        """
        # Çıxış biti (ən sağdakı bit)
        output_bit = self.state[-1]

        # Əks-əlaqə bitini hesabla (tapmaların XOR-u)
        feedback = 0
        for tap in self.taps:
            feedback ^= self.state[tap]

        # Registri sağa sürüşdür və feedback-i sola daxil et
        self.state = [feedback] + self.state[:-1]

        return output_bit

    def generate_bits(self, n):
        """
        n sayda bit generasiya edir

        Parametrlər:
            n: generasiya ediləcək bit sayı

        Qaytarır:
            bits: bitlər siyahısı
        """
        bits = []
        for _ in range(n):
            bits.append(self.next_bit())
        return bits

    def reset(self):
        """
        LFSR-i ilkin vəziyyətinə qaytarır
        """
        self.state = self.initial_state.copy()

    def get_state(self):
        """
        Cari vəziyyəti qaytarır
        """
        return self.state

    def __str__(self):
        state_str = ''.join(map(str, self.state))
        return f"LFSR(state={state_str}, taps={self.taps})"

# Test nümunəsi
print("=" * 60)
print("🔧 LFSR TESTİ")
print("=" * 60)

# 4-bit LFSR, tapmalar: 3 və 0 (polinom: x^4 + x^3 + 1)
lfsr = LFSR([1, 0, 0, 0], [3, 0])
print(f"LFSR: {lfsr}")

# 20 bit generasiya et
bits = lfsr.generate_bits(20)
print(f"Generasiya edilmiş bitlər: {bits}")
print(f"Bit ardıcıllığı: {''.join(map(str, bits))}")

# Sıfırlama testi
lfsr.reset()
bits2 = lfsr.generate_bits(5)
print(f"Sıfırlandıqdan sonra ilk 5 bit: {bits2}")
print(f"Orijinal ilk 5 bit: {bits[:5]}")
print(f"Eynidir? {bits[:5] == bits2}")
````

### 5.2 Müxtəlif LFSR konfiqurasiyaları

LFSR-lər adətən polinomla təsvir edilir. Məsələn, $x^4 + x^3 + 1$ polinomu 4-bit LFSR deməkdir.

````python
def taps_from_polynomial(polynomial):
    """
    Polinomu bu notebookdakı LFSR tap formatına çevirir.

    Qəbul edilən polinom forması:
        [n, ..., 1, 0]  <=>  x^n + ...

    Bu notebookun konvensiyasında:
        tap indeksi = degree - 1 - exponent
    və sabit 1 termini (0 dərəcə) də taplara daxildir.
    """
    if not polynomial:
        raise ValueError("polynomial boş ola bilməz")

    polynomial = sorted({int(p) for p in polynomial}, reverse=True)
    degree = max(polynomial)

    if degree <= 0:
        raise ValueError("Polinomun dərəcəsi ən azı 1 olmalıdır")
    if 0 not in polynomial:
        raise ValueError("Polinom sabit 1 terminini (0 dərəcə) daxil etməlidir")

    taps = sorted([degree - 1 - p for p in polynomial if p != degree], reverse=True)
    return taps


def create_lfsr_from_polynomial(init_state, polynomial):
    """
    Polinomdan LFSR yaradır
    """
    init_state = list(init_state)
    polynomial = sorted({int(p) for p in polynomial}, reverse=True)
    degree = max(polynomial)

    if len(init_state) != degree:
        raise ValueError(f"init_state uzunluğu polinom dərəcəsinə bərabər olmalıdır ({degree})")

    taps = taps_from_polynomial(polynomial)
    return LFSR(init_state, taps)

# Maksimum periodlu LFSR-lər üçün məşhur polinomlar
print("\n" + "-" * 60)
print("📊 MƏŞHUR LFSR POLİNOMLARI")
print("-" * 60)

# x^3 + x + 1 (3-bit, period 7)
lfsr3 = create_lfsr_from_polynomial([1, 0, 0], [3, 1, 0])
print(f"3-bit LFSR (x^3 + x + 1): {lfsr3}")
bits3 = lfsr3.generate_bits(15)
print(f"   Bitlər: {''.join(map(str, bits3))}")

# x^4 + x^3 + 1 (4-bit, period 15)
lfsr4 = create_lfsr_from_polynomial([1, 0, 0, 0], [4, 3, 0])
print(f"\n4-bit LFSR (x^4 + x^3 + 1): {lfsr4}")
bits4 = lfsr4.generate_bits(20)
print(f"   Bitlər: {''.join(map(str, bits4))}")

# x^5 + x^3 + 1 (5-bit, period 31)
lfsr5 = create_lfsr_from_polynomial([1, 0, 0, 0, 0], [5, 3, 0])
print(f"\n5-bit LFSR (x^5 + x^3 + 1): {lfsr5}")
bits5 = lfsr5.generate_bits(20)
print(f"   Bitlər: {''.join(map(str, bits5))}")
````

## 📊 Period analizi (20 dəq)

<div style="background-color: #f0f8ff; padding: 15px; border-radius: 10px; border-left: 5px solid #4682b4;">
<h4>🔄 LFSR Periodu</h4>
<p>LFSR-in periodu onun təkrarlanmağa başladığı uzunluqdur. n-bit LFSR üçün maksimum mümkün period <b>$2^n - 1$</b>-dir (çünki sıfır vəziyyəti qadağandır).</p>
<p>Maksimum perioda nail olmaq üçün polinom <b>primitiv</b> olmalıdır.</p>
</div>

````python
def find_period(lfsr, max_period=1000):
    """
    LFSR-in periodunu tapır

    Parametrlər:
        lfsr: LFSR obyekti
        max_period: maksimum yoxlanacaq period

    Qaytarır:
        period: tapılan period (max_period-dan böyük ola bilər)
        bits: generasiya edilmiş bitlər
        states: görülmüş vəziyyətlər
    """
    lfsr.reset()
    states = []
    bits = []

    for i in range(max_period):
        current_state = tuple(lfsr.get_state())
        if current_state in states:
            # Period tapıldı
            period = i - states.index(current_state)
            return period, bits, states

        states.append(current_state)
        bits.append(lfsr.next_bit())

    return max_period, bits, states

def analyze_lfsr_period(init_state, taps, max_period=1000):
    """
    LFSR-in periodunu analiz edir

    Parametrlər:
        init_state: ilkin vəziyyət
        taps: tapmalar
        max_period: maksimum yoxlanacaq period

    Qaytarır:
        period: tapılan period
    """
    lfsr = LFSR(init_state, taps)
    period, bits, states = find_period(lfsr, max_period)

    print("=" * 60)
    print(f"📊 LFSR ANALİZİ: taps={taps}, init={init_state}")
    print("=" * 60)
    print(f"📏 Registr uzunluğu: {len(init_state)} bit")
    print(f"🔄 Period: {period}")
    print(f"📈 Maksimum mümkün period: {2**len(init_state) - 1}")
    print(f"🔢 İlk 20 bit: {bits[:20]}")

    if period > 0:
        ones = sum(bits[:period])
        zeros = period - ones
        print(f"   1-lərin sayı: {ones}")
        print(f"   0-ların sayı: {zeros}")
        print(f"   Nisbət: {ones/period:.2f}")

    return period, bits

print("\n" + "=" * 60)
print("🔄 PERİOD ANALİZİ")
print("=" * 60)

# 3-bit LFSR (x^3 + x + 1)
period3, _ = analyze_lfsr_period([1, 0, 0], [2, 0], 50)
print(f"\n3-bit üçün maksimum: {2**3 - 1} - {'✓ maksimum' if period3 == 2**3-1 else '✗ deyil'}")

# 4-bit LFSR (x^4 + x^3 + 1)
period4, _ = analyze_lfsr_period([1, 0, 0, 0], [3, 2], 50)
print(f"\n4-bit üçün maksimum: {2**4 - 1} - {'✓ maksimum' if period4 == 2**4-1 else '✗ deyil'}")

# 5-bit LFSR (x^5 + x^3 + 1)
period5, _ = analyze_lfsr_period([1, 0, 0, 0, 0], [4, 2], 100)
print(f"\n5-bit üçün maksimum: {2**5 - 1} - {'✓ maksimum' if period5 == 2**5-1 else '✗ deyil'}")
````

### 6.2 Maksimum periodlu LFSR-lər

Bütün polinomlar maksimum period vermir. Primitiv polinomlar maksimum period təmin edir.

````python
def find_maximum_period_lfsr(length, max_trials=100):
    """
    Verilmiş uzunluq üçün maksimum periodlu LFSR tapmağa çalışır

    Parametrlər:
        length: LFSR uzunluğu (bit)
        max_trials: maksimum sınaq sayı

    Qaytarır:
        found_lfsrs: tapılan LFSR-lərin siyahısı
    """
    max_period_possible = 2**length - 1
    found_lfsrs = []

    print(f"\n🔍 {length}-bit üçün maksimum periodlu LFSR axtarışı...")
    print(f"   Maksimum mümkün period: {max_period_possible}")
    print("-" * 50)

    for trial in range(max_trials):
        # Təsadüfi taps seç (ən azı 1 tap)
        tap_count = random.randint(1, length)
        taps = random.sample(range(length), tap_count)
        taps.sort(reverse=True)

        # Təsadüfi ilkin vəziyyət (sıfır olmayan)
        init_state = [random.randint(0, 1) for _ in range(length)]
        while sum(init_state) == 0:  # Sıfır vəziyyətindən qaç
            init_state = [random.randint(0, 1) for _ in range(length)]

        lfsr = LFSR(init_state, taps)
        period, _, _ = find_period(lfsr, max_period_possible + 10)

        if period == max_period_possible:
            found_lfsrs.append((taps, init_state))
            print(f"   ✓ Tapıldı! taps={taps}, init={init_state}")

    if not found_lfsrs:
        print(f"   ✗ Heç bir maksimum periodlu LFSR tapılmadı (cəmi {max_trials} sınaq)")

    return found_lfsrs

# Məlum maksimum periodlu polinomlar (primitiv polinomlar)
max_period_polynomials = {
    2: [2, 1, 0],      # x^2 + x + 1
    3: [3, 1, 0],      # x^3 + x + 1
    4: [4, 1, 0],      # x^4 + x + 1
    5: [5, 2, 0],      # x^5 + x^2 + 1
    6: [6, 1, 0],      # x^6 + x + 1
    7: [7, 1, 0],      # x^7 + x + 1
    8: [8, 4, 3, 2, 0] # x^8 + x^4 + x^3 + x^2 + 1
}

print("\n" + "=" * 70)
print("📚 MƏLUM MAKSİMUM PERİODLU LFSR-LƏR (Primitiv Polinomlar)")
print("=" * 70)

for length, poly in max_period_polynomials.items():
    # Polinomu gözəl formatda göstər
    terms = []
    for p in poly:
        if p == 0:
            terms.append("1")
        elif p == 1:
            terms.append("x")
        else:
            terms.append(f"x^{p}")
    polynomial_str = " + ".join(terms)

    print(f"\n{length} bit: {polynomial_str}")
    print(f"   Period: {2**length - 1}")

    # Tapmaları göstər
    taps = taps_from_polynomial(poly)
    print(f"   Tapmalar: {taps}")
````

## 🔐 LFSR əsaslı axın şifrələmə (20 dəq)

````python
# ⚠️ XƏBƏRDARLIQ: Bu sinif yalnız pedaqoji məqsədli oyuncaq nümunədir.
# Kiçik LFSR-lərə əsaslanan bu konstruktsiya təhlükəsiz axın şifrəsi deyil və
# real sistemlərdə istifadə olunmamalıdır. Praktik tətbiqlərdə düzgün nonce
# idarəetməsi ilə ChaCha20-Poly1305 və ya AES-GCM kimi yoxlanılmış AEAD
# konstruktsiyalarından istifadə edin.

class StreamCipher:
    """
    LFSR əsaslı axın şifrəsi

    Açar axını (keystream) LFSR tərəfindən generasiya edilir
    və açıq mətnlə XOR-lanır.
    """

    def __init__(self, lfsr):
        """
        Parametrlər:
            lfsr: açar axını generasiya edəcək LFSR obyekti
        """
        self.lfsr = lfsr
        self.keystream = []

    def generate_keystream(self, length):
        """
        Verilmiş uzunluqda açar axını generasiya edir

        Parametrlər:
            length: tələb olunan bit sayı

        Qaytarır:
            keystream: açar axını bitləri
        """
        self.keystream = self.lfsr.generate_bits(length)
        return self.keystream

    def encrypt(self, plaintext):
        """
        Açıq mətni şifrləyir (hər simvol üçün 8 bit istifadə edir)

        Parametrlər:
            plaintext: şifrlənəcək mətn (str və ya bytes)

        Qaytarır:
            ciphertext: şifrəli mətn (bytes)
        """
        if isinstance(plaintext, str):
            plaintext = plaintext.encode('utf-8')

        # Açar axını generasiya et (hər bayt üçün 8 bit)
        keystream_bits = self.generate_keystream(len(plaintext) * 8)

        ciphertext = bytearray()
        for i, byte in enumerate(plaintext):
            # Bu bayt üçün 8 bit açar götür
            key_byte = 0
            for j in range(8):
                key_byte = (key_byte << 1) | keystream_bits[i*8 + j]

            # XOR tətbiq et
            ciphertext.append(byte ^ key_byte)

        return bytes(ciphertext)

    def decrypt(self, ciphertext):
        """
        Şifrəli mətni deşifrləyir (XOR simmetrikdir)

        Parametrlər:
            ciphertext: şifrəli mətn (bytes)

        Qaytarır:
            plaintext: deşifrələnmiş mətn (bytes)
        """
        # Eyni açar axınını generasiya et (LFSR-i sıfırla)
        self.lfsr.reset()
        return self.encrypt(ciphertext)

# Test
print("\n" + "=" * 70)
print("🔐 LFSR ƏSASLI AXIN ŞİFRƏSİ TESTİ")
print("=" * 70)

# LFSR yarat (maksimum periodlu - x^5 + x^2 + 1)
lfsr = LFSR([1, 0, 0, 0, 0], [4, 2])
cipher = StreamCipher(lfsr)

plaintext = "Salam, dünya! Bu gizli mesajdır."
print(f"📨 Orijinal mətn: {plaintext}")

ciphertext = cipher.encrypt(plaintext)
print(f"🔒 Şifrəli mətn (hex): {ciphertext.hex()}")

decrypted = cipher.decrypt(ciphertext)
print(f"🔓 Deşifrələnmiş: {decrypted.decode('utf-8')}")

print(f"\n✅ Uğurlu? {plaintext == decrypted.decode('utf-8')}")
````

### 7.2 Xəttilik testi

LFSR-in xətti olması onun ən böyük zəifliyidir. Bu xəttiliyi test edək:

````python
def linearity_test(lfsr, num_tests=100):
    """
    LFSR-in xətti olub-olmadığını test edir

    Xəttilik prinsipi: iki ardıcıllığın XOR-u,
    onların ayrı-ayrı XOR-larının cəminə bərabər olmalıdır
    """
    print("\n" + "=" * 60)
    print("📐 XƏTTİLİK TESTİ")
    print("=" * 60)

    lfsr.reset()
    # İki müxtəlif ilkin vəziyyət
    state1 = lfsr.get_state().copy()

    lfsr.reset()
    # İkinci ilkin vəziyyət (bir bit dəyişdir)
    state2 = state1.copy()
    state2[0] ^= 1

    print(f"İlkin vəziyyət 1: {state1}")
    print(f"İlkin vəziyyət 2: {state2}")
    print(f"XOR vəziyyəti: {[a^b for a,b in zip(state1, state2)]}")

    # Hər iki LFSR-dən ardıcıllıq generasiya et
    lfsr1 = LFSR(state1, lfsr.taps)
    lfsr2 = LFSR(state2, lfsr.taps)

    seq1 = lfsr1.generate_bits(num_tests)
    seq2 = lfsr2.generate_bits(num_tests)

    # seq1 XOR seq2
    xor_seq = [a ^ b for a, b in zip(seq1, seq2)]

    # İlkin vəziyyətlərin XOR-dan yeni LFSR yarat
    xor_state = [a ^ b for a, b in zip(state1, state2)]
    lfsr_xor = LFSR(xor_state, lfsr.taps)
    seq_xor = lfsr_xor.generate_bits(num_tests)

    # Müqayisə
    print(f"\nseq1 XOR seq2 (ilk 20): {xor_seq[:20]}")
    print(f"XOR LFSR ardıcıllığı (ilk 20): {seq_xor[:20]}")
    print(f"\nBərabərdir? {xor_seq == seq_xor}")

    if xor_seq == seq_xor:
        print("✅ LFSR xətti xassəyə malikdir")
        print("   Bu, Berlekamp-Massey alqoritmi ilə asanlıqla bərpa edilə bilər")
    else:
        print("❌ LFSR xətti deyil")

# Test
lfsr_test = LFSR([1, 0, 0, 0, 0], [4, 2])
linearity_test(lfsr_test)
````

## 🔬 Berlekamp-Massey alqoritmi (25 dəq)

<div style="background-color: #fff3e0; padding: 15px; border-radius: 10px; border-left: 5px solid #ff9800;">
<h4>📜 Berlekamp-Massey alqoritmi</h4>
<p>Berlekamp-Massey alqoritmi verilmiş ardıcıllığı generasiya edən <b>ən qısa LFSR-i</b> tapmaq üçün istifadə olunur. Alqoritm 1969-cu ildə <b>James Massey</b> tərəfindən Elwyn Berlekampın işinə əsaslanaraq inkişaf etdirilmişdir.</p>
<p>Bu alqoritm xətti rekursiv ardıcıllıqların analizində mühüm rol oynayır və LFSR-in xətti olmasının ən böyük zəiflik olduğunu göstərir.</p>
</div>

````python
def berlekamp_massey(sequence):
    """
    Berlekamp-Massey alqoritmi - verilmiş ardıcıllığı generasiya edən
    ən qısa LFSR-i tapır

    Parametrlər:
        sequence: bit ardıcıllığı (0/1 siyahısı)

    Qaytarır:
        taps: tapmalar siyahısı
        length: LFSR uzunluğu
    """
    n = len(sequence)

    c = [0] * (n + 1)
    b = [0] * (n + 1)
    c[0] = 1
    b[0] = 1
    l = 0
    m = -1

    for i in range(n):
        d = sequence[i]
        for j in range(1, l + 1):
            if c[j] == 1:
                d ^= sequence[i - j]

        if d == 1:
            t = c.copy()
            shift = i - m

            for j in range(n + 1 - shift):
                if b[j] == 1:
                    c[shift + j] ^= 1

            if l <= i // 2:
                l = i + 1 - l
                m = i
                b = t.copy()

    taps = sorted([j - 1 for j in range(1, l + 1) if c[j] == 1], reverse=True)
    return taps, l

def test_berlekamp_massey():
    """
    Berlekamp-Massey alqoritmini test edir
    """
    print("\n" + "=" * 70)
    print("🔬 BERLEKAMP-MASSEY ALQORİTMİ TESTİ")
    print("=" * 70)

    # Məlum LFSR-dən ardıcıllıq generasiya et
    lfsr = LFSR([1, 0, 0, 0], [3, 2])  # 4-bit LFSR
    sequence = lfsr.generate_bits(30)

    print(f"Original LFSR: taps={lfsr.taps}, length={lfsr.length}")
    print(f"Ardıcıllıq (ilk 20): {sequence[:20]}")

    # Berlekamp-Massey tətbiq et
    taps, length = berlekamp_massey(sequence)

    print(f"\nTapılan LFSR: taps={taps}, length={length}")

    # Tapılan LFSR ilə ardıcıllığı generasiya et
    if taps:
        # Bu LFSR sağa sürüşdürür və sağ biti çıxış verir; buna görə
        # başlanğıc vəziyyət ilk length çıxış bitlərinin tərsidir.
        recovered_state = list(reversed(sequence[:length]))
        recovered_lfsr = LFSR(recovered_state, taps)
        recovered_seq = recovered_lfsr.generate_bits(len(sequence))

        print(f"Bərpa edilmiş başlanğıc vəziyyət: {recovered_state}")
        print(f"Bərpa edilmiş ardıcıllıq (ilk 20): {recovered_seq[:20]}")
        print(f"Uyğunluq: {sequence == recovered_seq}")

        if sequence == recovered_seq:
            print("✅ Alqoritm düzgün işləyir!")

test_berlekamp_massey()
````

<div style="background-color: #f0f8ff; padding: 15px; border-radius: 10px; border-left: 5px solid #4682b4;">
<h4>🔀 Geffe generatoru</h4>
<p>LFSR-in xətti olması onu kriptoanalizə qarşı həssas edir. Geffe generatoru <b>üç LFSR-in çıxışlarını qeyri-xətti kombinasiya edərək</b> daha təhlükəsiz açar axını yaradır:</p>

<p style="text-align: center; font-size: 1.2em;">$$F(x,y,z) = (x \land y) \oplus (\lnot x \land z)$$</p>

<p>Bu, seçici (selector) əsasında işləyir: x=1 olduqda y, x=0 olduqda z seçilir.</p>
</div>

````python
class GeffeGenerator:
    """
    Geffe generatoru - üç LFSR-in qeyri-xətti kombinasiyası

    F(x,y,z) = (x ∧ y) ⊕ (¬x ∧ z)
    """

    def __init__(self, lfsr1, lfsr2, lfsr3):
        """
        Parametrlər:
            lfsr1, lfsr2, lfsr3: üç LFSR obyekti
        """
        self.lfsr1 = lfsr1
        self.lfsr2 = lfsr2
        self.lfsr3 = lfsr3

    def next_bit(self):
        """
        Növbəti biti generasiya edir
        F(x,y,z) = (x ∧ y) ⊕ (¬x ∧ z)
        """
        x = self.lfsr1.next_bit()
        y = self.lfsr2.next_bit()
        z = self.lfsr3.next_bit()

        # (x ∧ y) XOR ((not x) ∧ z)
        return (x & y) ^ ((1 - x) & z)

    def generate_bits(self, n):
        """
        n sayda bit generasiya edir
        """
        return [self.next_bit() for _ in range(n)]

    def reset(self):
        """Bütün LFSR-ləri sıfırlayır"""
        self.lfsr1.reset()
        self.lfsr2.reset()
        self.lfsr3.reset()

# Test
print("\n" + "=" * 70)
print("🔀 GEFFE GENERATORU TESTİ")
print("=" * 70)

# Üç LFSR yarat
lfsr1 = LFSR([1, 0, 0, 0], [3, 2])  # 4-bit
lfsr2 = LFSR([1, 0, 0, 0, 0], [4, 2])  # 5-bit
lfsr3 = LFSR([1, 0, 0, 0, 0, 0], [5, 2])  # 6-bit

geffe = GeffeGenerator(lfsr1, lfsr2, lfsr3)
bits = geffe.generate_bits(50)

print(f"Geffe generatoru bitləri (ilk 20): {bits[:20]}")
print(f"1-lərin sayı: {sum(bits)}")
print(f"0-ların sayı: {len(bits) - sum(bits)}")
print(f"1-lərin nisbəti: {sum(bits)/len(bits):.2f}")

# Xəttilik testi (təxmini)
# Məlum LFSR-lərlə müqayisə
lfsr1_bits = LFSR([1, 0, 0, 0], [3, 2]).generate_bits(50)
correlation = sum(1 for a, b in zip(bits, lfsr1_bits) if a == b)
print(f"LFSR1 ilə korrelyasiya: {correlation/50:.2f}")

# Berlekamp-Massey ilə test
taps_geffe, len_geffe = berlekamp_massey(bits)
print(f"Geffe ardıcıllığı üçün LFSR uzunluğu: {len_geffe}")
print(f"Bu, fərdi LFSR-lərin uzunluqları cəmindən böyükdür")
````

4. **Alternativ funksiyalar:** Başqa qeyri-xətti kombinasiya funksiyaları araşdırın:
   * Majority function (əksəriyyət funksiyası)
   * Threshold function

5. **Öz generatorunuz:** Öz qeyri-xətti generatorunuzu dizayn edin və test edin.

````python
# Çalışma 5 - Cavablar

print("📝 ÇALIŞMA 5 CAVABLARI")
print("=" * 80)

# 1. Geffe generatoru analizi
print("\n1. GEFFE GENERATORU ANALİZİ:")
l1 = LFSR([1,0,0,0], [3,2])
l2 = LFSR([1,0,0,0,0], [4,2])
l3 = LFSR([1,0,0,0,0,0], [5,2])
g = GeffeGenerator(l1,l2,l3)
bits_geffe = g.generate_bits(1000)

ones = sum(bits_geffe)
print(f"   1000 bitdə 1-lərin sayı: {ones}")
print(f"   1-lərin nisbəti: {ones/1000:.3f}")

# 2. Müqayisə
print("\n2. MÜQAYİSƏ:")
simple_lfsr = LFSR([1,0,0,0], [3,2])
simple_bits = simple_lfsr.generate_bits(1000)

taps_simple, len_simple = berlekamp_massey(simple_bits[:100])
taps_geffe, len_geffe = berlekamp_massey(bits_geffe[:100])

print(f"   Sadə LFSR üçün LFSR uzunluğu: {len_simple}")
print(f"   Geffe üçün LFSR uzunluğu: {len_geffe}")
print(f"   Geffe daha 'təsadüfi' görünür, LFSR uzunluğu daha böyük")

# 3. Korrelyasiya hücumu
print("\n3. KORELYASİYA HÜCUMU:")
print("   • Geffe generatorunda çıxış y və z LFSR-ləri ilə korrelyasiyadadır")
print("   • P(y = çıxış) = 3/4 və P(z = çıxış) = 3/4")
print("   • P(x = çıxış) = 1/2")
print("   • Bu statistik zəiflik hücum üçün istifadə edilə bilər")

# 4. Alternativ funksiyalar
print("\n4. ALTERNATİV FUNKSİYALAR:")
print("   • Majority function: F(x,y,z) = (x∧y) ∨ (x∧z) ∨ (y∧z)")
print("   • Threshold function: ən azı 2 bit 1 olduqda 1")

# 5. Öz generatoru
print("\n5. ÖZ GENERATORU:")
def custom_generator(l1, l2, l3):
    """Sadə majority funksiyası"""
    x = l1.next_bit()
    y = l2.next_bit()
    z = l3.next_bit()
    # Majority: ən azı 2 bit 1 olarsa 1
    return 1 if (x + y + z) >= 2 else 0

print("   Majority funksiyası əsasında generator yaradıldı")
````
