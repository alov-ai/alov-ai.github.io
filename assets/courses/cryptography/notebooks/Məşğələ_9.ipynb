## 🔧 DES-in komponentləri (25 dəq)

### 5.1 İlkin permutasiya (IP) və son permutasiya (FP)

````python
# DES ilkin permutasiya cədvəli (IP)
IP_TABLE = [
    58, 50, 42, 34, 26, 18, 10, 2,
    60, 52, 44, 36, 28, 20, 12, 4,
    62, 54, 46, 38, 30, 22, 14, 6,
    64, 56, 48, 40, 32, 24, 16, 8,
    57, 49, 41, 33, 25, 17, 9, 1,
    59, 51, 43, 35, 27, 19, 11, 3,
    61, 53, 45, 37, 29, 21, 13, 5,
    63, 55, 47, 39, 31, 23, 15, 7
]

# Son permutasiya (IP-in tərsi)
FP_TABLE = [
    40, 8, 48, 16, 56, 24, 64, 32,
    39, 7, 47, 15, 55, 23, 63, 31,
    38, 6, 46, 14, 54, 22, 62, 30,
    37, 5, 45, 13, 53, 21, 61, 29,
    36, 4, 44, 12, 52, 20, 60, 28,
    35, 3, 43, 11, 51, 19, 59, 27,
    34, 2, 42, 10, 50, 18, 58, 26,
    33, 1, 41, 9, 49, 17, 57, 25
]

def apply_permutation(block, table, n=64):
    """
    Permutasiya cədvəlini tətbiq edir

    Parametrlər:
        block: giriş bloku (int)
        table: permutasiya cədvəli
        n: blok ölçüsü (bit)

    Qaytarır:
        result: permutasiya olunmuş blok
    """
    result = 0
    for i, pos in enumerate(table):
        # pos 1-indekslidir, ona görə pos-1
        bit = (block >> (n - pos)) & 1
        if bit:
            result |= (1 << (n - i - 1))
    return result

def initial_permutation(block):
    """İlkin permutasiya (IP)"""
    return apply_permutation(block, IP_TABLE)

def final_permutation(block):
    """Son permutasiya (FP)"""
    return apply_permutation(block, FP_TABLE)

# Test
print("=" * 70)
print("🔄 DES PERMUTASİYA TESTİ")
print("=" * 70)

test_block = 0x0123456789ABCDEF
print(f"Test bloku: {test_block:016x}")

ip_result = initial_permutation(test_block)
print(f"IP-dən sonra: {ip_result:016x}")

fp_result = final_permutation(ip_result)
print(f"FP-dən sonra: {fp_result:016x}")
print(f"✅ Orijinala bərabər? {test_block == fp_result}")
````

### 5.2 Genişləndirmə (Expansion) cədvəli

````python
# DES genişləndirmə cədvəli (E)
E_TABLE = [
    32, 1, 2, 3, 4, 5,
    4, 5, 6, 7, 8, 9,
    8, 9, 10, 11, 12, 13,
    12, 13, 14, 15, 16, 17,
    16, 17, 18, 19, 20, 21,
    20, 21, 22, 23, 24, 25,
    24, 25, 26, 27, 28, 29,
    28, 29, 30, 31, 32, 1
]

def expand_block(block, table=E_TABLE, input_bits=32, output_bits=48):
    """
    Bloku genişləndirir (32 bit -> 48 bit)

    Parametrlər:
        block: 32-bit giriş
        table: genişləndirmə cədvəli
        input_bits: giriş bit sayı
        output_bits: çıxış bit sayı

    Qaytarır:
        result: 48-bit genişləndirilmiş blok
    """
    result = 0
    for i, pos in enumerate(table):
        # pos 1-indekslidir
        bit = (block >> (input_bits - pos)) & 1
        if bit:
            result |= (1 << (output_bits - i - 1))
    return result

# Test
print("\n" + "=" * 70)
print("📏 GENİŞLƏNDİRMƏ TESTİ")
print("=" * 70)

R = 0xF0F0F0F0  # 32-bit sağ yarım
expanded = expand_block(R)
print(f"R (32-bit): {R:08x}")
print(f"E(R) (48-bit): {expanded:012x}")

# Bitlərin təkrarlanmasını yoxlayaq
print("\nBit təkrarlanması:")
for i in range(0, 48, 6):
    block = (expanded >> (42 - i)) & 0x3F
    print(f"  Blok {i//6 + 1}: {block:06b}")
````

````python
# DES S-qutuları (8 ədəd, hər biri 6-bit giriş, 4-bit çıxış)
S_BOXES = [
    # S1
    [
        [14, 4, 13, 1, 2, 15, 11, 8, 3, 10, 6, 12, 5, 9, 0, 7],
        [0, 15, 7, 4, 14, 2, 13, 1, 10, 6, 12, 11, 9, 5, 3, 8],
        [4, 1, 14, 8, 13, 6, 2, 11, 15, 12, 9, 7, 3, 10, 5, 0],
        [15, 12, 8, 2, 4, 9, 1, 7, 5, 11, 3, 14, 10, 0, 6, 13]
    ],
    # S2
    [
        [15, 1, 8, 14, 6, 11, 3, 4, 9, 7, 2, 13, 12, 0, 5, 10],
        [3, 13, 4, 7, 15, 2, 8, 14, 12, 0, 1, 10, 6, 9, 11, 5],
        [0, 14, 7, 11, 10, 4, 13, 1, 5, 8, 12, 6, 9, 3, 2, 15],
        [13, 8, 10, 1, 3, 15, 4, 2, 11, 6, 7, 12, 0, 5, 14, 9]
    ],
    # S3
    [
        [10, 0, 9, 14, 6, 3, 15, 5, 1, 13, 12, 7, 11, 4, 2, 8],
        [13, 7, 0, 9, 3, 4, 6, 10, 2, 8, 5, 14, 12, 11, 15, 1],
        [13, 6, 4, 9, 8, 15, 3, 0, 11, 1, 2, 12, 5, 10, 14, 7],
        [1, 10, 13, 0, 6, 9, 8, 7, 4, 15, 14, 3, 11, 5, 2, 12]
    ],
    # S4
    [
        [7, 13, 14, 3, 0, 6, 9, 10, 1, 2, 8, 5, 11, 12, 4, 15],
        [13, 8, 11, 5, 6, 15, 0, 3, 4, 7, 2, 12, 1, 10, 14, 9],
        [10, 6, 9, 0, 12, 11, 7, 13, 15, 1, 3, 14, 5, 2, 8, 4],
        [3, 15, 0, 6, 10, 1, 13, 8, 9, 4, 5, 11, 12, 7, 2, 14]
    ],
    # S5
    [
        [2, 12, 4, 1, 7, 10, 11, 6, 8, 5, 3, 15, 13, 0, 14, 9],
        [14, 11, 2, 12, 4, 7, 13, 1, 5, 0, 15, 10, 3, 9, 8, 6],
        [4, 2, 1, 11, 10, 13, 7, 8, 15, 9, 12, 5, 6, 3, 0, 14],
        [11, 8, 12, 7, 1, 14, 2, 13, 6, 15, 0, 9, 10, 4, 5, 3]
    ],
    # S6
    [
        [12, 1, 10, 15, 9, 2, 6, 8, 0, 13, 3, 4, 14, 7, 5, 11],
        [10, 15, 4, 2, 7, 12, 9, 5, 6, 1, 13, 14, 0, 11, 3, 8],
        [9, 14, 15, 5, 2, 8, 12, 3, 7, 0, 4, 10, 1, 13, 11, 6],
        [4, 3, 2, 12, 9, 5, 15, 10, 11, 14, 1, 7, 6, 0, 8, 13]
    ],
    # S7
    [
        [4, 11, 2, 14, 15, 0, 8, 13, 3, 12, 9, 7, 5, 10, 6, 1],
        [13, 0, 11, 7, 4, 9, 1, 10, 14, 3, 5, 12, 2, 15, 8, 6],
        [1, 4, 11, 13, 12, 3, 7, 14, 10, 15, 6, 8, 0, 5, 9, 2],
        [6, 11, 13, 8, 1, 4, 10, 7, 9, 5, 0, 15, 14, 2, 3, 12]
    ],
    # S8
    [
        [13, 2, 8, 4, 6, 15, 11, 1, 10, 9, 3, 14, 5, 0, 12, 7],
        [1, 15, 13, 8, 10, 3, 7, 4, 12, 5, 6, 11, 0, 14, 9, 2],
        [7, 11, 4, 1, 9, 12, 14, 2, 0, 6, 10, 13, 15, 3, 5, 8],
        [2, 1, 14, 7, 4, 10, 8, 13, 15, 12, 9, 0, 3, 5, 6, 11]
    ]
]

def apply_sbox(input_48bit):
    """
    48-bit girişə S-qutularını tətbiq edir, 32-bit çıxış qaytarır

    Parametrlər:
        input_48bit: 48-bit giriş (int)

    Qaytarır:
        32-bit çıxış
    """
    result = 0
    for i in range(8):
        # 6-bit bloku götür
        block = (input_48bit >> (42 - 6*i)) & 0x3F

        # Sətir: birinci və sonuncu bit (b1b6)
        row = ((block >> 4) & 0x2) | (block & 0x1)

        # Sütun: orta 4 bit (b2b3b4b5)
        col = (block >> 1) & 0xF

        # S-box-dan dəyəri götür
        sbox_value = S_BOXES[i][row][col]

        # 4-bit çıxışı əlavə et
        result |= (sbox_value << (28 - 4*i))

    return result

def sbox_lookup(sbox_num, input_6bit):
    """
    Tək S-qutusuna müraciət edir
    """
    row = ((input_6bit >> 4) & 0x2) | (input_6bit & 0x1)
    col = (input_6bit >> 1) & 0xF
    return S_BOXES[sbox_num][row][col]

# Test
print("\n" + "=" * 70)
print("🔢 S-QUTULARI TESTİ")
print("=" * 70)

test_48bit = 0x123456789ABC
sbox_output = apply_sbox(test_48bit)
print(f"Giriş (48-bit): {test_48bit:012x}")
print(f"Çıxış (32-bit): {sbox_output:08x}")

# S1 qutusunun testi
print("\nS1 qutusu testi:")
for val in [0x00, 0x3F, 0x20, 0x15]:
    out = sbox_lookup(0, val)
    print(f"  S1({val:02x}) = {out:01x}")
````

### 5.4 P-qutusu (permutasiya)

````python
# DES P-qutusu
P_TABLE = [
    16, 7, 20, 21, 29, 12, 28, 17,
    1, 15, 23, 26, 5, 18, 31, 10,
    2, 8, 24, 14, 32, 27, 3, 9,
    19, 13, 30, 6, 22, 11, 4, 25
]

def apply_pbox(input_32bit):
    """
    P-qutusunu tətbiq edir (32-bit permutasiya)

    Parametrlər:
        input_32bit: 32-bit giriş

    Qaytarır:
        32-bit permutasiya olunmuş çıxış
    """
    result = 0
    for i, pos in enumerate(P_TABLE):
        bit = (input_32bit >> (32 - pos)) & 1
        if bit:
            result |= (1 << (32 - i - 1))
    return result

# Test
print("\n" + "=" * 70)
print("🔄 P-QUTUSU TESTİ")
print("=" * 70)

test_32bit = sbox_output
pbox_output = apply_pbox(test_32bit)
print(f"Giriş: {test_32bit:08x}")
print(f"Çıxış: {pbox_output:08x}")

# Bitlərin hərəkətini yoxlayaq
print("\nBit permutasiyası:")
for i, pos in enumerate(P_TABLE[:8]):
    print(f"  Bit {i+1:2d} <- Bit {pos:2d}")
````

### 5.5 DES raund funksiyası (F)

````python
def des_round_function(R, K):
    """
    DES raund funksiyası F(R, K)

    Parametrlər:
        R: 32-bit sağ yarım
        K: 48-bit raund açarı

    Qaytarır:
        32-bit nəticə
    """
    # 1. Genişləndirmə (32 -> 48)
    expanded = expand_block(R)

    # 2. Açarın əlavə edilməsi (XOR)
    xored = expanded ^ K

    # 3. S-qutuları (48 -> 32)
    sboxed = apply_sbox(xored)

    # 4. P-qutusu
    result = apply_pbox(sboxed)

    return result

# Test
print("\n" + "=" * 70)
print("🔧 DES RAUND FUNKSİYASI TESTİ")
print("=" * 70)

test_R = 0xF0F0F0F0
test_K = 0x123456789ABC
f_output = des_round_function(test_R, test_K)
print(f"R (32-bit): {test_R:08x}")
print(f"K (48-bit): {test_K:012x}")
print(f"F(R,K) (32-bit): {f_output:08x}")
````

## 🔐 3DES (Triple DES) (15 dəq)

<div style="background-color: #f0f8ff; padding: 15px; border-radius: 10px; border-left: 5px solid #4682b4;">
<h4>🔑 3DES (Triple DES)</h4>
<p>3DES üç müxtəlif rejimdə işləyə bilər:</p>
<ul>
    <li><b>DES-EEE3:</b> Üç müstəqil açarla üç dəfə şifrləmə: $C = E_{K3}(E_{K2}(E_{K1}(P)))$</li>
    <li><b>DES-EDE3:</b> Şifrləmə-deşifrləmə-şifrləmə: $C = E_{K3}(D_{K2}(E_{K1}(P)))$</li>
    <li><b>DES-EDE2:</b> İki açarla (K1=K3): $C = E_{K1}(D_{K2}(E_{K1}(P)))$</li>
</ul>
<p>EDE rejimi geriyə uyğunluq təmin edir: K1=K2=K3 olduqda DES-ə bərabər olur.</p>
</div>

````python
def des_encrypt_block(key_8, block_8):
    """1 blok DES şifrləmə (ECB yalnız xam blok primitivini çağırmaq üçündür)."""
    if len(key_8) != 8 or len(block_8) != 8:
        raise ValueError("DES üçün açar və blok 8 bayt olmalıdır")
    # TƏDRİS ÜÇÜN XƏBƏRDARLIQ: ECB çoxbloklu real şifrələmə üçün təhlükəsiz deyil;
    # eyni açıq mətn blokları eyni şifrəli mətn blokları verir və strukturu sızdırır.
    # Burada yalnız DES-in tək-blok çevirməsini göstərmək üçün istifadə olunur.
    return DES.new(key_8, DES.MODE_ECB).encrypt(block_8)


def des_decrypt_block(key_8, block_8):
    """1 blok DES deşifrləmə (ECB yalnız xam blok primitivini çağırmaq üçündür)."""
    if len(key_8) != 8 or len(block_8) != 8:
        raise ValueError("DES üçün açar və blok 8 bayt olmalıdır")
    # Tək blok nümunəsidir; real çoxbloklu tətbiqlərdə ECB istifadə etməyin.
    return DES.new(key_8, DES.MODE_ECB).decrypt(block_8)


def triple_des_eee3_block(block_8, k1, k2, k3):
    """DES-EEE3: E_K3(E_K2(E_K1(P)))"""
    return des_encrypt_block(k3, des_encrypt_block(k2, des_encrypt_block(k1, block_8)))


def triple_des_ede3_block(block_8, k1, k2, k3):
    """DES-EDE3: E_K3(D_K2(E_K1(P)))"""
    return des_encrypt_block(k3, des_decrypt_block(k2, des_encrypt_block(k1, block_8)))


def triple_des_ede2_block(block_8, k1, k2):
    """DES-EDE2: E_K1(D_K2(E_K1(P)))  (K1 = K3)"""
    return des_encrypt_block(k1, des_decrypt_block(k2, des_encrypt_block(k1, block_8)))


def triple_des_demo():
    """
    PyCryptodome ilə 3DES istifadəsi
    """
    if not DES_AVAILABLE:
        return

    print("\n" + "=" * 80)
    print("🔐 3DES (TRIPLE DES) NÜMAYİŞİ")
    print("=" * 80)
    print("⚠️  Qeyd: MODE_ECB burada yalnız 8 baytlıq tək blok nümunəsi üçündür; real çoxbloklu şifrələmədə ECB istifadə etməyin.")

    # 3 açarlı TDES üçün 24-bayt xammal açar -> düzgün parity bitləri ilə tənzimlənir
    raw_key_24 = bytes.fromhex('0123456789ABCDEFFEDCBA98765432100011223344556677')
    key_24 = DES3.adjust_key_parity(raw_key_24)

    print(f"3 açarlı TDES açarı: {key_24.hex()}")
    print(f"Açar uzunluğu: {len(key_24)} bayt")

    # 3DES şifrəsi (PyCryptodome DES-EDE3 istifadə edir).
    # ECB semantik təhlükəsiz rejim deyil; bu demo yalnız bir blok üzərində işləyir.
    cipher = DES3.new(key_24, DES3.MODE_ECB)

    # Test məlumatı (tam 1 blok = 8 bayt)
    plaintext = b'Message1'
    print(f"\n📨 Açıq mətn: {plaintext}")

    # Şifrləmə
    ciphertext = cipher.encrypt(plaintext)
    print(f"🔒 Şifrəli mətn (hex): {ciphertext.hex()}")

    # Deşifrləmə
    cipher_dec = DES3.new(key_24, DES3.MODE_ECB)
    decrypted = cipher_dec.decrypt(ciphertext)
    print(f"🔓 Deşifrələnmiş: {decrypted}")

    # İki açarlı 3DES (DES-EDE2)
    print("\n" + "-" * 50)
    print("📌 İKİ AÇARLI 3DES (DES-EDE2):")
    raw_key_16 = bytes.fromhex('0123456789ABCDEFFEDCBA9876543210')
    key_16 = DES3.adjust_key_parity(raw_key_16)
    print(f"2 açarlı TDES açarı: {key_16.hex()}")
    print(f"Açar uzunluğu: {len(key_16)} bayt")

    # ECB burada da yalnız tək bloklu demonstrasiya üçündür.
    cipher2 = DES3.new(key_16, DES3.MODE_ECB)
    ciphertext2 = cipher2.encrypt(plaintext)
    print(f"Şifrəli: {ciphertext2.hex()}")

    cipher2_dec = DES3.new(key_16, DES3.MODE_ECB)
    decrypted2 = cipher2_dec.decrypt(ciphertext2)
    print(f"Deşifrə: {decrypted2}")

if DES_AVAILABLE:
    triple_des_demo()
````

## 🎯 Meet-in-the-middle hücumu (20 dəq)

<div style="background-color: #fff3e0; padding: 15px; border-radius: 10px; border-left: 5px solid #ff9800;">
<h4>🎯 Meet-in-the-middle hücumu</h4>
<p>İkiqat DES (2DES): $C = E_{K2}(E_{K1}(P))$</p>

<p><b>Meet-in-the-middle hücumu:</b></p>
<ol>
    <li>Bütün mümkün K1 açarları üçün $E_{K1}(P)$ hesabla və cədvələ yaz (həfizədə saxla)</li>
    <li>Bütün mümkün K2 açarları üçün $D_{K2}(C)$ hesabla</li>
    <li>Cədvəldə uyğun gələn dəyərləri tap ($E_{K1}(P) = D_{K2}(C)$)</li>
</ol>

<p><b>Mürəkkəblik:</b> $2^{56} + 2^{56} = 2^{57}$, $2^{112}$ deyil!</p>
<p><b>Zaman-Həfizə kompromisi:</b> Vaxtdan qənaət etmək üçün həfizə istifadə olunur.</p>
</div>

````python
def small_key_to_des_key(key_int, key_size=16):
    """
    Kiçik ölçülü açarı (məs. 12 və ya 16 bit) unikal DES açarına çevirir.
    Açar bitləri 56-bit DES açar fəzasına yerləşdirilir və hər bayt üçün odd parity qurulur.
    """
    if not (1 <= key_size <= 16):
        raise ValueError("key_size 1 ilə 16 arasında olmalıdır")
    if not (0 <= key_int < (1 << key_size)):
        raise ValueError("Açar dəyəri verilmiş bit ölçüsünə uyğun deyil")

    key56 = key_int << (56 - key_size)
    key_bytes = []

    for i in range(8):
        chunk = (key56 >> (7 * (7 - i))) & 0x7F  # 7 məlumat biti
        byte = chunk << 1                         # parity biti üçün yer saxla
        if bin(byte).count('1') % 2 == 0:        # odd parity
            byte |= 1
        key_bytes.append(byte)

    return bytes(key_bytes)


def double_des_encrypt_block(block_8, k1, k2, key_size=16):
    """Kiçik açarlardan istifadə edən 2DES: C = E_K2(E_K1(P))"""
    key1 = small_key_to_des_key(k1, key_size)
    key2 = small_key_to_des_key(k2, key_size)
    return des_encrypt_block(key2, des_encrypt_block(key1, block_8))


def meet_in_the_middle_attack_simulation(plaintext, ciphertext, key_size=12, verbose=True):
    """
    Həqiqi (amma kiçildilmiş açar fəzası ilə) 2DES üçün meet-in-the-middle hücumunun simulyasiyası.

    Parametrlər:
        plaintext: 8-bayt açıq mətn və ya 64-bit int
        ciphertext: 8-bayt şifrəli mətn və ya 64-bit int
        key_size: açar ölçüsü (bit), praktik nümayiş üçün adətən 8-16
        verbose: ekrana detallı məlumat çap etsinmi

    Qaytarır:
        matches: tapılan (K1, K2) cütləri
    """
    if not DES_AVAILABLE:
        raise RuntimeError("PyCryptodome lazımdır")

    if isinstance(plaintext, int):
        plaintext = plaintext.to_bytes(8, 'big')
    if isinstance(ciphertext, int):
        ciphertext = ciphertext.to_bytes(8, 'big')

    if len(plaintext) != 8 or len(ciphertext) != 8:
        raise ValueError("2DES nümayişi üçün plaintext və ciphertext 8 bayt olmalıdır")

    total_keys = 1 << key_size

    if verbose:
        print("\n" + "=" * 80)
        print("🎯 MEET-IN-THE-MIDDLE HÜCUMU SİMULYASİYASI")
        print("=" * 80)
        print(f"Açar ölçüsü: {key_size} bit")
        print(f"Mümkün açar sayı: {total_keys}")
        print(f"Açıq mətn: {plaintext.hex()}")
        print(f"Şifrəli mətn: {ciphertext.hex()}")
        print("\n1. CƏDVƏL QURULMASI (Forward):")

    forward_table = {}
    for k1 in range(total_keys):
        middle = des_encrypt_block(small_key_to_des_key(k1, key_size), plaintext)
        forward_table.setdefault(middle, []).append(k1)

        if verbose and k1 < 5:
            print(f"  K1={k1:04x} -> E_K1(P)={middle.hex()}")

    if verbose:
        print(f"  Cədvəldə {len(forward_table)} unikal orta dəyər var")
        print("\n2. DEŞİFRLƏMƏ (Backward):")

    matches = []
    for k2 in range(total_keys):
        middle = des_decrypt_block(small_key_to_des_key(k2, key_size), ciphertext)
        if middle in forward_table:
            for k1 in forward_table[middle]:
                matches.append((k1, k2))
                if verbose:
                    print(f"  ✅ UYĞUNLUQ! K1={k1:04x}, K2={k2:04x}, orta dəyər={middle.hex()}")

    if verbose:
        print("\n3. NƏTİCƏ:")
        print(f"  Tapılan uyğunluqlar: {len(matches)}")
        print(f"  Hücum mürəkkəbliyi: O(2^{key_size} + 2^{key_size}) = O(2^{key_size + 1})")
        print(f"  Brute-force mürəkkəbliyi: O(2^{2 * key_size})")
        if matches:
            print("\n  Uğurlu hücum! (K1, K2) cütlüyü tapıldı")

    return matches


# Test: həqiqi 2DES nümunəsi (kiçik açar fəzası ilə)
demo_plain = b'ABCDEFGH'
demo_k1 = 0x123
demo_k2 = 0x2AB
demo_cipher = double_des_encrypt_block(demo_plain, demo_k1, demo_k2, key_size=12)
matches = meet_in_the_middle_attack_simulation(demo_plain, demo_cipher, key_size=12)
print(f"\nƏsl açarlar: K1={demo_k1:03x}, K2={demo_k2:03x}")
print(f"Əsl açarlar tapıldı? {(demo_k1, demo_k2) in matches}")
````

Bu hissə notebookdakı əsas implementasiyaların düzgünlüyünü avtomatik yoxlayır.

````python
def validate_cipher_implementations():
    import secrets

    print("\n" + "=" * 80)
    print("✅ AVTOMATİK DOĞRULAMA")
    print("=" * 80)

    checks = []

    # 1. IP və FP-in tərs olması
    perm_ok = True
    test_values = [0x0123456789ABCDEF] + [int.from_bytes(secrets.token_bytes(8), 'big') for _ in range(25)]
    for value in test_values:
        if final_permutation(initial_permutation(value)) != value:
            perm_ok = False
            break
    checks.append(("IP/FP tərsliyi", perm_ok))

    # 2. Raund funksiyası üçün məlum DES nümunəsi
    round_ok = des_round_function(0xF0AAF0AA, 0x1B02EFFC7072) == 0x234AA9BB
    checks.append(("DES raund funksiyası", round_ok))

    des_available = bool(
        globals().get("DES_AVAILABLE", globals().get("DES") is not None and globals().get("DES3") is not None)
    )

    if des_available:
        # Qeyd: MODE_ECB real şifrələmə üçün təhlükəsiz deyil; burada PyCryptodome
        # yalnız tək bloklu DES/3DES primitivini test vektoru kimi yoxlamaq üçün istifadə olunur.

        # 3. Zəif açarlar
        weak_keys_hex = [
            "0101010101010101",
            "1F1F1F1F0E0E0E0E",
            "E0E0E0E0F1F1F1F1",
            "FEFEFEFEFEFEFEFE"
        ]
        weak_ok = True
        for key_hex in weak_keys_hex:
            key = bytes.fromhex(key_hex)
            p = b'Message1'
            c = DES.new(key, DES.MODE_ECB).encrypt(p)
            p2 = DES.new(key, DES.MODE_ECB).encrypt(c)
            if p2 != p:
                weak_ok = False
                break
        checks.append(("DES zəif açarları", weak_ok))

        # 4. 3DES EDE3 və EDE2 implementasiyası
        key_24 = DES3.adjust_key_parity(bytes.fromhex('0123456789ABCDEFFEDCBA98765432100011223344556677'))
        key_16 = DES3.adjust_key_parity(bytes.fromhex('0123456789ABCDEFFEDCBA9876543210'))
        k1_3, k2_3, k3_3 = key_24[:8], key_24[8:16], key_24[16:24]
        k1_2, k2_2 = key_16[:8], key_16[8:16]
        block = b'Message1'

        ede3_ok = triple_des_ede3_block(block, k1_3, k2_3, k3_3) == DES3.new(key_24, DES3.MODE_ECB).encrypt(block)
        ede2_ok = triple_des_ede2_block(block, k1_2, k2_2) == DES3.new(key_16, DES3.MODE_ECB).encrypt(block)
        checks.append(("3DES EDE3", ede3_ok))
        checks.append(("3DES EDE2", ede2_ok))

        # 5. Kiçik açar fəzası ilə 2DES MITM
        demo_plain = b'ABCDEFGH'
        demo_k1 = 0x123
        demo_k2 = 0x2AB
        demo_cipher = double_des_encrypt_block(demo_plain, demo_k1, demo_k2, key_size=12)
        matches = meet_in_the_middle_attack_simulation(demo_plain, demo_cipher, key_size=12, verbose=False)
        mitm_ok = (demo_k1, demo_k2) in matches
        checks.append(("2DES MITM (kiçik açar fəzası)", mitm_ok))
    else:
        checks.append(("DES zəif açarları (PyCryptodome yoxdur — ötürüldü)", None))
        checks.append(("3DES EDE3/EDE2 (PyCryptodome yoxdur — ötürüldü)", None))
        checks.append(("2DES MITM (PyCryptodome yoxdur — ötürüldü)", None))

    all_ok = True
    for name, ok in checks:
        if ok is None:
            print(f"⚠️ {name}")
            continue
        print(f"{'✅' if ok else '❌'} {name}")
        all_ok = all_ok and ok

    print("\nYekun:")
    print("✅ Bütün icra olunan əsas implementasiyalar yoxlamadan keçdi" if all_ok else "❌ Bəzi implementasiyalarda problem var")

validate_cipher_implementations()
````
