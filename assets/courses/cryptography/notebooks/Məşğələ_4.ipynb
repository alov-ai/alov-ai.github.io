## 📖 Xatırlatma: Əsas ehtimal anlayışları (10 dəq)

<div style="background-color: #f0f8ff; padding: 15px; border-radius: 10px; border-left: 5px solid #4682b4;">
<h4>🔑 Kriptoqrafiyada ehtimalın rolu</h4>
<p>Kriptoqrafiyada təhlükəsizlik <b>ehtimallarla ifadə olunur</b>. Bir sistemin "təhlükəsiz" olması o deməkdir ki, onu sındırmaq ehtimalı çox kiçikdir - adətən $2^{-128}$ və ya daha az. Formal olaraq bu ehtimal hücumçunun resurslarına və təhlükəsizlik oyunundakı üstünlüyünə görə ölçülür.</p>
<p>Bu məşğələdə kriptoqrafiya üçün vacib olan ehtimal anlayışlarını Python-da simulyasiya edəcəyik.</p>
</div>

**Əsas anlayışlar:**

| Anlayış      | Düstur      | İzah |
|--------------|-------------|------|
| **Ehtimal kütlə funksiyası (PMF)** | $$ p(x) = P(X = x)$$ | Təsadüfi dəyişənin müəyyən qiymət alması ehtimalı |
| **Gözlənilən dəyər** | $E[X] = \sum x \cdot p(x)$ | Təsadüfi dəyişənin orta qiyməti |
| **Variasiya** | $Var(X) = E[(X - E[X])^2]$ | Qiymətlərin ortalama ətrafında yayılması |
| **Şərti ehtimal** | $P(A\mid B) = \frac{P(A \cap B)}{P(B)}$ | B hadisəsi baş verdikdə A-nın ehtimalı |
| **Bayes teoremi** | $P(B_i\mid A) = \frac{P(A\mid B_i)P(B_i)}{\sum_j P(A\mid B_j)P(B_j)}$ | Yeni məlumatla ehtimalların yenilənməsi |


## 🎲 Bircins paylanmanın simulyasiyası (15 dəq)

Bircins paylanmada bütün mümkün qiymətlər **bərabər ehtimala** malikdir. Bu, kriptoqrafiyada açar generasiyasında istifadə olunur - hər bir mümkün açar bərabər ehtimalla seçilməlidir.

> **Qeyd:** Bu notebookda `random` modulu yalnız simulyasiya və tədris nümunələri üçündür; real kriptoqrafik açarlar, sirlər və nonce-lar üçün `secrets` modulundan və ya əməliyyat sisteminin CSPRNG mənbəyindən istifadə edin.

````python
def uniform_distribution(n, trials=10000):
    """
    Bircins paylanmanı simulyasiya edir

    Parametrlər:
        n: mümkün qiymətlərin sayı (1-dən n-ə qədər)
        trials: sınaq sayı

    Qaytarır:
        results: təsadüfi ədədlər siyahısı
        freq: hər qiymətin tezliyi
    """
    results = []
    for _ in range(trials):
        # random.randint(1, n) - 1-dən n-ə qədər bircins təsadüfi ədəd
        value = random.randint(1, n)
        results.append(value)

    # Nəticələrin tezliyini hesabla
    freq = Counter(results)

    # Gözlənilən dəyəri hesabla
    expected = (n + 1) / 2
    actual_mean = sum(results) / trials

    # Variasiyanı hesabla
    expected_var = (n**2 - 1) / 12
    variance = sum((x - actual_mean)**2 for x in results) / trials

    print(f"🎲 Bircins paylanma (1-{n}): {trials} sınaq")
    print(f"   Gözlənilən ortalama: {expected:.2f}")
    print(f"   Faktiki ortalama: {actual_mean:.2f}")
    print(f"   Gözlənilən variasiya: {expected_var:.2f}")
    print(f"   Faktiki variasiya: {variance:.2f}")

    # Hər bir qiymətin tezliyini göstər
    print("\n   Qiymət | Tezlik | Gözlənilən")
    print("   " + "-" * 30)
    for i in range(1, n+1):
        count = freq.get(i, 0)
        expected_count = trials / n
        bar = '█' * int(30 * count / trials)
        print(f"     {i:2d}   | {count:5d} | {expected_count:5.1f} {bar}")

    return results, freq

# Test: 6 üzlü zar atma simulyasiyası
results_uniform, freq_uniform = uniform_distribution(6, 10000)
````

### 5.2 Bircins paylanmanın qrafiki

````python
def plot_uniform_distribution(results, n):
    """
    Bircins paylanmanın histogramını çəkir
    """
    if not HAS_MATPLOTLIB:
        print("⚠️ matplotlib yüklü deyil, qrafik çəkilə bilməz")
        return

    plt.figure(figsize=(10, 6))

    # Histogram
    counts, bins, patches = plt.hist(results, bins=range(1, n+2),
                                      align='left', rwidth=0.8,
                                      alpha=0.7, density=True,
                                      edgecolor='black')

    # Nəzəri bircins xətt
    plt.axhline(y=1/n, color='r', linestyle='--', linewidth=2,
                label=f'Nəzəri ehtimal (1/{n} = {1/n:.3f})')

    plt.xlabel('Qiymət')
    plt.ylabel('Ehtimal')
    plt.title(f'🎲 Bircins Paylanma (1-{n}) - {len(results)} sınaq')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.xticks(range(1, n+1))
    plt.show()

# Qrafiki çəkək
if HAS_MATPLOTLIB:
    plot_uniform_distribution(results_uniform, 6)
````

## 📊 Binomial paylanmanın simulyasiyası (20 dəq)

Binomial paylanma **müstəqil sınaqlarda** müəyyən sayda uğur qazanma ehtimalını modelləşdirir.

**Kriptoqrafiyada tətbiqi:**
* Hücumçunun müəyyən sayda cəhddən sonra uğur qazanma ehtimalı
* Səhv ehtimallarının hesablanması
* Şifrə sındırma cəhdlərinin modelləşdirilməsi

````python
def binomial_distribution(n, p, trials=10000):
    """
    Binomial paylanmanı simulyasiya edir

    Parametrlər:
        n: hər simulyasiyada sınaq sayı
        p: hər sınaqda uğur ehtimalı
        trials: simulyasiya sayı

    Qaytarır:
        results: hər simulyasiyada uğur sayı
        freq: hər uğur sayının tezliyi
    """
    results = []

    for _ in range(trials):
        successes = 0
        for _ in range(n):
            if random.random() < p:
                successes += 1
        results.append(successes)

    # Nəticələrin tezliyini hesabla
    freq = Counter(results)

    # Gözlənilən dəyəri hesabla
    expected = n * p
    actual_mean = sum(results) / trials

    # Variasiyanı hesabla
    expected_var = n * p * (1 - p)
    variance = sum((x - actual_mean)**2 for x in results) / trials

    print(f"📊 Binomial paylanma: n={n}, p={p}, {trials} simulyasiya")
    print(f"   Gözlənilən ortalama: {expected:.2f}")
    print(f"   Faktiki ortalama: {actual_mean:.2f}")
    print(f"   Gözlənilən variasiya: {expected_var:.2f}")
    print(f"   Faktiki variasiya: {variance:.2f}")

    # Ən çox rast gəlinən dəyərləri göstər
    print("\n   Ən çox rast gəlinən dəyərlər:")
    common = freq.most_common(5)
    for k, count in common:
        prob = count / trials
        bar = '█' * int(50 * prob)
        print(f"     k={k:2d}: {count:4d} dəfə ({prob:.3f}) {bar}")

    return results, freq

def binomial_probability(k, n, p):
    """
    Binomial ehtimalı hesablayır: P(X = k)
    """
    from math import comb
    return comb(n, k) * (p**k) * ((1-p)**(n-k))

# Test: 10 sikkə atışı, 5 yazı gəlmə ehtimalı
n_binom, p_binom = 10, 0.5
prob_theo = binomial_probability(5, n_binom, p_binom)
print(f"\n🧪 P(X=5) üçün nəzəri ehtimal: {prob_theo:.4f}")

results_binom, freq_binom = binomial_distribution(n_binom, p_binom, 10000)
actual_prob = freq_binom.get(5, 0) / 10000
print(f"   Simulyasiya ehtimalı: {actual_prob:.4f}")
````

### 6.2 Binomial paylanmanın qrafiki

````python
def plot_binomial_distribution(results, n, p):
    """
    Binomial paylanmanın histogramını çəkir
    """
    if not HAS_MATPLOTLIB:
        print("⚠️ matplotlib yüklü deyil, qrafik çəkilə bilməz")
        return

    plt.figure(figsize=(12, 6))

    # Simulyasiya nəticələri
    counts, bins, patches = plt.hist(results, bins=range(n+2),
                                      align='left', rwidth=0.8,
                                      alpha=0.7, density=True,
                                      label='Simulyasiya',
                                      edgecolor='black')

    # Nəzəri ehtimallar
    x = range(n+1)
    theo_probs = [binomial_probability(k, n, p) for k in x]
    plt.plot(x, theo_probs, 'ro-', label='Nəzəri',
             markersize=6, linewidth=2)

    plt.xlabel('Uğur sayı (k)')
    plt.ylabel('Ehtimal')
    plt.title(f'📊 Binomial Paylanma (n={n}, p={p})')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.xticks(x)
    plt.show()

# Qrafiki çəkək
if HAS_MATPLOTLIB:
    plot_binomial_distribution(results_binom, n_binom, p_binom)
````

## 🎯 Puasson paylanmasının simulyasiyası (15 dəq)

Puasson paylanması **nadir hadisələrin** sayını modelləşdirir. Binomial paylanmanın limit yaxınlaşmasıdır: $n$ böyük, $p$ kiçik və $np = \lambda$ sabit olduqda $\mathrm{Bin}(n,p) \approx \mathrm{Pois}(\lambda)$.

**Kriptoqrafiyada tətbiqi:**
* Kriptosistemlərdə sındırılma ehtimalları
* Səhv hadisələrinin sayı
* Hücum cəhdlərinin modelləşdirilməsi


````python
def poisson_distribution(lambda_param, trials=10000):
    """
    Puasson paylanmasını simulyasiya edir

    Parametrlər:
        lambda_param: hadisələrin orta sayı (λ)
        trials: simulyasiya sayı

    Qaytarır:
        results: hər simulyasiyada hadisə sayı
        freq: hər hadisə sayının tezliyi
    """
    if lambda_param <= 0:
        raise ValueError("λ must be positive")

    results = []

    for _ in range(trials):
        # Puasson təsadüfi dəyişəni generasiya etmək üçün
        # eksponensial paylanmadan istifadə
        t = 0
        count = 0
        while t < 1:
            # Eksponensial paylanmış təsadüfi ədəd
            u = random.random()
            t += -math.log(u) / lambda_param
            if t < 1:
                count += 1
        results.append(count)

    # Nəticələrin tezliyini hesabla
    freq = Counter(results)

    # Gözlənilən dəyəri hesabla
    expected = lambda_param
    actual_mean = sum(results) / trials

    # Variasiyanı hesabla
    expected_var = lambda_param
    variance = sum((x - actual_mean)**2 for x in results) / trials

    print(f"🎯 Puasson paylanması: λ={lambda_param}, {trials} simulyasiya")
    print(f"   Gözlənilən ortalama: {expected:.2f}")
    print(f"   Faktiki ortalama: {actual_mean:.2f}")
    print(f"   Gözlənilən variasiya: {expected_var:.2f}")
    print(f"   Faktiki variasiya: {variance:.2f}")

    # Ən çox rast gəlinən dəyərləri göstər
    print("\n   Ən çox rast gəlinən dəyərlər:")
    common = freq.most_common(5)
    for k, count in common:
        prob = count / trials
        bar = '█' * int(50 * prob)
        print(f"     k={k:2d}: {count:4d} dəfə ({prob:.3f}) {bar}")

    return results, freq

def poisson_probability(k, lambda_param):
    """
    Puasson ehtimalını hesablayır: P(X = k)
    """
    from math import factorial
    return (math.exp(-lambda_param) * lambda_param**k) / factorial(k)

# Test: λ=3 üçün Puasson paylanması
results_poisson, freq_poisson = poisson_distribution(3, 10000)
````

### 7.2 Puasson və binomial müqayisəsi

## 🧠 Bayes teoreminin tətbiqi (15 dəq)

Bayes teoremi **yeni məlumat əsasında ehtimalları yeniləməyə** imkan verir. Kriptoqrafiyada bu, hücumların analizində və sistemlərin təhlükəsizlik qiymətləndirilməsində istifadə olunur.

$$P(B_i|A) = \frac{P(A|B_i)P(B_i)}{\sum_j P(A|B_j)P(B_j)}$$

````python
def bayes_theorem(prior_probs, likelihoods, evidence_name="Müşahidə"):
    """
    Bayes teoremini tətbiq edir

    Parametrlər:
        prior_probs: hipotezlərin aprior ehtimalları [P(H1), P(H2), ...]
        likelihoods: hər hipotez üçün P(evidence|hipotez) [P(E|H1), P(E|H2), ...]
        evidence_name: müşahidə edilən hadisənin adı

    Qaytarır:
        posterior_probs: aposterior ehtimallar [P(H1|E), P(H2|E), ...]
    """
    if len(prior_probs) != len(likelihoods):
        raise ValueError("prior_probs və likelihoods eyni uzunluqda olmalıdır.")
    if len(prior_probs) == 0:
        raise ValueError("Ən azı bir hipotez verilməlidir.")
    if not all(0 <= p <= 1 for p in prior_probs):
        raise ValueError("Bütün aprior ehtimallar [0, 1] intervalında olmalıdır.")
    if not all(0 <= l <= 1 for l in likelihoods):
        raise ValueError("Bütün şərti ehtimallar P(E|H) [0, 1] intervalında olmalıdır.")
    if abs(sum(prior_probs) - 1.0) > 1e-9:
        raise ValueError("Aprior ehtimalların cəmi 1 olmalıdır.")

    n = len(prior_probs)

    # Tam ehtimalı hesabla: P(evidence)
    total_prob = sum(prior_probs[i] * likelihoods[i] for i in range(n))
    if total_prob <= 0:
        raise ValueError("P(evidence) sıfırdan böyük olmalıdır.")

    # Aposterior ehtimalları hesabla
    posterior_probs = []
    for i in range(n):
        posterior = (prior_probs[i] * likelihoods[i]) / total_prob
        posterior_probs.append(posterior)

    # Nəticələri çap et
    print("=" * 70)
    print(f"🧠 BAYES TEOREMİ - Müşahidə: {evidence_name}")
    print("=" * 70)
    print("Hipotez | Aprior P | P(E|H) | Aposterior P")
    print("-" * 50)
    for i in range(n):
        print(f"   H{i+1}   |  {prior_probs[i]:.3f}   |  {likelihoods[i]:.3f}   |    {posterior_probs[i]:.3f}")
    print("=" * 70)

    return posterior_probs

# Nümunə: Şifrələmə alqoritmi təhlili
print("\n🔐 ŞİFRLƏMƏ ALQORİTMİ TƏHLİLİ")
prior = [0.6, 0.4]  # H1: təhlükəsiz, H2: təhlükəsiz deyil
likelihood = [0.01, 0.1]  # Hücumun uğur ehtimalı
posterior = bayes_theorem(prior, likelihood, "Hücum uğurlu")
````

   * Bir sistem sındırıldısa, bunun Alqoritm A ilə olma ehtimalı nədir?

````python
# Çalışma 4 - Cavablar

print("📝 ÇALIŞMA 4 CAVABLARI")
print("=" * 70)

# 1. Şifrələmə sistemi təhlili
print("\n1. ŞİFRLƏMƏ SİSTEMİ TƏHLİLİ:")
prior1 = [0.7, 0.3]  # H1: təhlükəsiz, H2: zəif
likelihood1 = [0.02, 0.15]  # sındırılma ehtimalları
posterior1 = bayes_theorem(prior1, likelihood1, "Mesaj sındırıldı")
print(f"\n   P(təhlükəsiz | sındırıldı) = {posterior1[0]:.4f}")

# 2. Alqoritm seçimi problemi
print("\n2. ALQORİTM SEÇİMİ PROBLEMİ:")
prior2 = [0.8, 0.2]  # H1: Alqoritm A, H2: Alqoritm B
likelihood2 = [0.001, 0.05]  # sındırılma ehtimalları
posterior2 = bayes_theorem(prior2, likelihood2, "Sistem sındırıldı")
print(f"\n   P(Alqoritm A | sındırıldı) = {posterior2[0]:.4f}")

# 3. Öz problemi
print("\n3. ÖZ PROBLEMİ (Açar sındırma):")
print("   Problem: İki tip açar var - güclü (60%) və zəif (40%)")
print("   Güclü açarların sındırılma ehtimalı 0.1%, zəiflərin 5%")
print("   Bir açar sındırıldısa, onun zəif olma ehtimalı nədir?")

prior3 = [0.6, 0.4]  # H1: güclü, H2: zəif
likelihood3 = [0.001, 0.05]
posterior3 = bayes_theorem(prior3, likelihood3, "Açar sındırıldı")
print(f"\n   P(zəif | sındırıldı) = {posterior3[1]:.4f}")
````

## 🎂 Ad günü paradoksu (20 dəq)

<div style="background-color: #fff3e0; padding: 15px; border-radius: 10px; border-left: 5px solid #ff9800;">
<h4>📜 Ad günü paradoksu nədir?</h4>
<p>Ad günü paradoksuna görə, 23 nəfərlik bir qrupda ən azı iki nəfərin eyni ad gününə sahib olma ehtimalı <b>50%-dən çoxdur</b>. Bu, intuitiv olaraq gözləniləndən çoxdur (insanlar adətən 183 nəfər lazım olduğunu düşünür).</p>

<p><b>Kriptoqrafiyada əhəmiyyəti:</b> Ad günü paradoksu <b>heş funksiyalarında toqquşma tapmaq</b> üçün istifadə olunan "ad günü hücumu"nun əsasını təşkil edir. $N$ mümkün heş dəyəri üçün toqquşma tapmaq mürəkkəbliyi $N$ yox, təxminən $\sqrt{N}$-dir.</p>
</div>

````python
def birthday_paradox_simulation(n, days=365, trials=10000):
    """
    Ad günü paradoksunu simulyasiya edir

    Parametrlər:
        n: otaqdakı insan sayı
        days: ildəki gün sayı (365)
        trials: simulyasiya sayı

    Qaytarır:
        probability: ən azı bir toqquşma ehtimalı
    """
    collisions = 0

    for _ in range(trials):
        # Təsadüfi ad günləri generasiya et
        birthdays = [random.randint(1, days) for _ in range(n)]

        # Təkrarlanan ad günləri varmı?
        if len(set(birthdays)) < n:
            collisions += 1

    probability = collisions / trials

    # Nəzəri ehtimalı hesabla (yaxınlaşdırma)
    theoretical = 1 - math.exp(-n**2 / (2 * days))

    # Dəqiq nəzəri ehtimal (kiçik n üçün)
    exact = 1.0
    for i in range(n):
        exact *= (days - i) / days
    exact = 1 - exact

    print(f"   n = {n:2d} nəfər: Simulyasiya = {probability:.4f}, "
          f"Nəzəri (dəqiq) = {exact:.4f}, Yaxınlaşdırma = {theoretical:.4f}")

    return probability

print("🎂 AD GÜNÜ PARADOKSU SİMULYASİYASI")
print("-" * 70)
for n in [10, 20, 23, 30, 40, 50, 60]:
    birthday_paradox_simulation(n, 365, 5000)
````

### 9.2 Ad günü paradoksu qrafiki

````python
def plot_birthday_paradox(max_n=100, days=365, trials=1000):
    """
    Ad günü paradoksunun qrafikini çəkir
    """
    if not HAS_MATPLOTLIB:
        print("⚠️ matplotlib yüklü deyil, qrafik çəkilə bilməz")
        return

    n_values = list(range(1, max_n + 1))
    sim_probs = []
    exact_probs = []

    for n in n_values:
        # Simulyasiya
        sim_prob = birthday_paradox_simulation(n, days, trials)
        sim_probs.append(sim_prob)

        # Dəqiq nəzəri
        exact = 1.0
        for i in range(n):
            exact *= (days - i) / days
        exact_probs.append(1 - exact)

    plt.figure(figsize=(14, 8))
    plt.plot(n_values, sim_probs, 'bo-', label='Simulyasiya',
             markersize=4, linewidth=1, alpha=0.7)
    plt.plot(n_values, exact_probs, 'r-', label='Nəzəri (dəqiq)',
             linewidth=2)

    # Əhəmiyyətli nöqtələr
    plt.axhline(y=0.5, color='g', linestyle='--', linewidth=2,
                label='p=0.5')
    plt.axvline(x=23, color='orange', linestyle='--', linewidth=2,
                label='n=23')
    plt.axvline(x=57, color='purple', linestyle='--', linewidth=2,
                label='n=57 (p=0.99)')

    plt.xlabel('Otaqdakı insan sayı (n)')
    plt.ylabel('Ən azı bir toqquşma ehtimalı')
    plt.title(f'🎂 Ad Günü Paradoksu (il = {days} gün)')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.xlim(0, max_n)
    plt.ylim(0, 1.05)
    plt.show()

# Qrafiki çəkək
if HAS_MATPLOTLIB:
    plot_birthday_paradox(80, 365, 1000)
````

### 9.3 Ad günü hücumu simulyasiyası

````python
def birthday_attack_simulation(m, trials=1000):
    """
    Ad günü hücumunu simulyasiya edir

    Parametrlər:
        m: heş dəyərinin bit uzunluğu
        trials: simulyasiya sayı
    """
    N = 2**m  # Mümkün heş dəyərlərinin sayı

    # Müxtəlif q dəyərləri üçün toqquşma ehtimalı
    q_values = [int(2**(m/2) * f) for f in [0.5, 1, 1.5, 2, 3]]

    print("=" * 80)
    print(f"🎯 AD GÜNÜ HÜCUMU SİMULYASİYASI (m={m} bit, N=2^{m} ≈ {N:.0e})")
    print("=" * 80)
    print(f"{'q (sınaq sayı)':<15} | {'Simulyasiya P':<15} | {'Nəzəri P':<15} | {'Fərq':<10}")
    print("-" * 65)

    for q in q_values:
        if q > N:
            continue

        collisions = 0
        for _ in range(trials):
            # q sayda təsadüfi heş dəyəri generasiya et
            hashes = [random.randint(0, N-1) for _ in range(q)]

            # Toqquşma varmı?
            if len(set(hashes)) < q:
                collisions += 1

        sim_prob = collisions / trials
        theo_prob = 1 - math.exp(-q**2 / (2 * N))

        print(f"{q:<15} | {sim_prob:.6f}      | {theo_prob:.6f}      | {abs(sim_prob-theo_prob):.6f}")

    # Nəzəri qiymətləndirmə
    print("\n" + "-" * 65)
    print(f"📌 50% ehtimal üçün tələb olunan q ≈ √(2N·ln(2)) ≈ {math.sqrt(2*N*math.log(2)):.0f}")
    print(f"📌  √N ≈ {math.sqrt(N):.0f}")
    return

# Test: müxtəlif bit uzunluqları üçün
birthday_attack_simulation(8, 2000)   # 8-bit heş
birthday_attack_simulation(12, 2000)  # 12-bit heş
birthday_attack_simulation(16, 2000)  # 16-bit heş
````
