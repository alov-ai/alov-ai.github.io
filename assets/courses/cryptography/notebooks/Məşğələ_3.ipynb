## 🔤 Vijener şifrəsinin implementasiyası (30 dəq)

### 5.1 Köməkçi funksiyalar

Əvvəlcə hərf-rəqəm çevirmə funksiyalarını yazaq:

````python
import string


def letter_to_num(letter):
    """Hərfi rəqəmə çevirir (A=0, B=1, ..., Z=25)"""
    return ord(letter.upper()) - ord('A')


def num_to_letter(num):
    """Rəqəmi hərfə çevirir (0=A, 1=B, ..., 25=Z)"""
    return chr(num + ord('A'))


def clean_text(text):
    """Mətndən yalnız A-Z hərflərini götürür və böyüdür"""
    return ''.join([c.upper() for c in text if c.upper() in string.ascii_uppercase])


# Test
print(f"'A' → {letter_to_num('A')}")
print(f"0 → '{num_to_letter(0)}'")
print(f"Təmizlənmiş mətn: {clean_text('Salam, Dünya! 123')}")
````

### 5.2 Vijener şifrləmə funksiyası

Vijener şifrəsinin riyazi modeli:

* **Şifrləmə:** $C_i = (P_i + K_{i \mod m}) \mod 26$
* **Deşifrləmə:** $P_i = (C_i - K_{i \mod m}) \mod 26$

Burada $P_i$ açıq mətnin i-ci hərfi, $C_i$ şifrəli mətnin i-ci hərfi, $K_j$ açarın j-ci hərfi, $m$ isə açarın uzunluğudur.

````python
def vigenere_encrypt(plaintext, key):
    """
    Vijener şifrəsi ilə mətni şifrləyir
    C_i = (P_i + K_{i mod m}) mod 26

    Parametrlər:
        plaintext: şifrlənəcək mətn
        key: açar söz

    Qaytarır:
        şifrələnmiş mətn (yalnız böyük hərflər)
    """
    plaintext = clean_text(plaintext)
    key = clean_text(key)

    if not key:
        raise ValueError("❌ Açar boş ola bilməz")

    ciphertext = []
    key_len = len(key)

    for i, char in enumerate(plaintext):
        p = letter_to_num(char)
        k = letter_to_num(key[i % key_len])
        c = (p + k) % 26
        ciphertext.append(num_to_letter(c))

    return ''.join(ciphertext)

def vigenere_decrypt(ciphertext, key):
    """
    Vijener şifrəsi ilə şifrələnmiş mətni deşifrləyir
    P_i = (C_i - K_{i mod m}) mod 26
    """
    ciphertext = clean_text(ciphertext)
    key = clean_text(key)

    if not key:
        raise ValueError("❌ Açar boş ola bilməz")

    plaintext = []
    key_len = len(key)

    for i, char in enumerate(ciphertext):
        c = letter_to_num(char)
        k = letter_to_num(key[i % key_len])
        p = (c - k) % 26
        plaintext.append(num_to_letter(p))

    return ''.join(plaintext)
````

### 5.3 Vijener şifrəsinin test edilməsi

````python
# Test 1: Sadə nümunə
plaintext = "KRIPTOQRAFIYA KURSU COX MARAQLIDIR"
key = "ACAR"

print("📨 VİJENER ŞİFRƏ TESTİ")
print("=" * 60)
print(f"Orijinal mətn: {plaintext}")
print(f"Təmizlənmiş mətn: {clean_text(plaintext)}")
print(f"Açar: {key}")

encrypted = vigenere_encrypt(plaintext, key)
decrypted = vigenere_decrypt(encrypted, key)

print(f"\n🔒 Şifrələnmiş: {encrypted}")
print(f"🔓 Deşifrələnmiş: {decrypted}")
print(f"\n✓ Uğurlu? {clean_text(plaintext) == decrypted}")

# Test 2: Başqa nümunə
print("\n" + "-" * 60)
plaintext2 = "SALAM DUNYA"
key2 = "GIZLI"
encrypted2 = vigenere_encrypt(plaintext2, key2)
decrypted2 = vigenere_decrypt(encrypted2, key2)

print(f"Orijinal: {plaintext2}")
print(f"Açar: {key2}")
print(f"Şifrəli: {encrypted2}")
print(f"Deşifrə: {decrypted2}")
print(f"Uğurlu? {clean_text(plaintext2) == decrypted2}")
````

## 🔍 Kasiski testi ilə açar uzunluğunun tapılması (25 dəq)

### 6.1 Tarixi fon

<div style="background-color: #fff3e0; padding: 15px; border-radius: 10px; border-left: 5px solid #ff9800;">
<h4>📜 Fridrix Kasiski (1863)</h4>
<p>1863-cü ildə Prussiya ordusunun zabiti <b>Fridrix Kasiski</b> Vijener şifrəsini sındırmaq üçün bir üsul kəşf etdi. Onun metodu şifrəli mətndə təkrarlanan qrupları axtarmağa əsaslanırdı.</p>
<p><b>Əsas ideya:</b> Əgər eyni açıq mətn hissələri eyni açar hissələri ilə şifrlənirsə, şifrəli mətndə təkrarlanan qruplar yarana bilər. Bu təkrarlar arasındakı məsafələrin ən böyük ortaq böləni və ya onun bölənləri açar uzunluğu üçün <b>namizədlər</b> verir.</p>
<p><b>Qeyd:</b> Kasiski testi açar uzunluğunu birbaşa sübut etmir; alınan namizədləri adətən təsadüfi uyğunluqlar, bölənlər və qatlar arasından ayırmaq üçün IC (uyğunluq indeksi), tezlik analizi və ya uğurlu deşifrələmə ilə yoxlamaq lazımdır.</p>
</div>

### 6.2 Təkrarlanan qrupların tapılması

````python
def find_factors(n):
    """n ədədinin bölənlərini tapır (1 istisna)"""
    factors = [n] if n >= 2 else []
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            factors.append(i)
            if i != n // i:
                factors.append(n // i)
    return sorted(factors)

def find_repeating_sequences(ciphertext, min_length=3):
    """
    Şifrəli mətndə təkrarlanan ardıcıllıqları tapır

    Parametrlər:
        ciphertext: şifrəli mətn
        min_length: minimum ardıcıllıq uzunluğu

    Qaytarır:
        sequences: təkrarlanan ardıcıllıqlar haqqında məlumat
    """
    ciphertext = clean_text(ciphertext)
    sequences = {}

    # Müxtəlif uzunluqlu ardıcıllıqları axtar
    for length in range(min_length, min(7, len(ciphertext) // 2) + 1):
        for i in range(len(ciphertext) - length):
            seq = ciphertext[i:i+length]

            # Ardıcıllığın daha sonra təkrar olub-olmadığını yoxla
            for j in range(i + length, len(ciphertext) - length + 1):
                if ciphertext[j:j+length] == seq:
                    distance = j - i

                    if seq not in sequences:
                        sequences[seq] = []

                    sequences[seq].append({
                        'position1': i,
                        'position2': j,
                        'distance': distance,
                        'factors': find_factors(distance)
                    })

    return sequences
````

### 6.3 Açar uzunluğu ehtimallarının hesablanması

````python
def kasiski_analysis(ciphertext, min_length=3):
    """
    Kasiski testi ilə açar uzunluğunu təxmin edir

    Parametrlər:
        ciphertext: şifrəli mətn
        min_length: minimum ardıcıllıq uzunluğu

    Qaytarır:
        best_guess: ən çox ehtimal edilən açar uzunluğu
    """
    sequences = find_repeating_sequences(ciphertext, min_length)

    # Bütün bölənləri topla
    all_factors = []
    for seq, occurrences in sequences.items():
        for occ in occurrences:
            all_factors.extend(occ['factors'])

    # Bölənlərin tezliyini hesabla
    factor_counts = Counter(all_factors)

    print("=" * 70)
    print("🔍 KASISKI TESTİ NƏTİCƏLƏRİ")
    print("=" * 70)

    if not sequences:
        print("⚠️ Heç bir təkrarlanan ardıcıllıq tapılmadı.")
        print("Daha qısa ardıcıllıqları axtarmağa çalışın.")
        return None

    for seq, occurrences in sequences.items():
        print(f"\n'{seq}' ardıcıllığı {len(occurrences)} dəfə təkrarlanır:")
        for occ in occurrences:
            print(f"  📍 {occ['position1']}-{occ['position2']}: "
                  f"məsafə={occ['distance']}, bölənlər={occ['factors']}")

    print("\n" + "-" * 70)
    print("📊 BÖLƏNLƏRİN TEZLİYİ:")
    print("-" * 70)

    # Tezliyə görə sırala
    sorted_factors = sorted(factor_counts.items(), key=lambda x: x[1], reverse=True)

    for factor, count in sorted_factors:
        print(f"  {factor:3d}: {count} dəfə")

    # Ən çox təkrarlanan bölən açar uzunluğu ola bilər
    if sorted_factors:
        best_guess = sorted_factors[0][0]
        print("\n" + "=" * 70)
        print(f"🎯 TƏXMİN EDİLƏN AÇAR UZUNLUĞU: {best_guess}")
        print("=" * 70)
        return best_guess

    return None

# Test nümunəsi
test_cipher = vigenere_encrypt("BU BIR TEST MESAJIDIR", "KEY")
print(f"Test şifrəli mətn: {test_cipher}")
kasiski_analysis(test_cipher)
````

### 6.4 Uyğunluq indeksi (Index of Coincidence)

Uyğunluq indeksi mətnin təsadüfi olub-olmadığını ölçən statistik bir göstəricidir:

$$IC = \frac{\sum_{i=1}^{26} n_i (n_i - 1)}{N (N - 1)}$$

Burada:
* $n_i$ - i-ci hərfin mətndə sayı
* $N$ - mətndəki ümumi hərf sayı

**İngilis dili üçün:**
* Normal mətn üçün IC ≈ 0.0667
* Təsadüfi mətn üçün IC ≈ 0.0385

Vijener şifrəsində düzgün period tapıldıqda, hər sütunun IC-si ingilis dilinə yaxın olur.

````python
def calculate_ic(text):
    """
    Uyğunluq indeksini hesablayır
    IC = Σ(n_i * (n_i - 1)) / (N * (N - 1))
    """
    text = clean_text(text)
    N = len(text)

    if N < 2:
        return 0

    # Hərflərin tezliyini hesabla
    freq = Counter(text)

    # IC hesabla
    ic_sum = sum(count * (count - 1) for count in freq.values())
    ic = ic_sum / (N * (N - 1))

    return ic

def ic_for_different_periods(ciphertext, max_period=20, min_column_length=5):
    """
    Müxtəlif periodlar üçün orta uyğunluq indeksini hesablayır.

    Qeyd: IC yalnız diaqnostik göstəricidir. Qısa sütunlar və həqiqi
    periodun qatları süni olaraq yüksək nəticə verə bilər; namizədlər
    deşifrələmə cəhdi ilə yoxlanmalıdır.

    Parametrlər:
        ciphertext: şifrəli mətn
        max_period: maksimum sınanacaq period
        min_column_length: IC hesablamaq üçün hər sütunun minimum uzunluğu

    Qaytarır:
        results: IC-ə görə azalan sıralanmış (period, orta_IC) cütlərinin siyahısı
    """
    ciphertext = clean_text(ciphertext)
    min_column_length = max(2, min_column_length)
    results = []
    skipped_periods = []

    for period in range(1, min(max_period, len(ciphertext)) + 1):
        # Mətni period uzunluğunda sütunlara böl
        columns = [''] * period
        for i, char in enumerate(ciphertext):
            columns[i % period] += char

        column_lengths = [len(col) for col in columns]
        min_len = min(column_lengths) if column_lengths else 0

        # Çox qısa sütunlarda IC etibarsız ola bilər
        if min_len < min_column_length:
            skipped_periods.append((period, min_len))
            continue

        # Hər sütun üçün IC hesabla
        ics = [calculate_ic(col) for col in columns]
        avg_ic = sum(ics) / len(ics)
        results.append((period, avg_ic, min_len))

    ranked_results = sorted(results, key=lambda x: x[1], reverse=True)

    print("=" * 70)
    print("📊 UYĞUNLUQ İNDEKSİ ANALİZİ — NAMİZƏDLƏRİN SIRALANMASI")
    print("=" * 70)
    print("Sıra | Period | Orta IC | Min. sütun | Qiymətləndirmə")
    print("-" * 65)

    for rank, (period, avg_ic, min_len) in enumerate(ranked_results, start=1):
        # Qiymətləndirmə
        if avg_ic > 0.06:
            rating = "✅ İngilis dilinə yaxın"
        elif avg_ic > 0.05:
            rating = "⚠️ Şübhəli"
        else:
            rating = "❌ Təsadüfi mətn"

        print(f" {rank:3d} | {period:6d} | {avg_ic:.4f} | {min_len:10d} | {rating}")

    print("-" * 65)
    print("📌 İngilis dili üçün orta IC ≈ 0.0667")
    print("📌 Təsadüfi mətn üçün orta IC ≈ 0.0385")
    print(f"📌 Minimum sütun uzunluğu: {min_column_length}")

    if skipped_periods:
        skipped = ", ".join(f"{p} (min={m})" for p, m in skipped_periods)
        print(f"⚠️ Qısa sütunlara görə buraxılan periodlar: {skipped}")

    print("⚠️ Ən yüksək IC mütləq doğru period demək deyil; qısa mətnlərdə")
    print("   təsadüfi yüksəlmələr və həqiqi periodun qatları da yüksək görünə bilər.")
    print("   Namizədləri deşifrələmə nəticəsi ilə yoxlayın.")

    return [(period, avg_ic) for period, avg_ic, _ in ranked_results]
````

## 🔓 Açar uzunluğu məlum olduqda şifrənin sındırılması (25 dəq)

Açar uzunluğu $m$ məlum olduqda, Vijener şifrəsi $m$ sayda Sezar şifrəsinə parçalanır. Hər bir sütun üçün tezlik analizi apararaq açarın hər bir hərfini tapa bilərik.

### 7.1 Mətnin sütunlara bölünməsi

````python
def split_into_columns(ciphertext, period):
    """
    Şifrəli mətni period uzunluğunda sütunlara bölür

    Parametrlər:
        ciphertext: şifrəli mətn
        period: açar uzunluğu

    Qaytarır:
        columns: hər biri bir sütun olan siyahı
    """
    ciphertext = clean_text(ciphertext)
    columns = [''] * period

    for i, char in enumerate(ciphertext):
        columns[i % period] += char

    return columns

# Test
test_text = "ABCDEFGHIJ"
print(f"Mətn: {test_text}")
print("Period=3 olduqda sütunlar:")
for i, col in enumerate(split_into_columns(test_text, 3)):
    print(f"  Sütun {i+1}: {col}")
````

### 7.2 Hər sütun üçün tezlik analizi

````python
# İngilis dilində hərflərin standart tezlik sırası
ENGLISH_FREQ_ORDER = ['E', 'T', 'A', 'O', 'I', 'N', 'S', 'H', 'R', 'D',
                      'L', 'C', 'U', 'M', 'W', 'F', 'G', 'Y', 'P', 'B',
                      'V', 'K', 'J', 'X', 'Q', 'Z']

def frequency_analysis_for_column(column):
    """
    Bir sütun üçün tezlik analizi aparır

    Qaytarır:
        (hərf, say, faiz) formatında sıralanmış siyahı
    """
    freq = Counter(column)
    total = len(column)

    # Tezlik faizlərini hesabla
    freq_percent = []
    for letter, count in freq.items():
        percent = count / total * 100 if total > 0 else 0
        freq_percent.append((letter, count, percent))

    # Tezliyə görə sırala
    sorted_freq = sorted(freq_percent, key=lambda x: x[2], reverse=True)

    return sorted_freq

def find_shift_for_column(column):
    """
    Bir sütun üçün ən çox işlənən hərfə əsasən shift-i təxmin edir
    Fərz edirik ki, ən çox işlənən hərf 'E'-dir
    """
    freq = frequency_analysis_for_column(column)

    if not freq:
        return 0

    # Ən çox işlənən hərf
    most_common = freq[0][0]

    # 'E' = 4 (A=0 olduğuna görə)
    e_num = letter_to_num('E')
    most_common_num = letter_to_num(most_common)

    # Shift = (c - p) mod 26
    shift = (most_common_num - e_num) % 26

    return shift

def find_key(ciphertext, period):
    """
    Şifrəli mətn və period əsasında açarı tapır

    Parametrlər:
        ciphertext: şifrəli mətn
        period: açar uzunluğu

    Qaytarır:
        key: tapılan açar
    """
    columns = split_into_columns(ciphertext, period)
    key = []

    print("\n📊 SÜTUNLAR ÜZRƏ ANALİZ:")
    print("-" * 60)

    for i, col in enumerate(columns):
        freq = frequency_analysis_for_column(col)
        if freq:
            most_common = freq[0][0]
            shift = find_shift_for_column(col)
            key_letter = num_to_letter(shift)
            key.append(key_letter)

            print(f"Sütun {i+1}: ən çox işlənən = {most_common} "
                  f"({freq[0][2]:.2f}%), shift = {shift} ({key_letter})")

    return ''.join(key)
````

### 7.3 Tam şifrə sındırma prosesi

````python
def break_vigenere(ciphertext, max_period=20, known_key_length=None):
    """
    Vijener şifrəsini sındırmaq üçün tam proses

    Parametrlər:
        ciphertext: şifrəli mətn
        max_period: maksimum sınanacaq period
        known_key_length: əgər məlumdursa, birbaşa istifadə et

    Qaytarır:
        key: tapılan açar
        decrypted: deşifrələnmiş mətn
    """
    print("=" * 80)
    print("🔓 VİJENER ŞİFRƏSİNİ SINDIRMA PROSESİ")
    print("=" * 80)

    # Addım 1: Açar uzunluğunu təxmin et
    if known_key_length:
        period_guess = known_key_length
        print(f"\n1. MƏLUM AÇAR UZUNLUĞU: {period_guess}")
    else:
        print("\n1. KASISKI TESTİ İLƏ AÇAR UZUNLUĞUNUN TƏXMİNİ:")
        print("-" * 50)
        period_guess = kasiski_analysis(ciphertext)

        if not period_guess:
            print("Kasiski testi nəticə vermədi. Uyğunluq indeksindən istifadə edirik.")

            # Uyğunluq indeksi ilə yoxla
            print("\n2. UYĞUNLUQ İNDEKSİ ANALİZİ:")
            print("-" * 50)
            ic_results = ic_for_different_periods(ciphertext, max_period)

            if ic_results:
                period_guess = max(ic_results, key=lambda x: x[1])[0]
                print(f"\nIC analizinə əsasən seçilən period: {period_guess}")
            else:
                period_guess = 3  # default

    # Addım 3: Açarı tap
    print(f"\n3. AÇARIN TAPILMASI (period={period_guess}):")
    print("-" * 50)
    key = find_key(ciphertext, period_guess)
    print(f"\n🎯 TƏXMİN EDİLƏN AÇAR: {key}")

    # Addım 4: Mətni deşifrələ
    decrypted = vigenere_decrypt(ciphertext, key)

    print(f"\n4. DEŞİFRLƏNMIŞ MƏTN (ilk 200 simvol):")
    print("-" * 50)
    print(decrypted[:200])

    return key, decrypted

# Test üçün nümunə
# Uzun nümunə istifadə edirik ki, tezlik analizi hər sütunda E hərfini düzgün seçsin.
test_message = "weneedmoreevidencebeforewedecidethesecret" * 30
test_key = "SECRET"
test_cipher = vigenere_encrypt(test_message, test_key)
print(f"\n📨 TEST NÜMUNƏSİ:")
print(f"Orijinal mətn: {test_message[:50]}...")
print(f"Açar: {test_key}")
print(f"Şifrəli mətn: {test_cipher[:50]}...")

# Şifrəni sındır
key_found, decrypted = break_vigenere(test_cipher, known_key_length=len(test_key))
````

### ✍️ Çalışma 3: Vijener şifrəsinin sındırılması (2.5 bal)

Aşağıdakı qısa şifrəli mətn Vijener şifrəsi ilə şifrlənib. Onu analiz edin:

**Şifrəli mətn:** `"RIJVS KKSPP KXQYV BXLRI JVSKK SPPKX QYVBX"`

**İpucu:** Bu mətndə `"RIJVS"` ardıcıllığı təkrarlanır. Bu təkrarlar arasındakı məsafə açar uzunluğu haqqında məlumat verə bilər.

**Qeyd:** Mətn çox qısa olduğuna görə tək bir açarı deterministik şəkildə tapmaq çətin ola bilər. Məqsədiniz mümkün açar uzunluqlarını və namizəd açarları müqayisə edib ən inandırıcı deşifrələməni əsaslandırmaqdır.

**Addımlar:**
1. Kasiski testi ilə mümkün açar uzunluqlarını təxmin edin
2. Uyğunluq indeksi ilə namizəd uzunluqları müqayisə edin
3. Hər sütun üçün tezlik analizi aparın
4. Bir neçə namizəd açar/deşifrələmə yazın
5. Ən uyğun namizədi seçin və seçiminizi qısa izah edin


````python
# Çalışma 3 - Cavablar

print("📝 ÇALIŞMA 3 CAVABLARI")
print("=" * 80)

ciphertext = "RIJVS KKSPP KXQYV BXLRI JVSKK SPPKX QYVBX"
print(f"Şifrəli mətn: {ciphertext}")
print(f"Təmizlənmiş: {clean_text(ciphertext)}")
print("-" * 80)

# Kasiski testi
print("\n1. KASISKI TESTİ:")
period_guess = kasiski_analysis(ciphertext)

# Uyğunluq indeksi
print("\n2. UYĞUNLUQ İNDEKSİ ANALİZİ:")
ic_results = ic_for_different_periods(ciphertext)

# Açarı tap
print("\n3. AÇARIN TAPILMASI:")
if period_guess:
    key = find_key(ciphertext, period_guess)
    print(f"\n🎯 TAPILAN AÇAR: {key}")

    # Mətni deşifrələ
    decrypted = vigenere_decrypt(ciphertext, key)
    print(f"\n4. DEŞİFRLƏNMIŞ MƏTN:")
    print(f"   {decrypted}")
else:
    print("Açar uzunluğu tapılmadı. Fərqli periodları sınayın.")

    # Mümkün periodları sınaqdan keçir
    for period in range(2, 10):
        print(f"\nPeriod {period} üçün açar:")
        key = find_key(ciphertext, period)
        decrypted = vigenere_decrypt(ciphertext, key)
        print(f"   Açar: {key}")
        print(f"   Mətn: {decrypted[:50]}...")
````
