<h4>🔑 İş rejimləri (Modes of Operation)</h4>
<p>Blok şifrələr (AES, DES) tək bir bloku (128-bit və ya 64-bit) şifrləyir. İxtiyari uzunluqlu mesajları şifrləmək üçün <b>iş rejimləri</b> istifadə olunur.</p>

<p><b>Əsas məqsədlər:</b></p>
<ol>
    <li>Uzun mesajları bloklara bölmək</li>
    <li>Eyni açıq mətn bloklarının eyni şifrəli mətn blokları verməsinin qarşısını almaq</li>
    <li>Təhlükəsizliyi təmin etmək</li>
    <li>Müxtəlif tətbiq tələblərini qarşılamaq (sürət, paralellik, təsadüfi giriş)</li>
</ol>
</div>

**İş rejimlərinin növləri:**

| Rejim | Tam adı | Xüsusiyyətlər |
|-------|---------|---------------|
| **ECB** | Electronic Codebook | Sadə, sürətli, AMA TƏHLÜKƏSİZ DEYİL |
| **CBC** | Cipher Block Chaining | Ən geniş istifadə olunan, IV tələb edir |
| **CFB** | Cipher Feedback | Axın şifrəsi kimi, özünü sinxronizasiya edir |
| **OFB** | Output Feedback | Axın şifrəsi, xəta yayılmır |
| **CTR** | Counter | Paralel, təsadüfi giriş, ən sürətli |
| **GCM** | Galois/Counter Mode | AEAD (həm məxfilik, həm tamlıq) |
| **XTS** | XEX Tweakable Block Cipher | Disk şifrləməsi üçün |

## 🔲 ECB rejimi (Electronic Codebook) (15 dəq)

<div style="background-color: #fff3e0; padding: 15px; border-radius: 10px; border-left: 5px solid #ff9800;">
<h4>📌 ECB rejimi</h4>
<p>ECB ən sadə iş rejimidir. Hər blok müstəqil şifrlənir:</p>
<p style="text-align: center; font-size: 1.2em;">$$C_i = E_K(P_i)$$</p>
<p style="text-align: center; font-size: 1.2em;">$$P_i = D_K(C_i)$$</p>
</div>

<div style="background-color: #ffebee; padding: 15px; border-radius: 10px; border-left: 5px solid #f44336; margin-top: 10px;">
<h4>⚠️ ECB-nin əsas zəifliyi</h4>
<p>Eyni açıq mətn blokları həmişə <b>eyni şifrəli mətn bloklarına</b> şifrlənir. Bu, şifrəli mətndə nümunələrin (pattern) qalmasına səbəb olur.</p>
<p><i>Nəticə: ECB heç vaxt istifadə edilməməlidir!</i></p>
</div>

````python
BLOCK_SIZE = 16  # AES blok ölçüsü

def xor_bytes(left, right):
    """İki bayt sətrini XOR edir."""
    return bytes(a ^ b for a, b in zip(left, right))

def ecb_encrypt(plaintext, key, block_cipher):
    """
    ECB rejimində şifrləmə

    Parametrlər:
        plaintext: şifrlənəcək mətn (bytes)
        key: açar
        block_cipher: 16-baytlıq blok şifrləmə funksiyası

    Qaytarır:
        ciphertext: şifrəli mətn
    """
    padded = pad(plaintext, BLOCK_SIZE)
    ciphertext = bytearray()

    for i in range(0, len(padded), BLOCK_SIZE):
        block = padded[i:i + BLOCK_SIZE]
        ciphertext.extend(block_cipher(block, key))

    return bytes(ciphertext)

def ecb_decrypt(ciphertext, key, block_cipher_decrypt):
    """ECB rejimində deşifrləmə."""
    if len(ciphertext) % BLOCK_SIZE != 0:
        raise ValueError("Ciphertext uzunluğu blok ölçüsünün qatına bərabər olmalıdır")

    plaintext_padded = bytearray()

    for i in range(0, len(ciphertext), BLOCK_SIZE):
        block = ciphertext[i:i + BLOCK_SIZE]
        plaintext_padded.extend(block_cipher_decrypt(block, key))

    return unpad(bytes(plaintext_padded), BLOCK_SIZE)

def simple_block_cipher(block, key):
    """
    Sadə nümayiş üçün blok şifrə (təhsil məqsədli).
    Real sistemlərdə AES istifadə edin!
    """
    if len(block) != BLOCK_SIZE:
        raise ValueError("simple_block_cipher yalnız 16-baytlıq blok qəbul edir")
    if not key:
        raise ValueError("Açar boş ola bilməz")

    return bytes(b ^ key[i % len(key)] for i, b in enumerate(block))

# Test
print("=" * 70)
print("🔲 ECB REJİMİ TESTİ")
print("=" * 70)

key = os.urandom(BLOCK_SIZE)  # nümayiş üçün təsadüfi 16 baytlıq açar
plaintext = b"Salam, dunya! Bu ECB test mesajidir."

cipher = ecb_encrypt(plaintext, key, simple_block_cipher)
decrypted = ecb_decrypt(cipher, key, simple_block_cipher)

print(f"📨 Orijinal: {plaintext}")
print(f"🔒 Şifrəli (hex): {cipher.hex()}")
print(f"🔓 Deşifrə: {decrypted}")
print(f"\n✅ Uğurlu? {plaintext == decrypted}")
````

### 5.1 ECB rejiminin problemi - eyni blokların eyni şifrlənməsi

````python
import secrets


def split_blocks(data, block_size=BLOCK_SIZE):
    return [data[i:i + block_size] for i in range(0, len(data), block_size)]

def ecb_problem_demo():
    """
    Eyni 16-baytlıq blokların ECB-də eyni şifrləndiyini göstərir.
    """
    print("\n" + "-" * 70)
    print("⚠️ ECB PROBLEMİ - Eyni bloklar eyni şifrlənir")
    print("-" * 70)

    key = secrets.token_bytes(BLOCK_SIZE)
    repeating_block = b"AAAAAAAABBBBBBBB"   # 16 bayt
    plaintext = repeating_block * 2         # iki eyni blok

    print(f"📨 Orijinal mətn: {plaintext}")
    print("\n16-baytlıq açıq mətn blokları:")
    for idx, block in enumerate(split_blocks(plaintext)):
        print(f"  Blok {idx}: {block}")

    cipher = ecb_encrypt(plaintext, key, simple_block_cipher)
    cipher_blocks = split_blocks(cipher)

    print("\n16-baytlıq şifrəli bloklar:")
    for idx, block in enumerate(cipher_blocks):
        print(f"  Blok {idx}: {block.hex()}")

    print(f"\nİlk iki şifrəli blok eynidirmi? {cipher_blocks[0] == cipher_blocks[1]}")
    print("📌 Nəticə: ECB eyni blokları eyni şəkildə şifrləyir və nümunələri gizlətmir.")

ecb_problem_demo()
````

## 🔗 CBC rejimi (Cipher Block Chaining) (20 dəq)

<div style="background-color: #f0f8ff; padding: 15px; border-radius: 10px; border-left: 5px solid #4682b4;">
<h4>🔗 CBC rejimi</h4>
<p>CBC rejimində hər blok şifrlənmədən əvvəl əvvəlki şifrəli mətn bloku ilə XOR edilir:</p>

<p style="text-align: center; font-size: 1.2em;">$$C_i = E_K(P_i \oplus C_{i-1})$$</p>
<p style="text-align: center; font-size: 1.2em;">$$C_1 = E_K(P_1 \oplus IV)$$</p>

<p><b>IV (İlkin Vektor):</b> İlk blok üçün heç bir əvvəlki şifrəli mətn yoxdur, ona görə IV istifadə olunur.</p>
</div>

<div style="background-color: #f0f8ff; padding: 15px; border-radius: 10px; border-left: 5px solid #4682b4; margin-top: 10px;">
<h4>🔑 IV tələbləri</h4>
<ul>
    <li><b>Təkrarolunmaz olmalıdır</b>: Eyni açar ilə eyni IV təkrar istifadə edilməməlidir</li>
    <li><b>Təsadüfi olmalıdır</b>: Proqnozlaşdırıla bilən IV hücum riskini artırır</li>
    <li><b>Gizli olmalı deyil</b>: IV açıq şəkildə ötürülə bilər (şifrəli mətnlə birlikdə)</li>
</ul>
</div>

````python
def cbc_encrypt(plaintext, key, iv, block_cipher):
    """
    CBC rejimində şifrləmə

    Parametrlər:
        plaintext: şifrlənəcək mətn (bytes)
        key: açar
        iv: 16 baytlıq ilkin vektor
        block_cipher: şifrləmə funksiyası

    Qaytarır:
        ciphertext: şifrəli mətn
        iv: istifadə olunan IV
    """
    if len(iv) != BLOCK_SIZE:
        raise ValueError("IV uzunluğu 16 bayt olmalıdır")

    padded = pad(plaintext, BLOCK_SIZE)
    ciphertext = bytearray()
    prev = iv

    for i in range(0, len(padded), BLOCK_SIZE):
        block = padded[i:i + BLOCK_SIZE]
        xored = xor_bytes(block, prev)
        encrypted = block_cipher(xored, key)
        ciphertext.extend(encrypted)
        prev = encrypted

    return bytes(ciphertext), iv

def cbc_decrypt(ciphertext, key, iv, block_cipher_decrypt):
    """CBC rejimində deşifrləmə."""
    if len(iv) != BLOCK_SIZE:
        raise ValueError("IV uzunluğu 16 bayt olmalıdır")
    if len(ciphertext) % BLOCK_SIZE != 0:
        raise ValueError("Ciphertext uzunluğu blok ölçüsünün qatına bərabər olmalıdır")

    plaintext_padded = bytearray()
    prev = iv

    for i in range(0, len(ciphertext), BLOCK_SIZE):
        block = ciphertext[i:i + BLOCK_SIZE]
        decrypted = block_cipher_decrypt(block, key)
        plaintext_padded.extend(xor_bytes(decrypted, prev))
        prev = block

    return unpad(bytes(plaintext_padded), BLOCK_SIZE)

# Test
print("\n" + "=" * 70)
print("🔗 CBC REJİMİ TESTİ")
print("=" * 70)

iv = get_random_bytes(BLOCK_SIZE)
ciphertext, iv_used = cbc_encrypt(plaintext, key, iv, simple_block_cipher)
decrypted = cbc_decrypt(ciphertext, key, iv_used, simple_block_cipher)

print(f"IV: {iv_used.hex()}")
print(f"🔒 Şifrəli mətn (hex): {ciphertext.hex()}")
print(f"🔓 Deşifrələnmiş: {decrypted}")
print(f"\n✅ Uğurlu? {plaintext == decrypted}")
````

### 6.1 IV təkrarı problemi

````python
def iv_reuse_problem_demo():
    """
    Eyni IV-nin CBC-də niyə təhlükəli olduğunu göstərir.
    Eyni ilk blok + eyni IV => eyni ilk şifrəli blok.
    """
    import secrets

    print("\n" + "-" * 70)
    print("⚠️ IV TƏKRARI PROBLEMİ")
    print("-" * 70)

    key = secrets.token_bytes(16)
    iv = secrets.token_bytes(16)  # Demo üçün eyni IV qəsdən təkrar istifadə olunur.

    first_block = b"0123456789ABCDEF"  # 16 bayt
    message1 = first_block + b" | Gizli A"
    message2 = first_block + b" | Gizli B"

    c1, _ = cbc_encrypt(message1, key, iv, simple_block_cipher)
    c2, _ = cbc_encrypt(message2, key, iv, simple_block_cipher)

    print(f"Mesaj 1 şifrəli: {c1.hex()}")
    print(f"Mesaj 2 şifrəli: {c2.hex()}")

    print(f"\nİlk bloklar eynidir? {c1[:16] == c2[:16]}")
    print("📌 Nəticə: eyni IV təkrarlanarsa, ortaq prefiks haqqında məlumat sızır.")

if AES_AVAILABLE:
    iv_reuse_problem_demo()
````

## 🔄 CFB rejimi (Cipher Feedback) (15 dəq)

<div style="background-color: #f0f8ff; padding: 15px; border-radius: 10px; border-left: 5px solid #4682b4;">
<h4>🔄 CFB rejimi</h4>
<p>CFB rejimi blok şifrəni axın şifrəsinə çevirir. Registr şifrlənir, sonra nəticə açıq mətn seqmenti ilə XOR edilir.</p>

<p style="text-align: center; font-size: 1.2em;">$$O_i = E_K(R_i), \quad C_i = P_i \oplus O_i[:s]$$</p>
<p>Seqmentli CFB-də registr hər addımda <b>s</b> bit sola sürüşdürülür və sonuna cari şifrəli seqment əlavə olunur.</p>

<p><b>Xüsusiyyətlər:</b></p>
<ul>
    <li>Doldurma (padding) tələb etmir</li>
    <li>Özünü sinxronizasiya edir (bir neçə seqmentdən sonra bərpa oluna bilər)</li>
    <li>Segment ölçüsü (s) dəyişdirilə bilər</li>
</ul>
</div>


````python
def _cfb_segment_size_bytes(segment_size_bits):
    if segment_size_bits % 8 != 0:
        raise ValueError("segment_size_bits 8-in qatı olmalıdır")

    segment_size = segment_size_bits // 8
    if not (1 <= segment_size <= BLOCK_SIZE):
        raise ValueError("segment_size_bits 8 ilə 128 bit arasında olmalıdır")

    return segment_size

def cfb_encrypt(plaintext, key, iv, block_cipher, segment_size_bits=8):
    """
    CFB rejimində şifrləmə

    Parametrlər:
        plaintext: şifrlənəcək mətn (bytes)
        key: açar
        iv: ilkin vektor (16 bayt)
        block_cipher: blok şifrləmə funksiyası
        segment_size_bits: hər addımda işlənən seqment ölçüsü (bit)

    Qaytarır:
        ciphertext: şifrəli mətn
    """
    if len(iv) != BLOCK_SIZE:
        raise ValueError("IV uzunluğu 16 bayt olmalıdır")

    segment_size = _cfb_segment_size_bytes(segment_size_bits)
    ciphertext = bytearray()
    register = iv

    for i in range(0, len(plaintext), segment_size):
        keystream = block_cipher(register, key)
        segment = plaintext[i:i + segment_size]
        encrypted_segment = xor_bytes(segment, keystream[:len(segment)])
        ciphertext.extend(encrypted_segment)
        register = register[len(encrypted_segment):] + encrypted_segment

    return bytes(ciphertext)

def cfb_decrypt(ciphertext, key, iv, block_cipher, segment_size_bits=8):
    """
    CFB rejimində deşifrləmə.
    Qeyd: register plaintext ilə yox, şifrəli seqment ilə yenilənməlidir.
    """
    if len(iv) != BLOCK_SIZE:
        raise ValueError("IV uzunluğu 16 bayt olmalıdır")

    segment_size = _cfb_segment_size_bytes(segment_size_bits)
    plaintext = bytearray()
    register = iv

    for i in range(0, len(ciphertext), segment_size):
        encrypted_segment = ciphertext[i:i + segment_size]
        keystream = block_cipher(register, key)
        plaintext_segment = xor_bytes(encrypted_segment, keystream[:len(encrypted_segment)])
        plaintext.extend(plaintext_segment)
        register = register[len(encrypted_segment):] + encrypted_segment

    return bytes(plaintext)

# Test
print("\n" + "=" * 70)
print("🔄 CFB REJİMİ TESTİ")
print("=" * 70)

cfb_iv = get_random_bytes(BLOCK_SIZE)
cfb_cipher = cfb_encrypt(plaintext, key, cfb_iv, simple_block_cipher, 8)
cfb_decrypted = cfb_decrypt(cfb_cipher, key, cfb_iv, simple_block_cipher, 8)

print("Segment ölçüsü: 8 bit")
print(f"🔒 CFB şifrəli: {cfb_cipher.hex()}")
print(f"🔓 CFB deşifrə: {cfb_decrypted}")
print(f"\n✅ Uğurlu? {plaintext == cfb_decrypted}")
````

## ⏎ OFB rejimi (Output Feedback) (15 dəq)

<div style="background-color: #f0f8ff; padding: 15px; border-radius: 10px; border-left: 5px solid #4682b4;">
<h4>⏎ OFB rejimi</h4>
<p>OFB rejimində geri-əlaqə çıxışa tətbiq olunur:</p>

<p style="text-align: center; font-size: 1.2em;">$$O_0 = IV, \quad O_i = E_K(O_{i-1})$$</p>
<p style="text-align: center; font-size: 1.2em;">$$C_i = P_i \oplus O_i$$</p>

<p><b>Xüsusiyyətlər:</b></p>
<ul>
    <li>Açar axını (keystream) əvvəlcədən hesablana bilər</li>
    <li>Şifrəli mətndə bir bit xətası yalnız açıq mətndəki uyğun biti dəyişir; xəta yayılmır</li>
    <li>Paralel deyil (ardıcıl hesablama tələb edir)</li>
</ul>
</div>

````python
def ofb_encrypt(plaintext, key, iv, block_cipher):
    """OFB rejimində şifrləmə."""
    if len(iv) != BLOCK_SIZE:
        raise ValueError("IV uzunluğu 16 bayt olmalıdır")

    ciphertext = bytearray()
    output = iv

    for i in range(0, len(plaintext), BLOCK_SIZE):
        output = block_cipher(output, key)
        block = plaintext[i:i + BLOCK_SIZE]
        ciphertext.extend(xor_bytes(block, output[:len(block)]))

    return bytes(ciphertext)

def ofb_decrypt(ciphertext, key, iv, block_cipher):
    """OFB rejimində deşifrləmə (şifrləmə ilə eynidir)."""
    return ofb_encrypt(ciphertext, key, iv, block_cipher)

# Test
print("\n" + "=" * 70)
print("⏎ OFB REJİMİ TESTİ")
print("=" * 70)

ofb_iv = get_random_bytes(BLOCK_SIZE)
ofb_cipher = ofb_encrypt(plaintext, key, ofb_iv, simple_block_cipher)
ofb_decrypted = ofb_decrypt(ofb_cipher, key, ofb_iv, simple_block_cipher)

print(f"🔒 OFB şifrəli: {ofb_cipher.hex()}")
print(f"🔓 OFB deşifrə: {ofb_decrypted}")
print(f"\n✅ Uğurlu? {plaintext == ofb_decrypted}")
````

## 🔢 CTR rejimi (Counter) (20 dəq)

<div style="background-color: #f0f8ff; padding: 15px; border-radius: 10px; border-left: 5px solid #4682b4;">
<h4>🔢 CTR rejimi</h4>
<p>CTR rejimində sayğac dəyərləri şifrlənərək açar axını yaradılır:</p>

<p style="text-align: center; font-size: 1.2em;">$$C_i = P_i \oplus E_K(\text{CTR}_i), \quad \text{CTR}_i = \text{IV} + i$$</p>

<p><b>Xüsusiyyətlər:</b></p>
<ul>
    <li>✅ <b>Paralel şifrləmə/deşifrləmə</b> - hər blok müstəqil</li>
    <li>✅ <b>Təsadüfi giriş</b> - istənilən bloku deşifrləmək olar</li>
    <li>✅ <b>Padding tələb etmir</b> - axın şifrəsi kimi</li>
    <li>✅ <b>Ən sürətli rejim</b> - paralellik səbəbindən</li>
    <li>❌ <b>Sayğac təkrarı</b> - təhlükəlidir (OTP kimi)</li>
</ul>
</div>

````python
def ctr_encrypt(plaintext, key, iv, block_cipher):
    """
    CTR rejimində şifrləmə

    Parametrlər:
        plaintext: şifrlənəcək mətn (bytes)
        key: açar
        iv: 8-bayt nonce və ya 16-bayt nonce||counter
        block_cipher: şifrləmə funksiyası

    Qaytarır:
        ciphertext: şifrəli mətn
    """
    if len(iv) not in (8, BLOCK_SIZE):
        raise ValueError("iv ya 8-bayt nonce, ya da 16-bayt nonce||counter olmalıdır")

    if len(iv) == BLOCK_SIZE:
        nonce = iv[:8]
        counter = int.from_bytes(iv[8:], 'big')
    else:
        nonce = iv
        counter = 0

    ciphertext = bytearray()

    for i in range(0, len(plaintext), BLOCK_SIZE):
        counter_block = nonce + counter.to_bytes(8, 'big')
        encrypted_counter = block_cipher(counter_block, key)
        block = plaintext[i:i + BLOCK_SIZE]
        ciphertext.extend(xor_bytes(block, encrypted_counter[:len(block)]))
        counter = (counter + 1) & ((1 << 64) - 1)

    return bytes(ciphertext)

def ctr_decrypt(ciphertext, key, iv, block_cipher):
    """CTR rejimində deşifrləmə (şifrləmə ilə eynidir)."""
    return ctr_encrypt(ciphertext, key, iv, block_cipher)

# Test
print("\n" + "=" * 70)
print("🔢 CTR REJİMİ TESTİ")
print("=" * 70)

nonce = get_random_bytes(8)
initial_counter = 0
ctr_iv = nonce + initial_counter.to_bytes(8, 'big')

ctr_cipher = ctr_encrypt(plaintext, key, ctr_iv, simple_block_cipher)
ctr_decrypted = ctr_decrypt(ctr_cipher, key, ctr_iv, simple_block_cipher)

print(f"Nonce: {nonce.hex()}")
print(f"🔒 CTR şifrəli: {ctr_cipher.hex()}")
print(f"🔓 CTR deşifrə: {ctr_decrypted}")
print(f"\n✅ Uğurlu? {plaintext == ctr_decrypted}")
````

### 9.1 CTR rejiminin paralelləşdirilməsi

````python
def ctr_encrypt_parallel(plaintext, key, iv, block_cipher, num_threads=4):
    """
    CTR rejimində paralel şifrləmə

    Parametrlər:
        plaintext: şifrlənəcək mətn (bytes)
        key: açar
        iv: 16-bayt nonce||counter
        block_cipher: şifrləmə funksiyası
        num_threads: thread sayı

    Qaytarır:
        ciphertext: şifrəli mətn
    """
    if len(iv) != BLOCK_SIZE:
        raise ValueError("Paralel CTR üçün iv = nonce||counter formatında 16 bayt olmalıdır")
    if num_threads < 1:
        raise ValueError("num_threads ən azı 1 olmalıdır")

    nonce = iv[:8]
    start_counter = int.from_bytes(iv[8:], 'big')

    num_blocks = (len(plaintext) + BLOCK_SIZE - 1) // BLOCK_SIZE
    blocks_per_thread = (num_blocks + num_threads - 1) // num_threads
    results = [b""] * num_blocks

    def encrypt_chunk(thread_id):
        start_block = thread_id * blocks_per_thread
        end_block = min(start_block + blocks_per_thread, num_blocks)

        for block_idx in range(start_block, end_block):
            counter = start_counter + block_idx
            counter_block = nonce + counter.to_bytes(8, 'big')
            encrypted_counter = block_cipher(counter_block, key)

            start_byte = block_idx * BLOCK_SIZE
            end_byte = min(start_byte + BLOCK_SIZE, len(plaintext))
            block = plaintext[start_byte:end_byte]
            results[block_idx] = xor_bytes(block, encrypted_counter[:len(block)])

    with ThreadPoolExecutor(max_workers=num_threads) as executor:
        futures = [executor.submit(encrypt_chunk, i) for i in range(num_threads)]
        for future in futures:
            future.result()

    return b"".join(results)

print("\n" + "-" * 70)
print("⚡ CTR PARALEL ŞİFRLƏMƏ")
print("-" * 70)
print("CTR rejimi paralel şifrləməyə imkan verir!")
print("Bu, xüsusilə böyük fayllar üçün sürət üstünlüyü təmin edir.")
````

## 🔐 GCM (Galois/Counter Mode) - AEAD rejimi (15 dəq)

<div style="background-color: #f0f8ff; padding: 15px; border-radius: 10px; border-left: 5px solid #4682b4;">
<h4>🔐 GCM xüsusiyyətləri</h4>
<p>GCM CTR rejimi ilə Qalua meydanında vurmanı birləşdirən <b>AEAD</b> (Authenticated Encryption with Associated Data) rejimidir.</p>

<ul>
    <li>✅ <b>AEAD</b>: Həm məxfilik, həm də tamlıq təmin edir</li>
    <li>✅ <b>Paralel</b>: CTR əsaslı olduğu üçün paralelləşdirilə bilər</li>
    <li>✅ <b>Sürətli</b>: Müasir prosessorlarda sürətli implementasiya</li>
    <li>✅ <b>TLS 1.3</b>: Məcburi dəstəklənən rejim</li>
    <li>✅ <b>AAD</b>: Şifrlənməyən, ancaq tamlığı qorunan məlumat</li>
</ul>
</div>

````python
def gcm_demo():
    """
    PyCryptodome ilə GCM rejiminin nümayişi
    """
    if not AES_AVAILABLE:
        print("PyCryptodome yüklü deyil")
        return

    print("\n" + "=" * 80)
    print("🔐 GCM (GALOIS/COUNTER MODE) NÜMAYİŞİ")
    print("=" * 80)

    key = get_random_bytes(16)      # AES-128
    nonce = get_random_bytes(12)    # GCM üçün 96-bit nonce tövsiyə olunur

    plaintext = b"Bu cox gizli mesajdir. Hec kim oxuya bilmez."
    aad = "Əlaqəli məlumat (AAD) - şifrələnmir, amma tamligi qorunur".encode("utf-8")

    print(f"Açar: {key.hex()}")
    print(f"Nonce: {nonce.hex()}")
    print(f"📨 Açıq mətn: {plaintext}")
    print(f"📎 AAD: {aad}")

    cipher = AES.new(key, AES.MODE_GCM, nonce=nonce)
    cipher.update(aad)
    ciphertext, tag = cipher.encrypt_and_digest(plaintext)

    print(f"\n🔒 Şifrəli mətn: {ciphertext.hex()}")
    print(f"🏷️ Autentifikasiya teqi: {tag.hex()}")

    decipher = AES.new(key, AES.MODE_GCM, nonce=nonce)
    decipher.update(aad)

    try:
        decrypted = decipher.decrypt_and_verify(ciphertext, tag)
        print(f"\n🔓 Deşifrələnmiş: {decrypted}")
        print("✅ Autentifikasiya UĞURLU!")
    except ValueError:
        print("❌ Autentifikasiya XƏTASI! Məlumat dəyişdirilib")

    print("\n" + "-" * 50)
    print("AAD dəyişdirildikdə:")
    decipher2 = AES.new(key, AES.MODE_GCM, nonce=nonce)
    decipher2.update("Dəyişdirilmiş AAD".encode("utf-8"))

    try:
        decipher2.decrypt_and_verify(ciphertext, tag)
        print("❌ Bu işləməməlidir!")
    except ValueError:
        print("✅ Düzgün: AAD dəyişdirildikdə autentifikasiya xətası verir")

if AES_AVAILABLE:
    gcm_demo()
````

## 💽 XTS rejimi (Disk Şifrələməsi) (10 dəq)

<div style="background-color: #f0f8ff; padding: 15px; border-radius: 10px; border-left: 5px solid #4682b4;">
<h4>💽 XTS xüsusiyyətləri</h4>
<p>XTS (XEX Tweakable Block Cipher with Ciphertext Stealing) disk şifrləməsi üçün nəzərdə tutulmuş xüsusi rejimdir (IEEE 1619).</p>

<ul>
    <li><b>Tweak</b>: Hər blok üçün fərqli tweak dəyəri (adətən sektor nömrəsi əsasında)</li>
    <li><b>Təsadüfi giriş</b>: İstənilən sektoru müstəqil emal etmək olar</li>
    <li><b>Sektor asılılığı</b>: Eyni məlumat fərqli sektorda fərqli şifrlənir</li>
    <li><b>İki açar</b>: Biri məlumat, biri tweak üçün</li>
</ul>
</div>


````python
def gf_mul_alpha(tweak_bytes):
    """
    GF(2^128)-də α = x ilə vurma.
    XTS üçün little-endian təqdimat istifadə olunur.
    """
    tweak_int = int.from_bytes(tweak_bytes, "little")
    carry = (tweak_int >> 127) & 1
    tweak_int = (tweak_int << 1) & ((1 << 128) - 1)
    if carry:
        tweak_int ^= 0x87
    return tweak_int.to_bytes(16, "little")

def xts_tweak_calculation(sector_number, block_index, tweak_key):
    """
    XTS tweak dəyərinin hesablanması:
        T_0 = E_K2(sector_number)
        T_i = α^i · T_0
    """
    if not AES_AVAILABLE:
        raise RuntimeError("XTS tweak demo üçün PyCryptodome tələb olunur")
    if block_index < 0:
        raise ValueError("block_index mənfi ola bilməz")

    sector_bytes = sector_number.to_bytes(16, "little")
    # XTS-də tweak hesablanması üçün yalnız bir AES blok permutasiyası lazımdır.
    # PyCryptodome bunu ECB API-si ilə verir; ECB rejimi ümumi mesaj şifrələməsi
    # üçün semantik təhlükəsiz deyil və praktikada istifadə edilməməlidir.
    tweak = AES.new(tweak_key, AES.MODE_ECB).encrypt(sector_bytes)

    for _ in range(block_index):
        tweak = gf_mul_alpha(tweak)

    return tweak

print("\n" + "=" * 70)
print("💽 XTS TWEAK HESABLANMASI")
print("=" * 70)

sector = 42
tweak_key = get_random_bytes(16)
for i in range(5):
    tweak = xts_tweak_calculation(sector, i, tweak_key)
    print(f"Sektor {sector}, blok {i}: tweak = {tweak.hex()}")
````

````python
def modes_lab_menu():
    """İş rejimləri laboratoriyası interaktiv menyu."""
    while True:
        print("\n" + "=" * 70)
        print("🔄 İŞ REJİMLƏRİ LABORATORİYASI - Məşğələ 11")
        print("=" * 70)
        print("1. 🔲 ECB rejimi nümayişi")
        print("2. 🔗 CBC rejimi nümayişi")
        print("3. 🔄 CFB rejimi nümayişi")
        print("4. ⏎ OFB rejimi nümayişi")
        print("5. 🔢 CTR rejimi nümayişi")
        print("6. 🔐 GCM AEAD nümayişi (PyCryptodome)")
        print("7. 📊 Rejim müqayisəsi")
        print("8. 📁 Fayl şifrləmə (AES-GCM)")
        print("9. 💽 XTS disk şifrləməsi haqqında")
        print("0. ❌ Çıxış")
        print("=" * 70)

        choice = input("📌 Seçiminiz: ")

        if choice == '1':
            print("\n🔲 ECB REJİMİ:")
            plain = input("Mətn: ").encode()
            key = get_random_bytes(BLOCK_SIZE)

            cipher = ecb_encrypt(plain, key, simple_block_cipher)
            dec = ecb_decrypt(cipher, key, simple_block_cipher)

            print(f"🔒 Şifrəli: {cipher.hex()}")
            print(f"🔓 Deşifrə: {dec}")

        elif choice == '2':
            print("\n🔗 CBC REJİMİ:")
            plain = input("Mətn: ").encode()
            key = get_random_bytes(BLOCK_SIZE)
            iv = get_random_bytes(BLOCK_SIZE)

            cipher, used_iv = cbc_encrypt(plain, key, iv, simple_block_cipher)
            dec = cbc_decrypt(cipher, key, used_iv, simple_block_cipher)

            print(f"IV: {used_iv.hex()}")
            print(f"🔒 Şifrəli: {cipher.hex()}")
            print(f"🔓 Deşifrə: {dec}")

        elif choice == '3':
            print("\n🔄 CFB REJİMİ:")
            plain = input("Mətn: ").encode()
            key = get_random_bytes(BLOCK_SIZE)
            iv = get_random_bytes(BLOCK_SIZE)
            seg_bits = int(input("Segment ölçüsü (bit, 8/16/32/64/128): ") or "8")

            cipher = cfb_encrypt(plain, key, iv, simple_block_cipher, seg_bits)
            dec = cfb_decrypt(cipher, key, iv, simple_block_cipher, seg_bits)

            print(f"🔒 Şifrəli: {cipher.hex()}")
            print(f"🔓 Deşifrə: {dec}")

        elif choice == '4':
            print("\n⏎ OFB REJİMİ:")
            plain = input("Mətn: ").encode()
            key = get_random_bytes(BLOCK_SIZE)
            iv = get_random_bytes(BLOCK_SIZE)

            cipher = ofb_encrypt(plain, key, iv, simple_block_cipher)
            dec = ofb_decrypt(cipher, key, iv, simple_block_cipher)

            print(f"🔒 Şifrəli: {cipher.hex()}")
            print(f"🔓 Deşifrə: {dec}")

        elif choice == '5':
            print("\n🔢 CTR REJİMİ:")
            plain = input("Mətn: ").encode()
            key = get_random_bytes(BLOCK_SIZE)
            nonce = get_random_bytes(8)
            counter = 0
            iv = nonce + counter.to_bytes(8, 'big')

            cipher = ctr_encrypt(plain, key, iv, simple_block_cipher)
            dec = ctr_decrypt(cipher, key, iv, simple_block_cipher)

            print(f"Nonce: {nonce.hex()}")
            print(f"🔒 Şifrəli: {cipher.hex()}")
            print(f"🔓 Deşifrə: {dec}")

        elif choice == '6' and AES_AVAILABLE:
            print("\n🔐 GCM AEAD REJİMİ:")
            plain = input("Mətn: ").encode()
            aad = input("Əlaqəli məlumat (AAD, boş ola bilər): ").encode()

            key = get_random_bytes(16)
            nonce = get_random_bytes(12)

            cipher = AES.new(key, AES.MODE_GCM, nonce=nonce)
            cipher.update(aad)
            c, tag = cipher.encrypt_and_digest(plain)

            print(f"🔒 Şifrəli: {c.hex()}")
            print(f"🏷️ Teq: {tag.hex()}")

            decipher = AES.new(key, AES.MODE_GCM, nonce=nonce)
            decipher.update(aad)
            try:
                dec = decipher.decrypt_and_verify(c, tag)
                print(f"🔓 Deşifrə: {dec}")
                print("✅ Autentifikasiya uğurlu")
            except ValueError:
                print("❌ Autentifikasiya xətası")

        elif choice == '7':
            modes_comparison()

        elif choice == '8' and AES_AVAILABLE:
            filename = input("Fayl adı: ")
            password = input("Şifrə: ").encode()

            from Crypto.Hash import SHA256

            salt = get_random_bytes(16)
            key = PBKDF2(password, salt, 32, count=100000, hmac_hash_module=SHA256)
            nonce = get_random_bytes(12)

            with open(filename, 'rb') as f:
                data = f.read()

            cipher = AES.new(key, AES.MODE_GCM, nonce=nonce)
            encrypted, tag = cipher.encrypt_and_digest(data)

            with open(filename + '.enc', 'wb') as f:
                f.write(salt + nonce + tag + encrypted)

            print(f"✅ Fayl şifrləndi və autentifikasiya teqi əlavə edildi: {filename}.enc")
            print("   Format: salt(16) || nonce(12) || tag(16) || ciphertext")

        elif choice == '9':
            print("\n💽 XTS DISK ŞİFRLƏMƏSİ:")
            print("""
   XTS (XEX Tweakable Block Cipher with Ciphertext Stealing):
   • IEEE 1619 standartı
   • Disk şifrləməsi üçün nəzərdə tutulub
   • İstifadə edənlər: BitLocker, FileVault, LUKS
   • Tweak sektor nömrəsi və blok indeksindən asılıdır
   • Başqa sektora köçürülən məlumat yeni tweak tələb edir
   • İki açar: biri məlumat, biri tweak üçün
""")

        elif choice == '0':
            print("👋 Proqram bitdi. Sağ olun!")
            break

        else:
            print("❌ Yanlış seçim")

# Proqramı işə sal (istəyə bağlı)
# modes_lab_menu()
````
