<h4>🔑 Heş funksiyası</h4>
<p>Heş funksiyası $H: \{0,1\}^* \rightarrow \{0,1\}^n$ — ixtiyari uzunluqlu məlumatı sabit uzunluqlu $n$-bit çıxışa çevirən funksiyadır.</p>

<p><b>Əsas xüsusiyyətlər:</b></p>
<ul>
    <li><b>Deterministik:</b> Eyni giriş üçün həmişə eyni çıxış</li>
    <li><b>Sürətli hesablama:</b> İxtiyari giriş üçün tez hesablanır</li>
    <li><b>Birtərəflilik:</b> Heşdən orijinal məlumatı bərpa etmək çətindir</li>
    <li><b>Lavina effekti:</b> Bir bit dəyişikliyi çıxış bitlərinin təxminən yarısını dəyişməlidir</li>
</ul>

## 🐍 Python-da heş funksiyalarından istifadə (15 dəq)

````python
def hash_demo():
    """
    Müxtəlif heş alqoritmlərinin nümayişi
    """
    print("=" * 70)
    print("📦 HEŞ ALQORİTMLƏRİ NÜMAYİŞİ")
    print("=" * 70)

    message = b"Salam, dunya! Bu test mesajidir."
    print(f"📨 Mesaj: {message}")

    # MD5 (128-bit)
    md5_hash = hashlib.md5(message).hexdigest()
    print(f"\n🔹 MD5 (128-bit): {md5_hash}")
    print(f"   Uzunluq: {len(md5_hash)} hex simvol = {len(md5_hash)*4} bit")

    # SHA-1 (160-bit)
    sha1_hash = hashlib.sha1(message).hexdigest()
    print(f"\n🔹 SHA-1 (160-bit): {sha1_hash}")
    print(f"   Uzunluq: {len(sha1_hash)} hex simvol = {len(sha1_hash)*4} bit")

    # SHA-256 (256-bit)
    sha256_hash = hashlib.sha256(message).hexdigest()
    print(f"\n🔹 SHA-256 (256-bit): {sha256_hash}")
    print(f"   Uzunluq: {len(sha256_hash)} hex simvol = {len(sha256_hash)*4} bit")

    # SHA-512 (512-bit)
    sha512_hash = hashlib.sha512(message).hexdigest()
    print(f"\n🔹 SHA-512 (512-bit): {sha512_hash[:64]}...")
    print(f"   Uzunluq: {len(sha512_hash)} hex simvol = {len(sha512_hash)*4} bit")

    return {
        'md5': md5_hash,
        'sha1': sha1_hash,
        'sha256': sha256_hash,
        'sha512': sha512_hash
    }

hashes = hash_demo()
````

### 5.1 Lavina effekti nümayişi

````python
def avalanche_effect_demo():
    """
    Lavina effektini nümayiş etdirir
    """
    print("\n" + "-" * 70)
    print("❄️ LAVİNA EFFEKTİ NÜMAYİŞİ")
    print("-" * 70)

    # İki çox yaxın mesaj
    message1 = b"Kriptoqrafiya kursu"
    message2 = b"Kriptoqrafiya kursu."  # bir nöqtə əlavə olunub

    print(f"📨 Mesaj 1: {message1}")
    print(f"📨 Mesaj 2: {message2}")
    print(f"   Fərq: bir simvol ('.')")

    # SHA-256 heşlərini hesabla
    h1 = hashlib.sha256(message1).hexdigest()
    h2 = hashlib.sha256(message2).hexdigest()

    print(f"\n🔹 SHA-256(Message1): {h1}")
    print(f"🔹 SHA-256(Message2): {h2}")

    # Fərqli bitlərin sayını hesabla
    h1_bin = bin(int(h1, 16))[2:].zfill(256)
    h2_bin = bin(int(h2, 16))[2:].zfill(256)

    diff_count = sum(1 for a, b in zip(h1_bin, h2_bin) if a != b)
    diff_percent = (diff_count / 256) * 100

    print(f"\n📊 Fərqli bit sayı: {diff_count} / 256 ({diff_percent:.1f}%)")
    print(f"   İdeal lavina effekti: təxminən 50%")

avalanche_effect_demo()
````

````python
# Çalışma 1 - Cavablar

print("📝 ÇALIŞMA 1 CAVABLARI")
print("=" * 80)

# 1. Şəxsi heş
print("\n1. ŞƏXSİ HEŞ:")
my_name = "Səməd Vəkilov"  # Öz adınızla əvəz edin
name_bytes = my_name.encode()
print(f"   Ad: {my_name}")
print(f"   MD5: {hashlib.md5(name_bytes).hexdigest()}")
print(f"   SHA-1: {hashlib.sha1(name_bytes).hexdigest()}")
print(f"   SHA-256: {hashlib.sha256(name_bytes).hexdigest()}")

# 2. Böyük/kiçik hərf müqayisəsi
print("\n2. BÖYÜK/KİÇİK HƏRF MÜQAYİSƏSİ:")
lower_text = "salam"
upper_text = "SALAM"
h_lower = hashlib.sha256(lower_text.encode()).hexdigest()
h_upper = hashlib.sha256(upper_text.encode()).hexdigest()
print(f"   'salam' SHA-256: {h_lower}")
print(f"   'SALAM' SHA-256: {h_upper}")
print(f"   Heşlər eynidir? {h_lower == h_upper}")

# 3. Lavina testi (bir simvol dəyişikliyi)
print("\n3. LAVİNA TESTİ:")
text1 = "Kriptoqrafiya"
text2 = "Kriptoqrafiya!"  # bir simvol əlavə olunub
h1 = hashlib.sha256(text1.encode()).hexdigest()
h2 = hashlib.sha256(text2.encode()).hexdigest()

h1_bin = bin(int(h1, 16))[2:].zfill(256)
h2_bin = bin(int(h2, 16))[2:].zfill(256)
diff = sum(1 for a, b in zip(h1_bin, h2_bin) if a != b)

print(f"   Mətn 1: {text1}")
print(f"   Mətn 2: {text2}")
print(f"   Fərqli bit sayı: {diff}/256 ({diff/256*100:.1f}%)")
````

## 🛡️ Heş funksiyalarının təhlükəsizlik xüsusiyyətləri (15 dəq)

<div style="background-color: #f0f8ff; padding: 15px; border-radius: 10px; border-left: 5px solid #4682b4;">
<h4>🔐 Ön görüntü dayanıqlılığı (Preimage resistance)</h4>
<p>$y$ heş dəyəri verildikdə, $H(x) = y$ şərtini ödəyən hər hansı $x$ girişini tapmaq hesablama baxımından çətin olmalıdır.</p>
<p><b>Mürəkkəblik:</b> $2^n$ (n - heş uzunluğu)</p>
</div>

<div style="background-color: #f0f8ff; padding: 15px; border-radius: 10px; border-left: 5px solid #4682b4; margin-top: 10px;">
<h4>🔄 İkinci ön görüntü dayanıqlılığı (Second preimage resistance)</h4>
<p>$x_1$ girişi verildikdə, $H(x_1) = H(x_2)$ və $x_1 \neq x_2$ şərtini ödəyən fərqli $x_2$ girişini tapmaq hesablama baxımından çətin olmalıdır.</p>
<p><b>Mürəkkəblik:</b> $2^n$</p>
</div>

<div style="background-color: #f0f8ff; padding: 15px; border-radius: 10px; border-left: 5px solid #4682b4; margin-top: 10px;">
<h4>💥 Toqquşma dayanıqlılığı (Collision resistance)</h4>
<p>$H(x_1) = H(x_2)$ və $x_1 \neq x_2$ şərtini ödəyən iki fərqli $x_1, x_2$ girişi tapmaq hesablama baxımından çətin olmalıdır.</p>
<p><b>Mürəkkəblik:</b> $2^{n/2}$ (ad günü paradoksuna görə)</p>
</div>

````python
def preimage_search_demo(target_hash, hash_func=hashlib.sha256, max_trials=100000):
    """
    Ön görüntü axtarışını simulyasiya edir (çox kiçik axtarış fəzası ilə)
    """
    print("\n" + "-" * 70)
    print("🔎 ÖN GÖRÜNTÜ AXTARIŞI SİMULYASİYASI")
    print("-" * 70)

    print(f"Hədəf heş: {target_hash}")
    print(f"Axtarış fəzası: {max_trials} təsadüfi giriş")

    found = False
    start_time = time.time()

    for i in range(max_trials):
        # Təsadüfi giriş yarat
        test_input = f"test{i}".encode()
        test_hash = hash_func(test_input).hexdigest()

        if i % 10000 == 0:
            print(f"  Sınaq {i}: {test_input} -> {test_hash[:16]}...")

        if test_hash == target_hash:
            elapsed = time.time() - start_time
            print(f"\n✅ TAPILDI! Giriş: {test_input}")
            print(f"   Heş: {test_hash}")
            print(f"   Vaxt: {elapsed:.2f} saniyə")
            found = True
            break

    if not found:
        elapsed = time.time() - start_time
        print(f"\n❌ {max_trials} sınaqda tapılmadı")
        print(f"   Vaxt: {elapsed:.2f} saniyə")
        print(f"   Nəzəri mürəkkəblik: 2^{256} ≈ 10^77 sınaq!")

# Kiçik bir nümunə üçün
target = hashlib.sha256(b"gizli mesaj").hexdigest()
preimage_search_demo(target, max_trials=10000)
````

````python
def truncate_digest_bits(digest_bytes, bits):
    """
    Heş çıxışının ilk `bits` bitini tam ədəd kimi qaytarır.
    Bu funksiya 4-ə bölünməyən bit ölçüləri üçün də düzgün işləyir.
    """
    total_bits = len(digest_bytes) * 8
    if not 1 <= bits <= total_bits:
        raise ValueError(f"bits 1 ilə {total_bits} arasında olmalıdır")

    value = int.from_bytes(digest_bytes, "big")
    return value >> (total_bits - bits)

def format_bits(value, bits):
    return format(value, f"0{bits}b")

def collision_search_demo(hash_func=hashlib.sha256, bits=16, max_trials=10000):
    """
    Toqquşma axtarışını simulyasiya edir (kiçik heş ölçüsü ilə)
    """
    print("\n" + "-" * 70)
    print("💥 TOQQUŞMA AXTARIŞI SİMULYASİYASI")
    print("-" * 70)

    if bits <= 0:
        raise ValueError("bits müsbət olmalıdır")

    digest_bits = hash_func(b"").digest_size * 8
    if bits > digest_bits:
        raise ValueError(f"Bu heş funksiyası üçün maksimum {digest_bits} bit istifadə edilə bilər")

    print(f"İstifadə olunan heş: ilk {bits} bit")
    print(f"Mümkün dəyər sayı: 2^{bits} = {2**bits}")
    print(f"Ad günü həddi: ~2^({bits}/2)")

    seen_hashes = {}
    start_time = time.time()

    for i in range(max_trials):
        test_input = f"message{i}".encode()
        digest_bytes = hash_func(test_input).digest()

        # İlk `bits` biti düzgün şəkildə götür
        short_hash = truncate_digest_bits(digest_bytes, bits)

        if short_hash in seen_hashes:
            prev_input = seen_hashes[short_hash]
            elapsed = time.time() - start_time
            print(f"\n✅ TOQQUŞMA TAPILDI! ({i} sınaqdan sonra)")
            print(f"   Giriş 1: {prev_input}")
            print(f"   Giriş 2: {test_input}")
            print(f"   Ortaq heş (ilk {bits} bit): {format_bits(short_hash, bits)}")
            print(f"   Vaxt: {elapsed:.2f} saniyə")
            return short_hash, prev_input, test_input

        seen_hashes[short_hash] = test_input

        if i % 1000 == 0:
            print(f"  Sınaq {i}: cədvəldə {len(seen_hashes)} giriş")

    elapsed = time.time() - start_time
    print(f"\n❌ {max_trials} sınaqda toqquşma tapılmadı")
    print(f"   Vaxt: {elapsed:.2f} saniyə")
    return None

# 16-bit heş üçün toqquşma axtar
collision_search_demo(bits=16, max_trials=1000)
````

## 🎂 Ad günü paradoksu simulyasiyası (20 dəq)

````python
def birthday_collision_probability_exact(n, days=365):
    """
    n nəfər üçün ən azı bir ad günü toqquşmasının dəqiq ehtimalını hesablayır.
    """
    if n < 0:
        raise ValueError("n mənfi ola bilməz")
    if days <= 0:
        raise ValueError("days müsbət olmalıdır")
    if n > days:
        return 1.0

    unique_probability = 1.0
    for k in range(n):
        unique_probability *= (days - k) / days

    return 1 - unique_probability

def birthday_paradox_simulation(n, days=365, trials=10000):
    """
    Ad günü paradoksunu simulyasiya edir
    n: otaqdakı insan sayı
    days: ildəki gün sayı (365)
    trials: simulyasiya sayı
    """
    collisions = 0

    for _ in range(trials):
        birthdays = [random.randint(1, days) for _ in range(n)]
        if len(set(birthdays)) < n:
            collisions += 1

    probability = collisions / trials
    theoretical = birthday_collision_probability_exact(n, days)

    print(f"n = {n:2d} nəfər: Simulyasiya = {probability:.4f}, Nəzəri = {theoretical:.4f}")

    return probability, theoretical

def birthday_paradox_analysis():
    """
    Müxtəlif n dəyərləri üçün ad günü paradoksunu analiz edir
    """
    print("\n" + "=" * 70)
    print("🎂 AD GÜNÜ PARADOKSU ANALİZİ")
    print("=" * 70)

    n_values = [10, 15, 20, 22, 23, 25, 30, 35, 40, 50, 60]
    sim_probs = []
    theo_probs = []

    for n in n_values:
        sim, theo = birthday_paradox_simulation(n, 365, 5000)
        sim_probs.append(sim)
        theo_probs.append(theo)

    if PLT_AVAILABLE:
        plt.figure(figsize=(10, 6))
        plt.plot(n_values, sim_probs, 'bo-', label='Simulyasiya', markersize=8)
        plt.plot(n_values, theo_probs, 'r--', label='Dəqiq nəzəri', linewidth=2)
        plt.axhline(y=0.5, color='g', linestyle=':', linewidth=2, label='p=0.5')
        plt.axvline(x=23, color='orange', linestyle=':', linewidth=2, label='n=23')
        plt.xlabel('Otaqdakı insan sayı (n)')
        plt.ylabel('Toqquşma ehtimalı')
        plt.title('Ad Günü Paradoksu')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.show()

    return sim_probs, theo_probs

birthday_paradox_analysis()
````

````python
def hash_collision_probability(q, n_bits):
    """
    q sayda təsadüfi giriş üçün toqquşma ehtimalının yaxşı yaxınlaşdırması.
    n_bits: heş çıxışının bit uzunluğu
    """
    if q < 0:
        raise ValueError("q mənfi ola bilməz")
    if n_bits <= 0:
        raise ValueError("n_bits müsbət olmalıdır")

    N = 2**n_bits
    probability = 1 - math.exp(-(q * (q - 1)) / (2 * N))
    return probability

def trials_for_collision_probability(n_bits, probability=0.5):
    """
    Verilmiş ehtimalla ən azı bir toqquşma almaq üçün lazım olan q sayının yaxınlaşdırması.
    """
    if not 0 < probability < 1:
        raise ValueError("probability 0 ilə 1 arasında olmalıdır")

    N = 2**n_bits
    return math.sqrt(2 * N * math.log(1 / (1 - probability)))

def collision_probability_table():
    """
    Müxtəlif heş uzunluqları üçün toqquşma ehtimalı cədvəli
    """
    print("\n" + "-" * 70)
    print("📊 HEŞ FUNKSİYALARINDA TOQQUŞMA EHTİMALI")
    print("-" * 70)
    print(f"{'Heş uzunluğu':<15} {'q (sınaq sayı)':<20} {'Toqquşma ehtimalı':<20}")
    print("-" * 70)

    algorithms = [
        ("MD5 (128)", 128, 2**64),
        ("SHA-1 (160)", 160, 2**80),
        ("SHA-256", 256, 2**128),
        ("SHA-512", 512, 2**256)
    ]

    for name, bits, q in algorithms:
        prob = hash_collision_probability(q, bits)
        print(f"{name:<15} 2^{bits//2:<17} {prob:.6f}")

    print("\nQeyd: p = 0.5 üçün q təxminən 1.177 × 2^(n/2) olmalıdır.")

collision_probability_table()
````

## 🏗️ Merkl-Damqord quruluşu (20 dəq)

<div style="background-color: #f0f8ff; padding: 15px; border-radius: 10px; border-left: 5px solid #4682b4;">
<h4>🏗️ Merkl-Damqord quruluşu</h4>
<p>Bir çox klassik heş funksiyalarının (MD5, SHA-1, SHA-2) əsasını təşkil edir.</p>

<p style="text-align: center; font-size: 1.2em;">$$
\begin{align*}
h_0 &= IV \\
h_i &= f(h_{i-1}, M_i) \quad (i = 1,...,k) \\
H(M) &= h_k
\end{align*}
$$</p>

<p>Burada:</p>
<ul>
    <li>$IV$ — ilkin dəyər (Initial Value)</li>
    <li>$f$ — sıxışdırma funksiyası (compression function)</li>
    <li>$M_i$ — mesaj blokları</li>
</ul>
</div>

````python
class MerkleDamgard:
    """
    Merkl-Damqord quruluşunun sadə implementasiyası
    """

    def __init__(self, compression_func, iv, block_size=64, output_size=32):
        """
        compression_func: sıxışdırma funksiyası f(h, block) -> yeni h
        iv: ilkin dəyər (bytes)
        block_size: blok ölçüsü (bayt)
        output_size: çıxış ölçüsü (bayt)
        """
        if len(iv) != output_size:
            raise ValueError("IV uzunluğu output_size ilə eyni olmalıdır")

        self.compression_func = compression_func
        self.iv = iv
        self.block_size = block_size
        self.output_size = output_size

    def _pad(self, message):
        """
        MD doldurması: mesajın sonuna '1' + '0'* + uzunluq əlavə edir
        """
        original_length = len(message) * 8
        padded = message + b'\x80'

        while (len(padded) + 8) % self.block_size != 0:
            padded += b'\x00'

        padded += original_length.to_bytes(8, 'big')
        return padded

    def hash(self, message):
        """
        Mesajın heşini hesablayır
        """
        padded = self._pad(message)
        h = self.iv

        for i in range(0, len(padded), self.block_size):
            block = padded[i:i+self.block_size]
            h = self.compression_func(h, block)

            if len(h) != self.output_size:
                raise ValueError("compression_func səhv ölçüdə çıxış qaytardı")

        return h

# Sadə sıxışdırma funksiyası nümunəsi (təhsil üçün)
def simple_compression(h, block):
    """
    Təhsil məqsədli sıxışdırma funksiyası.
    Qeyd: Bu, real kriptoqrafik dizayn deyil; sadəcə Merkl-Damqord axınını göstərir.
    """
    return hashlib.sha256(h + block).digest()

# Test
print("\n" + "=" * 70)
print("🏗️ MERKL-DAMQORD QURULUŞU TESTİ")
print("=" * 70)

iv = b'\x00' * 32
md = MerkleDamgard(simple_compression, iv, block_size=64, output_size=32)

message = b"Salam, dunya! Bu Merkl-Damqord testidir."
hash_value = md.hash(message)

print(f"📨 Mesaj: {message}")
print(f"🔹 Heş: {hash_value.hex()}")
````

<p>Əgər $H(M)$ və $len(M)$ məlumdursa, hücumçu $M$-i bilmədən $H(M || pad || M')$ dəyərini hesablaya bilər.</p>

````python
def hmac_demo():
    """
    HMAC nümayişi
    """
    print("\n" + "-" * 70)
    print("🔏 HMAC (HASH-BASED MESSAGE AUTHENTICATION CODE)")
    print("-" * 70)

    # Gizli açar
    key = secrets.token_bytes(16)
    message = b"Bu vacib mesajdir"

    print(f"🔑 Açar (hex): {key.hex()}")
    print(f"📨 Mesaj: {message}")

    # HMAC hesabla
    mac = hmac.new(key, message, hashlib.sha256).hexdigest()
    print(f"\n🔏 HMAC-SHA256: {mac}")

    # Yoxlama
    recalculated_mac = hmac.new(key, message, hashlib.sha256).hexdigest()
    is_valid = hmac.compare_digest(mac, recalculated_mac)
    print("✓ HMAC uğurla yoxlandı" if is_valid else "✗ HMAC yoxlanması uğursuz")

    # Saxtalaşdırılmış mesaj nümunəsi
    tampered_message = b"Bu vacib mesajdir!"
    tampered_mac = hmac.new(key, tampered_message, hashlib.sha256).hexdigest()
    print(f"   Saxtalaşdırılmış mesaj üçün eyni MAC alınır? {hmac.compare_digest(mac, tampered_mac)}")

    # Sadə H(key || message) ilə müqayisə
    simple_mac = hashlib.sha256(key + message).hexdigest()
    print(f"\n📌 Sadə H(key || message): {simple_mac}")
    print("⚠ Uzunluq genişləndirmə hücumuna qarşı həssas!")

    print("\n✅ HMAC-ın üstünlükləri:")
    print("  • Uzunluq genişləndirmə hücumlarına qarşı davamlı")
    print("  • Heş funksiyasının zəifliklərindən az təsirlənir")
    print("  • TLS, IPsec, SSH kimi protokollarda istifadə olunur")

hmac_demo()
````

````python
class MerkleTree:
    """
    Sadə Merkle ağacı implementasiyası
    """

    def __init__(self, data_blocks):
        if not data_blocks:
            raise ValueError("Ən azı bir məlumat bloku verilməlidir")

        self.data_blocks = data_blocks
        self.leaves = [hashlib.sha256(block).digest() for block in data_blocks]
        self.root = self._build_tree(self.leaves)

    def _build_tree(self, nodes):
        """
        Merkle ağacını qurur
        """
        if len(nodes) == 1:
            return nodes[0]

        new_nodes = []
        for i in range(0, len(nodes), 2):
            if i + 1 < len(nodes):
                combined = nodes[i] + nodes[i+1]
            else:
                combined = nodes[i] + nodes[i]  # tək qaldıqda özünü təkrarla

            new_nodes.append(hashlib.sha256(combined).digest())

        return self._build_tree(new_nodes)

    def get_proof(self, index):
        """
        Verilmiş indeks üçün Merkle sübutu yaradır
        """
        if not 0 <= index < len(self.leaves):
            raise IndexError("index etibarlı yarpaq aralığında olmalıdır")

        proof = []
        nodes = self.leaves
        idx = index

        while len(nodes) > 1:
            if idx % 2 == 0:
                if idx + 1 < len(nodes):
                    proof.append(('right', nodes[idx + 1]))
                else:
                    proof.append(('right', nodes[idx]))  # tək qaldıqda özü
            else:
                proof.append(('left', nodes[idx - 1]))

            # Növbəti səviyyəyə keç
            new_nodes = []
            for i in range(0, len(nodes), 2):
                if i + 1 < len(nodes):
                    combined = nodes[i] + nodes[i+1]
                else:
                    combined = nodes[i] + nodes[i]
                new_nodes.append(hashlib.sha256(combined).digest())

            nodes = new_nodes
            idx //= 2

        return proof

    def verify_proof(self, data, index, proof, root):
        """
        Merkle sübutunu yoxlayır
        """
        if not 0 <= index < len(self.leaves):
            return False

        current_hash = hashlib.sha256(data).digest()
        idx = index
        level_size = len(self.leaves)

        for direction, sibling_hash in proof:
            if level_size <= 1:
                return False

            expected_direction = 'right' if idx % 2 == 0 else 'left'
            if direction != expected_direction:
                return False

            if direction == 'left':
                combined = sibling_hash + current_hash
            elif direction == 'right':
                combined = current_hash + sibling_hash
            else:
                return False

            current_hash = hashlib.sha256(combined).digest()
            idx //= 2
            level_size = (level_size + 1) // 2

        if level_size != 1:
            return False

        return current_hash == root

# Test
print("\n" + "-" * 70)
print("🌳 MERKLE AĞACI NÜMAYİŞİ")
print("-" * 70)

# 4 məlumat bloku
blocks = [
    b"Transaction 1",
    b"Transaction 2",
    b"Transaction 3",
    b"Transaction 4"
]

tree = MerkleTree(blocks)
print(f"🌲 Merkle kökü: {tree.root.hex()}")

# 2-ci blok üçün sübut
proof = tree.get_proof(1)
print(f"\n📋 Blok 2 üçün sübut:")
for direction, h in proof:
    print(f"   {direction}: {h.hex()}")

# Sübutu yoxla
is_valid = tree.verify_proof(blocks[1], 1, proof, tree.root)
print(f"\n✅ Sübut etibarlıdır? {is_valid}")
````
