<p>Yaxşı qarışdırma: <i>açarın bir biti dəyişdikdə, şifrəli mətndə təsadüfi dəyişikliklər olmalıdır</i>.</p>

## 🔢 S-qutuları və P-qutuları (15 dəq)

### 6.1 S-qutusu yaratma

S-qutuları (Substitution boxes) — qarışdırmanın əsas mənbəyidir. Onlar qeyri-xətti əvəzetmə həyata keçirir.

````python
def create_sbox(input_bits, output_bits, mapping=None):
    """
    S-qutusu yaradır.

    Qeydlər:
        - input_bits == output_bits olduqda təsadüfi permutasiya yaradılır
        - input_bits < output_bits olduqda çıxışlar təkrarsız seçilir
        - input_bits > output_bits olduqda çıxışlar balanslı şəkildə təkrarlanır
    """
    if input_bits <= 0 or output_bits <= 0:
        raise ValueError("input_bits və output_bits müsbət olmalıdır")

    input_size = 1 << input_bits
    output_range = 1 << output_bits

    if mapping is not None:
        if len(mapping) != input_size:
            raise ValueError("Verilmiş mapping uzunluğu 2**input_bits olmalıdır")
        if any((x < 0 or x >= output_range) for x in mapping):
            raise ValueError("S-qutusu dəyərləri çıxış aralığında olmalıdır")
        return list(mapping)

    # n x n S-qutusu: permutasiya
    if input_bits == output_bits:
        sbox = list(range(output_range))
        random.shuffle(sbox)
        return sbox

    # n x m (n < m): mümkün qədər fərqli çıxışlar seç
    if input_bits < output_bits:
        return random.sample(range(output_range), input_size)

    # n x m (n > m): balanslı təkrarlanan çıxışlar
    repeats = input_size // output_range
    sbox = list(range(output_range)) * repeats
    random.shuffle(sbox)
    return sbox

def apply_sbox(value, sbox, input_bits, output_bits=None, total_input_bits=None):
    """
    S-qutusunu tətbiq edir.

    Parametrlər:
        value: giriş dəyəri (int)
        sbox: S-qutusu cədvəli
        input_bits: hər blok üçün giriş bit sayı
        output_bits: hər blok üçün çıxış bit sayı
        total_input_bits: sabit enli giriş üçün ümumi bit sayı (istəyə bağlı)

    Qaytarır:
        result: S-qutusu tətbiq olunmuş dəyər
    """
    if input_bits <= 0:
        raise ValueError("input_bits müsbət olmalıdır")
    if len(sbox) != (1 << input_bits):
        raise ValueError("S-qutusunun uzunluğu 2**input_bits olmalıdır")

    if output_bits is None:
        output_bits = max(1, max(sbox).bit_length())
    elif output_bits <= 0:
        raise ValueError("output_bits müsbət olmalıdır")

    mask = (1 << input_bits) - 1

    if total_input_bits is None:
        total_input_bits = max(input_bits, value.bit_length())

    num_chunks = max(1, math.ceil(total_input_bits / input_bits))
    result = 0

    # Ən soldakı blokdan başlayaraq tətbiq et
    for chunk_index in range(num_chunks - 1, -1, -1):
        idx = (value >> (chunk_index * input_bits)) & mask
        result = (result << output_bits) | sbox[idx]

    return result

# Test
print("=" * 70)
print("🔢 S-QUTUSU TESTİ")
print("=" * 70)

# 3-bit giriş, 3-bit çıxış üçün S-qutusu
sbox = create_sbox(3, 3)
print(f"S-qutusu: {sbox}")

# Test dəyərləri
print("\nGiriş -> Çıxış:")
for i in range(8):
    out = apply_sbox(i, sbox, 3, 3)
    print(f"  {i:03b} ({i:2d}) -> {out:03b} ({out:2d})")
````

### 6.2 P-qutusu yaratma

P-qutuları (Permutation boxes) — yayılmanın əsas mənbəyidir. Onlar bitlərin yerini dəyişir.

````python
import random


def create_pbox(bits, rng=None):
    """
    Bit səviyyəsində P-qutusu yaradır
    """
    if bits <= 0:
        raise ValueError("bits müsbət olmalıdır")

    permutation = list(range(bits))
    (rng or random).shuffle(permutation)
    return permutation


def apply_pbox(value, pbox, bits):
    """
    P-qutusunu tətbiq edir (bit permutasiyası)

    pbox[source_bit] = destination_bit mənasını daşıyır.
    """
    if len(pbox) != bits or sorted(pbox) != list(range(bits)):
        raise ValueError("P-qutusu 0..bits-1 aralığının permutasiyası olmalıdır")

    value &= (1 << bits) - 1
    result = 0

    for src_pos, dst_pos in enumerate(pbox):
        if (value >> src_pos) & 1:
            result |= (1 << dst_pos)

    return result


# Test
print("\n" + "=" * 70)
print("🔄 P-QUTUSU TESTİ")
print("=" * 70)

# 8-bit P-qutusu: sabit toxumlu generator notebook çıxışını təkrarlana bilən edir.
pbox = create_pbox(8, rng=random.Random(0))
print(f"P-qutusu permutasiyası: {pbox}")

# Test dəyəri
test_value = 0b10101010
print(f"Test dəyəri: {test_value:08b} ({test_value:3d})")

permuted = apply_pbox(test_value, pbox, 8)
print(f"Permutasiya olunmuş: {permuted:08b} ({permuted:3d})")

# Tərs P-qutusu
def invert_pbox(pbox):
    inv = [0] * len(pbox)
    for src_pos, dst_pos in enumerate(pbox):
        inv[dst_pos] = src_pos
    return inv


inv_pbox = invert_pbox(pbox)
restored = apply_pbox(permuted, inv_pbox, 8)
print(f"Bərpa edilmiş: {restored:08b} ({restored:3d})")
print(f"Uğurlu? {test_value == restored}")
````

### ✍️ Çalışma 1: S-qutusu və P-qutusu (1.5 bal)

Aşağıdakı tapşırıqları yerinə yetirin:

1. **S-qutusu yaratma:** 4-bit giriş, 4-bit çıxış üçün S-qutusu yaradın. Cədvəli çap edin.

2. **Qeyri-xəttilik testi:** S-qutusunun qeyri-xətti olduğunu yoxlayın (xətti yaxınlaşma olub-olmadığını test edin).
   * Burada $\alpha \cdot x$ və $\beta \cdot S(x)$ skalyar hasilləri GF(2)-də paritet kimi hesablanır.
   * Bütün sıfır olmayan maskalar üçün Walsh əmsalını hesablayın:
     $$
     W(\alpha,\beta)=\sum_{x=0}^{15}(-1)^{(\alpha\cdot x)\oplus(\beta\cdot S(x))}.
     $$
   * Alternativ olaraq eyni məlumatı qərəz kimi verə bilərsiniz:
     $$
     \left|\Pr_x[\alpha\cdot x=\beta\cdot S(x)]-\frac12\right|=\frac{|W(\alpha,\beta)|}{32}.
     $$
   * Maksimum mütləq qərəzi hesabatda göstərin və nəticəyə əsasən S-qutusunun xətti olub-olmadığını izah edin.

3. **P-qutusu yaratma:** 16-bit P-qutusu yaradın və bir neçə test dəyəri üçün tətbiq edin.

4. **Birləşmə:** S-qutusu və P-qutusunu birləşdirən `sbox_pbox_round(x, sbox, pbox)` funksiyası yazın.
   * Əvvəlcə S-qutusunu tətbiq edin, sonra P-qutusunu


````python
# Çalışma 1 - Cavablar

print("📝 ÇALIŞMA 1 CAVABLARI")
print("=" * 80)

# 1. 4-bit S-qutusu
print("\n1. 4-bit S-QUTUSU:")
sbox4 = create_sbox(4, 4)
print("   S-qutusu cədvəli:")
for i in range(0, 16, 4):
    print(f"      {i:2d}-{i+3:2d}: {sbox4[i:i+4]}")

# 2. Qeyri-xəttilik testi (sadələşdirilmiş)
print("\n2. QEYRİ-XƏTTİLİK TESTİ:")
max_bias = 0
for alpha in range(1, 16):
    for beta in range(1, 16):
        count = 0
        for x in range(16):
            y = sbox4[x]
            input_parity = bin(x & alpha).count('1') % 2
            output_parity = bin(y & beta).count('1') % 2
            if input_parity == output_parity:
                count += 1
        bias = abs(count / 16 - 0.5)
        if bias > max_bias:
            max_bias = bias
print(f"   Maksimum bias: {max_bias:.4f}")
print(f"   Nəticə: {'✅ Yaxşı qeyri-xəttilik' if max_bias < 0.25 else '⚠️ Zəif qeyri-xəttilik'}")

# 3. 16-bit P-qutusu
print("\n3. 16-bit P-QUTUSU:")
pbox16 = create_pbox(16)
print(f"   Permutasiya: {pbox16}")

test_val = 0x1234
permuted16 = apply_pbox(test_val, pbox16, 16)
print(f"   Test dəyəri: {test_val:04x}")
print(f"   Permutasiya: {permuted16:04x}")

# 4. Birləşmiş funksiya
print("\n4. BİRLƏŞMİŞ FUNKSİYA:")
def sbox_pbox_round(x, sbox, pbox, input_bits=4, output_bits=4):
    """S-qutusu və P-qutusunu birləşdirir"""
    total_bits = len(pbox)
    after_sbox = apply_sbox(x, sbox, input_bits, output_bits, total_bits)
    after_pbox = apply_pbox(after_sbox, pbox, len(pbox))
    return after_pbox

test_round = 0x1234
result_round = sbox_pbox_round(test_round, sbox4, pbox16)
print(f"   Giriş: {test_round:04x}")
print(f"   Nəticə: {result_round:04x}")
````

## 🔄 Feystel şəbəkəsinin implementasiyası (25 dəq)

<div style="background-color: #f0f8ff; padding: 15px; border-radius: 10px; border-left: 5px solid #4682b4;">
<h4>🔄 Feystel raundu</h4>
<p>Bir Feystel raundu aşağıdakı əməliyyatlardan ibarətdir:</p>

<p style="text-align: center; font-size: 1.2em;">$$
\begin{align*}
L_{i+1} &= R_i \\
R_{i+1} &= L_i \oplus F(R_i, K_i)
\end{align*}
$$</p>

<p>Burada:</p>
<ul>
    <li>$F$ — raund funksiyası (qarışdırma və yayılma təmin edir)</li>
    <li>$K_i$ — i-ci raund açarı</li>
    <li>$\oplus$ — XOR əməliyyatı</li>
</ul>

<p><b>Üstünlükləri:</b></p>
<ul>
    <li>Şifrləmə və deşifrləmə eyni struktura malikdir</li>
    <li>Raund funksiyasının tərsinə olması tələb olunmur</li>
    <li>Aparat implementasiyası asandır</li>
</ul>
</div>

````python
import inspect

class FeistelCipher:
    """
    Feystel şəbəkəsi əsasında blok şifrə

    Bu sinif Feystel strukturunu implementasiya edir.
    İstənilən raund funksiyası istifadə edilə bilər.

    XƏBƏRDARLIQ: encrypt()/decrypt() metodları çoxbloklu məlumat üçün ECB
    rejimini göstərir. ECB təhlükəsiz deyil və yalnız nümayiş/tədris məqsədi
    ilə daxil edilib; praktik fayl şifrələməsi üçün autentifikasiyalı,
    təsadüfiləşdirilmiş rejimdən (məsələn, AEAD) istifadə edin.
    """

    def __init__(self, round_keys, round_function, block_size=16):
        """
        Parametrlər:
            round_keys: raund açarları siyahısı
            round_function: F(R, K) funksiyası
            block_size: blok ölçüsü (bit)
        """
        if not round_keys:
            raise ValueError("Ən azı bir raund açarı verilməlidir")
        if block_size <= 0 or block_size % 2 != 0:
            raise ValueError("block_size müsbət və cüt olmalıdır")

        self.round_keys = list(round_keys)
        self.rounds = len(round_keys)
        self.round_function = round_function
        self.block_size = block_size
        self.half_size = block_size // 2
        self.mask = (1 << self.half_size) - 1

        params = list(inspect.signature(round_function).parameters.values())
        positional = [
            p for p in params
            if p.kind in (inspect.Parameter.POSITIONAL_ONLY, inspect.Parameter.POSITIONAL_OR_KEYWORD)
        ]
        has_varargs = any(p.kind == inspect.Parameter.VAR_POSITIONAL for p in params)
        self._round_accepts_bits = has_varargs or len(positional) >= 3

    def _split(self, block):
        """
        Bloku sol və sağ yarımlara bölür
        """
        left = (block >> self.half_size) & self.mask
        right = block & self.mask
        return left, right

    def _combine(self, left, right):
        """
        Sol və sağ yarımları birləşdirir
        """
        return ((left & self.mask) << self.half_size) | (right & self.mask)

    def _round(self, right, key):
        if self._round_accepts_bits:
            value = self.round_function(right, key, self.half_size)
        else:
            value = self.round_function(right, key)
        return value & self.mask

    def encrypt_block(self, plaintext):
        """
        Bir bloku şifrləyir
        """
        if plaintext < 0 or plaintext >= (1 << self.block_size):
            raise ValueError("plaintext blok ölçüsünə sığmalıdır")

        L, R = self._split(plaintext)

        for key in self.round_keys:
            L, R = R, (L ^ self._round(R, key)) & self.mask

        return self._combine(L, R)

    def decrypt_block(self, ciphertext):
        """
        Bir bloku deşifrləyir
        """
        if ciphertext < 0 or ciphertext >= (1 << self.block_size):
            raise ValueError("ciphertext blok ölçüsünə sığmalıdır")

        L, R = self._split(ciphertext)

        for key in reversed(self.round_keys):
            L, R = (R ^ self._round(L, key)) & self.mask, L

        return self._combine(L, R)

    def _pad(self, data):
        block_bytes = self.block_size // 8
        if block_bytes == 0:
            raise ValueError("Bayt səviyyəsində şifrləmə üçün block_size ən azı 8 olmalıdır")

        pad_len = block_bytes - (len(data) % block_bytes)
        if pad_len == 0:
            pad_len = block_bytes
        return data + bytes([pad_len] * pad_len)

    def _unpad(self, data):
        if not data:
            return data

        block_bytes = self.block_size // 8
        pad_len = data[-1]

        if pad_len < 1 or pad_len > block_bytes:
            raise ValueError("Yanlış PKCS#7 doldurması")
        if data[-pad_len:] != bytes([pad_len] * pad_len):
            raise ValueError("Yanlış PKCS#7 doldurması")

        return data[:-pad_len]

    def encrypt(self, data):
        """
        Çoxbloklu məlumatı şifrləyir (ECB rejimi).

        XƏBƏRDARLIQ: ECB təhlükəsiz deyil və yalnız nümayiş üçündür; praktik
        fayl şifrələməsi üçün autentifikasiyalı, təsadüfiləşdirilmiş rejim
        (məsələn, AEAD) tələb olunur.
        """
        if self.block_size % 8 != 0:
            raise ValueError("Bayt səviyyəsində şifrləmə üçün block_size 8-ə bölünməlidir")

        if isinstance(data, str):
            data = data.encode('utf-8')

        data = self._pad(data)

        result = bytearray()
        block_bytes = self.block_size // 8

        for i in range(0, len(data), block_bytes):
            block_data = data[i:i + block_bytes]
            block = int.from_bytes(block_data, 'big')
            encrypted = self.encrypt_block(block)
            result.extend(encrypted.to_bytes(block_bytes, 'big'))

        return bytes(result)

    def decrypt(self, data):
        """
        Çoxbloklu məlumatı deşifrləyir (ECB rejimi).

        XƏBƏRDARLIQ: Bu metod yalnız ECB nümayişi ilə uyğun şifrəli mətni
        açmaq üçündür; ECB praktik fayl şifrələməsi üçün istifadə olunmamalıdır.
        """
        if self.block_size % 8 != 0:
            raise ValueError("Bayt səviyyəsində deşifrləmə üçün block_size 8-ə bölünməlidir")

        result = bytearray()
        block_bytes = self.block_size // 8

        if len(data) % block_bytes != 0:
            raise ValueError("Şifrəli məlumatın uzunluğu blok ölçüsünə bölünməlidir")

        for i in range(0, len(data), block_bytes):
            block_data = data[i:i + block_bytes]
            block = int.from_bytes(block_data, 'big')
            decrypted = self.decrypt_block(block)
            result.extend(decrypted.to_bytes(block_bytes, 'big'))

        return self._unpad(bytes(result))
````

### 7.2 Raund funksiyaları

````python
def simple_round_function(R, key, bits=None):
    """
    Sadə raund funksiyası - yalnız XOR
    """
    value = R ^ key
    if bits is not None:
        value &= (1 << bits) - 1
    return value

def xor_and_rotate(R, key, bits=8, rotation=5):
    """
    XOR və bit döndərmə

    Qeyd:
        bits Feystel şifrəsində yarım blokun ölçüsünə uyğun olmalıdır.
    """
    if bits <= 0:
        raise ValueError("bits müsbət olmalıdır")

    mask = (1 << bits) - 1
    x = (R ^ key) & mask

    shift = rotation % bits
    if shift:
        x = ((x << shift) | (x >> (bits - shift))) & mask

    return x

def sbox_round_function(R, key, sbox, pbox=None, bits=8):
    """
    S-qutusu və P-qutusu olan raund funksiyası

    Parametrlər:
        R: sağ yarım
        key: raund açarı
        sbox: 4x4 S-qutusu
        pbox: bit permutasiyası (istəyə bağlı)
        bits: yarım blok ölçüsü
    """
    if bits % 4 != 0:
        raise ValueError("Bu nümunədə bits 4-ə bölünməlidir")
    if len(sbox) != 16:
        raise ValueError("Bu nümunədə 4x4 S-qutusu gözlənilir")
    if pbox is not None and len(pbox) != bits:
        raise ValueError("P-qutusunun uzunluğu bits ilə eyni olmalıdır")

    x = (R ^ key) & ((1 << bits) - 1)

    # S-qutusunu 4-bit bloklara tətbiq et
    result = 0
    for i in range(0, bits, 4):
        block = (x >> i) & 0xF
        result |= (sbox[block] << i)

    if pbox is not None:
        result = apply_pbox(result, pbox, bits)

    return result & ((1 << bits) - 1)

def make_sbox_round_function(sbox, pbox=None, bits=8):
    """
    FeistelCipher üçün uyğun (R, K, bits) imzalı raund funksiyası qaytarır
    """
    def round_function(R, key, half_bits=None):
        use_bits = bits if half_bits is None else half_bits
        if use_bits != bits:
            raise ValueError("Hazır S/P raund funksiyası fərqli bit ölçüsü ilə çağırıldı")
        return sbox_round_function(R, key, sbox, pbox, use_bits)

    return round_function

# Test
print("\n" + "=" * 70)
print("🔄 RAUND FUNKSİYALARI TESTİ")
print("=" * 70)

test_R = 0xAB
test_key = 0xCD

print(f"simple_round_function({test_R:02x}, {test_key:02x}) = {simple_round_function(test_R, test_key, 8):02x}")
print(f"xor_and_rotate({test_R:02x}, {test_key:02x}) = {xor_and_rotate(test_R, test_key, 8):02x}")
````

## 📊 Lavina effektinin analizi (20 dəq)

````python
import secrets

def hamming_distance(a, b):
    """
    İki ədəd arasındakı Hamming məsafəsini hesablayır
    (fərqli bitlərin sayı)
    """
    xor = a ^ b
    distance = 0
    while xor:
        distance += xor & 1
        xor >>= 1
    return distance

def avalanche_effect_test(cipher, plaintext, bit_position):
    """
    Bir bit dəyişikliyinin təsirini ölçür

    Parametrlər:
        cipher: Feystel şifrəsi
        plaintext: açıq mətn
        bit_position: dəyişdiriləcək bit mövqeyi

    Qaytarır:
        diff_bits: fərqli bit sayı
        c1: orijinal şifrəli mətn
        c2: dəyişdirilmiş şifrəli mətn
    """
    # Orijinal şifrəli mətn
    c1 = cipher.encrypt_block(plaintext)

    # Bir bit dəyişdirilmiş açıq mətn
    modified = plaintext ^ (1 << bit_position)
    c2 = cipher.encrypt_block(modified)

    # Fərqli bitlərin sayı
    diff_bits = hamming_distance(c1, c2)

    return diff_bits, c1, c2

def avalanche_analysis(cipher, num_tests=100):
    """
    Lavina effektini analiz edir
    """
    print("\n" + "=" * 80)
    print("❄️ LAVİNA EFFEKTİ ANALİZİ")
    print("=" * 80)

    results = []
    block_size = cipher.block_size

    for test in range(num_tests):
        # Kriptoqrafik təhlükəsiz təsadüfi açıq mətn
        plaintext = secrets.randbits(block_size)

        for bit in range(min(block_size, 8)):  # İlk 8 biti test et
            diff, _, _ = avalanche_effect_test(cipher, plaintext, bit)
            results.append(diff)

    # Statistik məlumatlar
    avg_diff = sum(results) / len(results)
    ideal_avg = block_size / 2

    print(f"Test sayı: {len(results)}")
    print(f"Blok ölçüsü: {block_size} bit")
    print(f"Orta fərqli bit sayı: {avg_diff:.2f}")
    print(f"İdeal orta (50%): {ideal_avg:.2f}")
    print(f"Fərq: {abs(avg_diff - ideal_avg):.2f} bit")
    print(f"Nisbət: {avg_diff/block_size*100:.1f}%")

    # Paylanma histogramı
    if HAS_MATPLOTLIB:
        plt.figure(figsize=(10, 6))
        plt.hist(results, bins=range(block_size + 1), alpha=0.7, density=True, edgecolor='black')
        plt.axvline(x=avg_diff, color='r', linestyle='--', linewidth=2, label=f'Orta: {avg_diff:.2f}')
        plt.axvline(x=ideal_avg, color='g', linestyle='--', linewidth=2, label=f'Ideal: {ideal_avg:.2f}')
        plt.xlabel('Fərqli bit sayı')
        plt.ylabel('Tezlik')
        plt.title('Lavina Effekti - Bit Dəyişikliyinin Paylanması')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.show()

    return results, avg_diff

# Test
results, avg = avalanche_analysis(cipher, 50)
````

## 📐 S-qutularının qeyri-xəttilik testi (15 dəq)

````python
def linear_approximation_test(sbox, input_bits, output_bits):
    """
    S-qutusu üçün xətti yaxınlaşma testi

    Qeyri-xətti S-qutusu üçün bütün yaxınlaşmaların ehtimalı 0.5-ə yaxın olmalıdır.
    Kiçik S-qutuları üçün maksimum bias ölçüdən asılıdır; məsələn, yaxşı 4×4
    bijektiv S-qutularında belə maksimum bias 0.25 ola bilər.

    Parametrlər:
        sbox: S-qutusu cədvəli
        input_bits: giriş bit sayı
        output_bits: çıxış bit sayı

    Qaytarır:
        max_bias: maksimum bias
        best_approx: ən yaxşı yaxınlaşma (input_mask, output_mask, bias)
    """
    input_size = 2**input_bits
    output_size = 2**output_bits
    is_bijective = (
        input_size == output_size
        and len(sbox) == input_size
        and len(set(sbox)) == input_size
        and all(0 <= y < output_size for y in sbox)
    )

    max_bias = 0
    best_approx = None

    # Bütün mümkün giriş maskaları üçün
    for input_mask in range(1, input_size):
        for output_mask in range(1, output_size):
            count = 0
            total = input_size

            # Bütün giriş dəyərləri üçün
            for x in range(input_size):
                y = sbox[x]

                # Xətti yaxınlaşmanı yoxla
                input_parity = bin(x & input_mask).count('1') % 2
                output_parity = bin(y & output_mask).count('1') % 2

                if input_parity == output_parity:
                    count += 1

            # Bias = |count/total - 0.5|
            bias = abs(count/total - 0.5)
            if bias > max_bias:
                max_bias = bias
                best_approx = (input_mask, output_mask, bias)

    print("=" * 80)
    print("📐 S-QUTUSU XƏTTİLİK TESTİ")
    print("=" * 80)
    print(f"Ölçü: {input_bits}×{output_bits}; bijektiv: {'bəli' if is_bijective else 'xeyr'}")
    print(f"Maksimum bias: {max_bias:.4f}")
    if best_approx:
        print(f"Ən yaxşı yaxınlaşma: input_mask={best_approx[0]:02x}, output_mask={best_approx[1]:02x}")

    eps = 1e-12
    if input_bits == 4 and output_bits == 4 and is_bijective:
        # 4×4 bijektiv S-qutuları üçün ən yaxşı məlum maksimum bias 0.25-dir
        # (komponent qeyri-xəttiliyi 4). Buna görə 0.1/0.2 kimi AES-ölçülü
        # sabit hədlər burada düzgün interpretasiya vermir.
        optimal_bias = 0.25
        print(f"4×4 bijektiv S-qutusu üçün istinad həddi: {optimal_bias:.4f}")
        if max_bias <= optimal_bias + eps:
            print("✅ Bu 4×4 bijektiv S-qutusu üçün çox yaxşı xətti dayanıqlıqdır")
        elif max_bias < 0.5 - eps:
            print("⚠️ Bu 4×4 S-qutusu optimal deyil, lakin ölçüyə görə ayrıca qiymətləndirilməlidir")
        else:
            print("❌ Güclü xətti yaxınlaşma var; S-qutusu zəifdir")
    elif input_bits == output_bits and is_bijective:
        reference_bias = 2 ** (-input_bits / 2)
        print(f"{input_bits}×{output_bits} bijektiv S-qutusu üçün ölçüdən asılı istinad: {reference_bias:.4f}")
        if max_bias <= reference_bias + eps:
            print("✅ S-qutusu bu ölçü üçün yaxşı qeyri-xəttiliyə malikdir")
        elif max_bias <= 2 * reference_bias + eps:
            print("⚠️ S-qutusu orta qeyri-xəttiliyə malikdir")
        else:
            print("❌ S-qutusu zəif qeyri-xəttiliyə malikdir")
    else:
        print("ℹ️ Bu ölçü və ya qeyri-bijektiv S-qutusu üçün universal sabit hədd verilmir;")
        print("   maksimum bias-ı eyni ölçülü alternativ S-qutuları ilə müqayisə edin.")

    return max_bias, best_approx

# Test
print("\n" + "=" * 80)
print("📐 S-QUTUSU QEYRİ-XƏTTİLİK TESTİ")
print("=" * 80)

# Təsadüfi S-qutusu yarat
sbox_random = create_sbox(4, 4)
bias_random, _ = linear_approximation_test(sbox_random, 4, 4)

# Xətti S-qutusu (şərti olaraq)
sbox_linear = list(range(16))
bias_linear, _ = linear_approximation_test(sbox_linear, 4, 4)

print(f"\nTəsadüfi S-qutusu maksimum bias: {bias_random:.4f}")
print(f"Xətti S-qutusu maksimum bias: {bias_linear:.4f}")
````

## 🔑 Açar generasiyası (15 dəq)

````python
def generate_round_keys(master_key, rounds, key_size, round_key_size):
    """
    Master açardan raund açarları yaradır
    """
    if rounds <= 0:
        raise ValueError("rounds müsbət olmalıdır")
    if key_size <= 0 or round_key_size <= 0:
        raise ValueError("Açar ölçüləri müsbət olmalıdır")
    if key_size < 3:
        raise ValueError("key_size ən azı 3 olmalıdır")
    if round_key_size > key_size:
        raise ValueError("Raund açarı master açardan böyük ola bilməz")

    round_keys = []
    full_mask = (1 << key_size) - 1
    mask = (1 << round_key_size) - 1
    key = master_key & full_mask

    for i in range(rounds):
        # Sadə bir açar generasiyası (tədris məqsədi üçün)
        round_key = (key ^ (key >> 8) ^ (key << 4)) & mask
        round_keys.append(round_key)

        # Açarı dəyişdir
        key = ((key << 3) | (key >> (key_size - 3))) & full_mask
        key ^= i + 1

    return round_keys

def deterministic_sbox_8x8(seed=0xC0FFEE):
    """
    Təkrarolunan 8x8 S-qutusu yaradır.

    Qeyd:
        create_sbox(8, 8) hər çağırışda fərqli nəticə verdiyi üçün
        default açar generasiyası da qeyri-deterministik olurdu.
        Burada lokal RNG ilə sabit, təkrarolunan bir S-qutusu qurulur.
    """
    rng = random.Random(seed)
    sbox = list(range(256))
    rng.shuffle(sbox)
    return sbox

def improved_key_schedule(master_key, rounds, round_key_size, sbox=None):
    """
    Təkmilləşdirilmiş açar generasiyası (S-qutusu istifadə edir)

    Qeyd:
        Eyni master açar üçün eyni raund açarları alınması üçün
        default S-qutusu deterministik saxlanılır.
    """
    if rounds <= 0:
        raise ValueError("rounds müsbət olmalıdır")
    if round_key_size <= 0 or round_key_size > 64:
        raise ValueError("round_key_size 1..64 aralığında olmalıdır")

    round_keys = []
    key = master_key & ((1 << 64) - 1)

    if sbox is None:
        sbox = deterministic_sbox_8x8()
    if len(sbox) != 256:
        raise ValueError("improved_key_schedule üçün 8x8 S-qutusu tələb olunur")
    if any((x < 0 or x >= 256) for x in sbox):
        raise ValueError("S-qutusu dəyərləri 0..255 aralığında olmalıdır")

    for i in range(rounds):
        # Açarın hər baytına S-qutusu tətbiq et
        transformed = 0
        for byte_pos in range(0, 64, 8):
            byte = (key >> byte_pos) & 0xFF
            transformed |= (sbox[byte] << byte_pos)

        round_key = transformed & ((1 << round_key_size) - 1)
        round_keys.append(round_key)

        # Açarı döndür və sabit əlavə et
        key = ((key << 13) | (key >> (64 - 13))) & ((1 << 64) - 1)
        key ^= i * 0x9E3779B97F4A7C15  # Sabit ədəd

    return round_keys

def key_schedule_test():
    """
    Açar generasiyasını test edir
    """
    print("\n" + "=" * 80)
    print("🔑 RAUND AÇARLARININ GENERASİYASI")
    print("=" * 80)

    master_key = 0x123456789ABCDEF0
    rounds = 16
    key_size = 64
    round_key_size = 16

    round_keys = generate_round_keys(master_key, rounds, key_size, round_key_size)

    print(f"Master açar: {master_key:016x}")
    print("\nRaund açarları:")
    for i, key in enumerate(round_keys):
        print(f"  K{i+1:2d}: {key:04x}")

    # Təkmilləşdirilmiş versiya
    print("\nTəkmilləşdirilmiş açar generasiyası:")
    improved_keys = improved_key_schedule(master_key, rounds, round_key_size)
    for i, key in enumerate(improved_keys):
        print(f"  K{i+1:2d}: {key:04x}")

    # Təkrarolunma testi
    repeated_keys = improved_key_schedule(master_key, rounds, round_key_size)
    print(f"\nDeterministik? {improved_keys == repeated_keys}")

key_schedule_test()
````
