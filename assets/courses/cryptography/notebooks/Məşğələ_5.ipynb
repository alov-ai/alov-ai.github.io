## 📖 Xatırlatma: Entropiya anlayışı (10 dəq)

<div style="background-color: #f0f8ff; padding: 15px; border-radius: 10px; border-left: 5px solid #4682b4;">
<h4>🔑 Şannon entropiyası</h4>
<p>Klod Şannonun informasiya nəzəriyyəsində <b>entropiya</b> təsadüfi dəyişənin qeyri-müəyyənliyini ölçür. Diskret təsadüfi dəyişən $X$ üçün entropiya:</p>

<p style="text-align: center; font-size: 1.2em;">$$H(X) = -\sum_{x} P(x) \log_2 P(x)$$</p>

<p>Entropiya <b>bitlərlə</b> ölçülür və $X$-in qiymətini təxmin etmək üçün lazım olan ortalama informasiya miqdarını göstərir.</p>
</div>

**Əsas xüsusiyyətlər:**

| Xüsusiyyət | İzah |
|------------|------|
| **Qeyri-mənfilik** | $H(X) \geq 0$, bərabərlik yalnız deterministik halda (bir qiymət ehtimalı 1) |
| **Maksimum entropiya** | $n$ mümkün qiyməti olan bircins paylanma üçün $H(X) = \log_2 n$ |
| **Müstəqillik** | İki müstəqil dəyişən üçün $H(X,Y) = H(X) + H(Y)$ |
| **Konkavlik** | Entropiya ehtimal paylanmasının konkav funksiyasıdır |

### Entropiya Düsturunun Komponentləri

Şannon entropiyası düsturu $H(X) = -\sum P(x) \log_2 P(x)$ aşağıdakı hissələrdən ibarətdir:

*   **$P(x)$ (Ehtimal)**:
    *   Bu, diskret təsadüfi dəyişən $X$-in müəyyən bir $x$ qiymətini alması ehtimalını ifadə edir.
    *   Məsələn, bir sikkə atışında `yazı` gəlməsi ehtimalı $P(\text{yazı}) = 0.5$ olar. Yaxud 6 üzlü zarda bir `3` rəqəminin düşmə ehtimalı $P(3) = 1/6$ olar.
    *   Entropiya hesablamaq üçün bütün mümkün $x$ qiymətləri üçün ehtimalları bilməliyik.

*   **$\log_2 P(x)$ (İnformasiya Miqdarı)**:
    *   Bu hissə, $X$ təsadüfi dəyişənin $x$ qiymətini alması haqqında informasiya miqdarını təmsil edir. Daha dəqiq desək, informasiya miqdarı $-\log_2 P(x)$ düsturu ilə verilir.
    *   $\log_2$ istifadə edildiyi üçün informasiya **bitlərlə** ölçülür.
    *   **Niyə mənfi logarifm?** Ehtimal $P(x)$ hər zaman 0 ilə 1 arasındadır. Buna görə də $\log_2 P(x)$ dəyəri həmişə mənfi və ya sıfır (ehtimal 1 olduqda) olacaq. İnformasiya miqdarı müsbət olmalıdır, ona görə də qarşısına mənfi işarəsi qoyulur.
    *   **İnformasiyanın intuitiv mənası:**
        *   Əgər bir hadisənin ehtimalı **çox yüksəkdirsə** (məsələn, $P(x) \approx 1$), deməli o hadisənin baş verməsi bizi çox təəccübləndirmir və bu barədə öyrəndiyimiz informasiya **azdır**. Məsələn, $P(x)=1$ olduqda, $-\log_2 1 = 0$ bit informasiya var.
        *   Əgər bir hadisənin ehtimalı **çox aşağıdırsa** (məsələn, $P(x) \approx 0$), deməli o hadisənin baş verməsi bizi çox təəccübləndirir və bu barədə öyrəndiyimiz informasiya **çoxdur**. Məsələn, $P(x)=0.01$ olduqda, $-\log_2 0.01 \approx 6.64$ bit informasiya var.

Entropiya düsturu $H(X) = -\sum P(x) \log_2 P(x)$, təsadüfi dəyişənin hər bir mümkün qiymətinin ehtimalı ilə onun verdiyi informasiya miqdarını ($-\log_2 P(x)$) çarpıb, sonra bu məhsulların cəmini götürərək, **orta informasiya miqdarını** (və ya qeyri-müəyyənliyi) hesablayır.

## 🧮 Entropiya hesablamaları (20 dəq)

### 5.1 Əsas entropiya funksiyası

````python
def entropy(probabilities):
    """
    Diskret paylanmanın entropiyasını hesablayır

    Parametrlər:
        probabilities: ehtimallar siyahısı (cəmi 1 olmalıdır)

    Qaytarır:
        entropiya (bit)
    """
    entropy_value = 0
    for p in probabilities:
        if p > 0:
            entropy_value -= p * math.log2(p)

    return entropy_value

def entropy_from_counts(counts):
    """
    Tezlik saylarından entropiyanı hesablayır

    Parametrlər:
        counts: hər bir qiymətin sayı

    Qaytarır:
        entropiya (bit)
    """
    total = sum(counts)
    if total == 0:
        return 0
    probabilities = [c / total for c in counts]
    return entropy(probabilities)

def entropy_from_text(text):
    """
    Mətndəki hərflərin tezliyinə əsasən entropiyanı hesablayır

    Parametrlər:
        text: mətn

    Qaytarır:
        entropiya (bit)
    """
    # Yalnız hərfləri götür
    letters = [c.upper() for c in text if c.isalpha()]
    total = len(letters)

    if total == 0:
        return 0

    # Hərflərin tezliyini hesabla
    freq = Counter(letters)
    probabilities = [count / total for count in freq.values()]

    return entropy(probabilities)

# Test nümunələri
print("=" * 60)
print("📊 ENTROPİYA HESABLAMALARI")
print("=" * 60)

# Bircins paylanma (6 üzlü zar)
p_uniform = [1/6] * 6
h_uniform = entropy(p_uniform)
print(f"🎲 Bircins zar (6 üzlü): H = {h_uniform:.4f} bit")
print(f"   log2(6) = {math.log2(6):.4f} bit")

# Qərəzli sikkə (p=0.9)
p_biased = [0.9, 0.1]
h_biased = entropy(p_biased)
print(f"🪙 Qərəzli sikkə (p=0.9): H = {h_biased:.4f} bit")

# Bircins sikkə (p=0.5)
p_fair = [0.5, 0.5]
h_fair = entropy(p_fair)
print(f"🪙 Bircins sikkə (p=0.5): H = {h_fair:.4f} bit")

# Mətn nümunəsi
text = "Bu bir test mətnidir. Burada Azərbaycan hərfləri də var: ç, ş, ğ, ö, ü, ı"
h_text = entropy_from_text(text)
print(f"📝 Mətn entropiyası: {h_text:.4f} bit")
````

### 5.2 Müxtəlif paylanmaların entropiyası

````python
def compare_distributions():
    """
    Müxtəlif paylanmaların entropiyasını müqayisə edir
    """
    distributions = {
        "Deterministik (həmişə A)": [1.0],
        "Qərəzli sikkə (p=0.9)": [0.9, 0.1],
        "Qərəzli sikkə (p=0.75)": [0.75, 0.25],
        "Qərəzli sikkə (p=0.6)": [0.6, 0.4],
        "Bircins sikkə (p=0.5)": [0.5, 0.5],
        "4 üzlü bircins zar": [0.25, 0.25, 0.25, 0.25],
        "6 üzlü bircins zar": [1/6] * 6,
        "8 üzlü bircins zar": [1/8] * 8,
        "26 üzlü (əlifba)": [1/26] * 26,
        "32 üzlü (Azərbaycan əlifbası)": [1/32] * 32,
    }

    print("=" * 70)
    print("📊 MÜXTƏLİF PAYLANMALARIN ENTROPİYASI")
    print("=" * 70)
    print(f"{'Paylanma':<30} | {'Entropiya (bit)':<15} | {'Maksimum':<10}")
    print("-" * 70)

    for name, probs in distributions.items():
        h = entropy(probs)
        max_h = math.log2(len(probs)) if len(probs) > 1 else 0
        bar = '█' * int(30 * h / max_h) if max_h > 0 else ''
        print(f"{name:<30} | {h:<15.4f} | {max_h:<10.2f} {bar}")

    print("=" * 70)

compare_distributions()
````

### 5.3 Entropiya qrafiki (ikili seçim)

````python
def plot_binary_entropy():
    """
    İkili seçim üçün entropiya funksiyasının qrafikini çəkir
    H(p) = -p log2(p) - (1-p) log2(1-p)
    """
    if not HAS_MATPLOTLIB:
        print("⚠️ matplotlib yüklü deyil, qrafik çəkilə bilməz")
        return

    p_values = np.linspace(0.01, 0.99, 100)
    h_values = [-p * math.log2(p) - (1-p) * math.log2(1-p) for p in p_values]

    plt.figure(figsize=(12, 7))
    plt.plot(p_values, h_values, 'b-', linewidth=3)

    # Əhəmiyyətli nöqtələri işarələ
    plt.plot(0.5, 1, 'ro', markersize=10, label='Maksimum (p=0.5, H=1)')
    plt.plot(0.1, entropy([0.1, 0.9]), 'go', markersize=8, label='p=0.1')
    plt.plot(0.2, entropy([0.2, 0.8]), 'go', markersize=8)
    plt.plot(0.3, entropy([0.3, 0.7]), 'go', markersize=8)
    plt.plot(0.4, entropy([0.4, 0.6]), 'go', markersize=8)

    plt.xlabel('p (Ehtimal)', fontsize=12)
    plt.ylabel('Entropiya H(p) (bit)', fontsize=12)
    plt.title('📈 İkili Entropiya Funksiyası', fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.axvline(x=0.5, color='r', linestyle='--', alpha=0.5)
    plt.axhline(y=1, color='r', linestyle='--', alpha=0.5)
    plt.legend()
    plt.xlim(0, 1)
    plt.ylim(0, 1.1)
    plt.show()

# Qrafiki çəkək
if HAS_MATPLOTLIB:
    plot_binary_entropy()
````

## 🔗 Birgə entropiya və şərti entropiya (20 dəq)

### 6.1 Birgə entropiya

<div style="background-color: #f0f8ff; padding: 10px; border-radius: 5px; border-left: 3px solid #4682b4;">
<p><b>Birgə entropiya:</b> İki təsadüfi dəyişən $X$ və $Y$ üçün birgə entropiya:</p>
<p style="text-align: center;">$$H(X,Y) = -\sum_{x}\sum_{y} P(x,y) \log_2 P(x,y)$$</p>
</div>

````python
def joint_entropy(joint_probs):
    """
    Birgə entropiyanı hesablayır

    Parametrlər:
        joint_probs: 2D siyahı, joint_probs[x][y] = P(X=x, Y=y)

    Qaytarır:
        H(X,Y) birgə entropiya
    """
    h_joint = 0
    for row in joint_probs:
        for p in row:
            if p > 0:
                h_joint -= p * math.log2(p)

    return h_joint

# Nümunə: İki sikkə atışı
# X: birinci sikkə, Y: ikinci sikkə
print("\n" + "=" * 60)
print("🔗 BİRGƏ ENTROPİYA NÜMUNƏLƏRİ")
print("=" * 60)

# Müstəqil sikkələr
joint_independent = [[0.25, 0.25], [0.25, 0.25]]  # P(0,0)=0.25, P(0,1)=0.25, ...
h_joint_ind = joint_entropy(joint_independent)
h_x = entropy([0.5, 0.5])
h_y = entropy([0.5, 0.5])
print(f"\n🎲 Müstəqil sikkələr:")
print(f"   H(X) = {h_x:.4f} bit")
print(f"   H(Y) = {h_y:.4f} bit")
print(f"   H(X,Y) = {h_joint_ind:.4f} bit")
print(f"   H(X) + H(Y) = {h_x + h_y:.4f} bit")
print(f"   Müstəqillik şərti ödənir: H(X,Y) = H(X) + H(Y)")
````

### 6.2 Şərti entropiya

<div style="background-color: #f0f8ff; padding: 10px; border-radius: 5px; border-left: 3px solid #4682b4;">
<p><b>Şərti entropiya:</b> $Y$ məlum olduqda $X$-in şərti entropiyası:</p>
<p style="text-align: center;">$$H(X|Y) = -\sum_{x,y} P(x,y) \log_2 P(x|y) = H(X,Y) - H(Y)$$</p>
</div>

````python
def conditional_entropy(joint_probs):
    """
    H(X|Y) şərti entropiyasını hesablayır

    Parametrlər:
        joint_probs: 2D siyahı, joint_probs[x][y] = P(X=x, Y=y)

    Qaytarır:
        h_cond: H(X|Y)
        h_joint: H(X,Y)
        h_y: H(Y)
    """
    n_x = len(joint_probs)
    n_y = len(joint_probs[0])

    # P(Y) marjinalını hesabla
    p_y = [0] * n_y
    for x in range(n_x):
        for y in range(n_y):
            p_y[y] += joint_probs[x][y]

    # H(Y)-ni hesabla
    h_y = entropy(p_y)

    # Birgə entropiyanı hesabla
    h_joint = joint_entropy(joint_probs)

    # Şərti entropiya: H(X|Y) = H(X,Y) - H(Y)
    h_cond = h_joint - h_y

    return h_cond, h_joint, h_y

# Nümunə: Asılı dəyişənlər
# Tutaq ki, X və Y arasında əlaqə var
joint_probs_dependent = [
    [0.4, 0.1],  # P(X=0, Y=0)=0.4, P(X=0, Y=1)=0.1
    [0.1, 0.4]   # P(X=1, Y=0)=0.1, P(X=1, Y=1)=0.4
]

print("\n" + "=" * 60)
print("🔗 ŞƏRTİ ENTROPİYA NÜMUNƏSİ")
print("=" * 60)
h_cond, h_joint, h_y = conditional_entropy(joint_probs_dependent)

# P(X) marjinalını hesablayaq
p_x = [joint_probs_dependent[0][0] + joint_probs_dependent[0][1],
       joint_probs_dependent[1][0] + joint_probs_dependent[1][1]]
h_x = entropy(p_x)

print(f"\n📊 Asılı dəyişənlər üçün:")
print(f"   H(X) = {h_x:.4f} bit")
print(f"   H(Y) = {h_y:.4f} bit")
print(f"   H(X,Y) = {h_joint:.4f} bit")
print(f"   H(X|Y) = {h_cond:.4f} bit")
print(f"   H(X) - H(X|Y) = {h_x - h_cond:.4f} bit (qarşılıqlı informasiya)")
print(f"\n📌 Y məlum olduqda X-in qeyri-müəyyənliyi azaldı: {h_x:.4f} → {h_cond:.4f}")
````

### 6.3 Qarşılıqlı informasiya

<div style="background-color: #f0f8ff; padding: 10px; border-radius: 5px; border-left: 3px solid #4682b4;">
<p><b>Qarşılıqlı informasiya:</b> $X$ və $Y$ arasındakı qarşılıqlı informasiya:</p>
<p style="text-align: center;">$$I(X;Y) = H(X) - H(X|Y) = H(Y) - H(Y|X) = H(X) + H(Y) - H(X,Y)$$</p>
<p>Bu, bir dəyişənin digəri haqqında verdiyi informasiya miqdarını ölçür.</p>
</div>

````python
def mutual_information(joint_probs):
    """
    Qarşılıqlı informasiyanı hesablayır I(X;Y)

    Parametrlər:
        joint_probs: 2D siyahı, joint_probs[x][y] = P(X=x, Y=y)

    Qaytarır:
        mi: I(X;Y)
        h_x: H(X)
        h_y: H(Y)
        h_joint: H(X,Y)
    """
    n_x = len(joint_probs)
    n_y = len(joint_probs[0])

    # Marjinalları hesabla
    p_x = [0] * n_x
    p_y = [0] * n_y

    for x in range(n_x):
        for y in range(n_y):
            p = joint_probs[x][y]
            p_x[x] += p
            p_y[y] += p

    # Fərdi entropiyalar
    h_x = entropy(p_x)
    h_y = entropy(p_y)

    # Birgə entropiya
    h_joint = joint_entropy(joint_probs)

    # Qarşılıqlı informasiya
    mi = h_x + h_y - h_joint

    return mi, h_x, h_y, h_joint

# Müqayisə: müstəqil vs asılı dəyişənlər
joint_independent = [[0.25, 0.25], [0.25, 0.25]]
joint_dependent = [[0.4, 0.1], [0.1, 0.4]]

mi_ind, hx_ind, hy_ind, hjoint_ind = mutual_information(joint_independent)
mi_dep, hx_dep, hy_dep, hjoint_dep = mutual_information(joint_dependent)

print("\n" + "=" * 70)
print("🔄 QARŞILIQLI İNFORMASİYA MÜQAYİSƏSİ")
print("=" * 70)
print(f"{'Dəyişənlər':<15} | {'H(X)':<8} | {'H(Y)':<8} | {'H(X,Y)':<10} | {'I(X;Y)':<10} | {'Nisbi'}")
print("-" * 75)
print(f"{'Müstəqil':<15} | {hx_ind:<8.4f} | {hy_ind:<8.4f} | {hjoint_ind:<10.4f} | {mi_ind:<10.4f} | {mi_ind/hx_ind:.2f}")
print(f"{'Asılı':<15} | {hx_dep:<8.4f} | {hy_dep:<8.4f} | {hjoint_dep:<10.4f} | {mi_dep:<10.4f} | {mi_dep/hx_dep:.2f}")
print("=" * 75)
print("📌 Asılı dəyişənlərdə qarşılıqlı informasiya böyükdür (bir-biri haqqında çox məlumat verir)")
print("📌 Müstəqil dəyişənlərdə I(X;Y)=0 (bir-biri haqqında heç nə demir)")
````

## 🔐 Bir dəfəlik bloknot (OTP) implementasiyası (20 dəq)

<div style="background-color: #f0f8ff; padding: 15px; border-radius: 10px; border-left: 5px solid #4682b4;">
<h4>🔑 Bir dəfəlik bloknot (One-Time Pad)</h4>
<p>OTP <b>mükəmməl məxfiliyin</b> klassik və ən sadə nümunəsidir. Shannon mənasında mükəmməl məxfilik o deməkdir ki, ciphertext mesaj haqqında əlavə məlumat vermir:</p>
<p>$$\Pr[M=m \mid C=c] = \Pr[M=m]$$</p>
<p>OTP-də bu xassənin ödənməsi üçün şərtlər:</p>
<ol>
    <li>Açar həqiqətən <b>təsadüfi</b> olmalıdır</li>
    <li>Açar mesajla <b>eyni uzunluqda</b> olmalıdır</li>
    <li>Açar heç vaxt <b>təkrar istifadə</b> edilməməlidir</li>
    <li>Açar təhlükəsiz şəkildə <b>paylanmalıdır</b></li>
</ol>
<p><b>Şifrləmə:</b> $c = m \oplus k$</p>
<p><b>Deşifrləmə:</b> $m = c \oplus k$</p>
</div>

````python
def otp_encrypt_decrypt(message, key):
    """
    Bir dəfəlik bloknot (OTP) şifrləmə/deşifrləmə
    XOR əməliyyatına əsaslanır

    Parametrlər:
        message: şifrlənəcək məlumat (str və ya bytes)
        key: açar (bytes)

    Qaytarır:
        result: şifrlənmiş/deşifrlənmiş məlumat (bytes)
    """
    # Mesajı bayta çevir
    if isinstance(message, str):
        message = message.encode('utf-8')

    # Açar mesajla eyni uzunluqda olmalıdır
    if len(key) != len(message):
        raise ValueError(f"❌ Açar uzunluğu ({len(key)}) mesaj uzunluğuna ({len(message)}) bərabər olmalıdır!")

    # XOR əməliyyatı
    result = bytes([a ^ b for a, b in zip(message, key)])

    return result

def generate_random_key(length):
    """
    Kriptoqrafik təhlükəsiz təsadüfi açar generasiya edir

    Parametrlər:
        length: açar uzunluğu (bayt)

    Qaytarır:
        key: təsadüfi açar (bytes)
    """
    # secrets modulu kriptoqrafik təhlükəsiz təsadüfi ədədlər üçündür
    return secrets.token_bytes(length)

def otp_demo():
    """
    OTP nümayişi
    """
    print("=" * 70)
    print("🔐 BİR DƏFƏLİK BLOKNOT (OTP) NÜMAYİŞİ")
    print("=" * 70)

    message = "Çox gizli məlumat - OTP test"
    message_bytes = message.encode('utf-8')

    print(f"\n📨 Orijinal mesaj: {message}")
    print(f"📏 Mesaj uzunluğu: {len(message_bytes)} bayt")

    # Təsadüfi açar yarat
    key = generate_random_key(len(message_bytes))
    print(f"🔑 Açar (hex): {key.hex()[:50]}..." if len(key.hex()) > 50 else f"🔑 Açar (hex): {key.hex()}")

    # Şifrlə
    ciphertext = otp_encrypt_decrypt(message_bytes, key)
    print(f"🔒 Şifrəli mətn (hex): {ciphertext.hex()[:50]}..." if len(ciphertext.hex()) > 50 else f"🔒 Şifrəli mətn (hex): {ciphertext.hex()}")

    # Deşifrlə
    decrypted = otp_encrypt_decrypt(ciphertext, key)
    decrypted_text = decrypted.decode('utf-8')

    print(f"🔓 Deşifrələnmiş: {decrypted_text}")
    print(f"\n✅ Uğurlu? {message == decrypted_text}")

    return message_bytes, key, ciphertext

# Test
msg, key, ct = otp_demo()
````

### 7.2 Mükəmməl məxfiliyin yoxlanması

````python
def perfect_secrecy_demo():
    """
    OTP-nin mükəmməl məxfiliyini nümayiş etdirir
    P(M|C) = P(M) olduğunu göstərir
    """
    print("\n" + "=" * 80)
    print("🎭 MÜKƏMMƏL MƏXFİLİK NÜMAYİŞİ")
    print("=" * 80)

    # İki mümkün mesaj
    m1 = b"Attack at dawn!"
    m2 = b"Retreat at dusk"

    print(f"📨 Mesaj 1: {m1}")
    print(f"📨 Mesaj 2: {m2}")

    # Təsadüfi bir açar seçək
    key = generate_random_key(len(m1))

    # Hər iki mesajı bu açarla şifrlə (eyni uzunluqda olduqları üçün)
    c1 = otp_encrypt_decrypt(m1, key)
    c2 = otp_encrypt_decrypt(m2, key)

    print(f"\n🔑 Eyni açarla şifrlənmiş mesajlar:")
    print(f"   c1: {c1.hex()}")
    print(f"   c2: {c2.hex()}")
    print(f"   c1 və c2 fərqlidir? {c1 != c2}")

    # İndi fərqli açarlarla şifrləyək
    key1 = generate_random_key(len(m1))
    key2 = generate_random_key(len(m2))

    c1_diff = otp_encrypt_decrypt(m1, key1)
    c2_diff = otp_encrypt_decrypt(m2, key2)

    print(f"\n🔑 Fərqli açarlarla şifrlənmiş mesajlar:")
    print(f"   c1: {c1_diff.hex()}")
    print(f"   c2: {c2_diff.hex()}")

    print("\n" + "-" * 80)
    print("📌 NƏZƏRİ İZAH:")
    print("OTP-də hər bir şifrəli mətn istənilən mesajdan yarana bilər,")
    print("çünki hər bir c üçün elə bir k var ki, c = m ⊕ k")
    print("Buna görə də P(M|C) = P(M) - mükəmməl məxfilik təmin edilir.")
    print("-" * 80)
    print("🎯 NƏTİCƏ: OTP mükəmməl məxfilik təmin edir!")

perfect_secrecy_demo()
````

### ✍️ Çalışma 3: OTP implementasiyası (1.5 bal)

Aşağıdakı tapşırıqları yerinə yetirin:

````python
# Çalışma 3 - Cavablar

print("📝 ÇALIŞMA 3 CAVABLARI")
print("=" * 80)

# 1. Şəxsi mesajın şifrlənməsi
print("\n1. ŞƏXSİ MESAJIN ŞİFRLƏNMƏSİ:")
my_message = "Səməd Vəkilov"  # Öz adınızla əvəz edin
msg_bytes = my_message.encode('utf-8')
key_personal = generate_random_key(len(msg_bytes))
cipher_personal = otp_encrypt_decrypt(msg_bytes, key_personal)
print(f"   Mesaj: {my_message}")
print(f"   Açar (hex): {key_personal.hex()}")
print(f"   Şifrəli (hex): {cipher_personal.hex()}")

# 2. Açar faylı yaratmaq
print("\n2. AÇAR FAYLININ YARADILMASI:")
key_32 = generate_random_key(32)
with open('key_32.key', 'wb') as f:
    f.write(key_32)
print(f"   32 baytlıq açar yaradıldı və 'key_32.key' faylına yazıldı")
print(f"   Açar (hex): {key_32.hex()}")

# 3-4. Fayl şifrləmə və deşifrləmə
print("\n3-4. FAYL ŞİFRLƏMƏ VƏ DEŞİFRLƏMƏ:")
test_content = "Bu bir test faylıdır. OTP ilə şifrlənəcək."
with open('test.txt', 'w', encoding='utf-8') as f:
    f.write(test_content)

# Faylı oxu və şifrlə
with open('test.txt', 'rb') as f:
    plain_data = f.read()
key_file = generate_random_key(len(plain_data))
encrypted_data = otp_encrypt_decrypt(plain_data, key_file)
with open('test.enc', 'wb') as f:
    f.write(encrypted_data)
print(f"   'test.txt' faylı şifrləndi -> 'test.enc'")

# Deşifrlə
with open('test.enc', 'rb') as f:
    enc_data = f.read()
decrypted_data = otp_encrypt_decrypt(enc_data, key_file)
with open('test.dec', 'wb') as f:
    f.write(decrypted_data)
print(f"   'test.enc' faylı deşifrləndi -> 'test.dec'")

# Yoxlama
with open('test.dec', 'r', encoding='utf-8') as f:
    decrypted_content = f.read()
print(f"   Deşifrlənmiş məzmun: {decrypted_content}")
print(f"   Uğurlu? {test_content == decrypted_content}")

# 5. Praktik çatışmazlıqlar
print("\n5. OTP-NİN PRAKTİK ÇATIŞMAZLIQLARI:")
print("   • Açar paylanması problemi - açar mesajla eyni uzunluqda və təhlükəsiz paylanmalıdır")
print("   • Açar idarəçiliyi - hər yeni mesaj üçün yeni açar lazımdır")
print("   • Açarın təhlükəsiz saxlanması - açar mesaj qədər qorunmalıdır")
print("   • Açar təkrarı riski - eyni açarın təkrarı təhlükəsizliyi tamamilə məhv edir")
print("   • Böyük həcmli məlumat üçün yararsızdır - 1 GB məlumat üçün 1 GB açar lazımdır")
````

## 🔁 Açar təkrarı probleminin simulyasiyası (20 dəq)

<div style="background-color: #fff3e0; padding: 15px; border-radius: 10px; border-left: 5px solid #ff9800;">
<h4>⚠️ Açar təkrarı problemi</h4>
<p>OTP-nin ən vacib şərti açarın heç vaxt təkrar istifadə edilməməsidir. Əgər eyni açar iki fərqli mesajı şifrləmək üçün istifadə edilərsə:</p>
<p style="text-align: center;">$$c_1 = m_1 \oplus k, \quad c_2 = m_2 \oplus k$$</p>
<p style="text-align: center;">$$c_1 \oplus c_2 = (m_1 \oplus k) \oplus (m_2 \oplus k) = m_1 \oplus m_2$$</p>
<p>Bu, hücumçuya mesajlar haqqında məlumat verir. Hətta mesajların özləri bilinməsə belə, statistik analiz mümkündür.</p>
</div>

````python
def key_reuse_attack_demo():
    """
    Açar təkrarı problemini simulyasiya edir
    """
    print("=" * 80)
    print("⚠️ AÇAR TƏKRARI PROBLEMİ - HÜCUM SİMULYASİYASI")
    print("=" * 80)

    # Eyni açarla iki mesaj şifrlənir
    m1 = b"Top secret message A"
    m2 = b"Classified message B"

    print(f"\n📨 Mesaj 1: {m1}")
    print(f"📨 Mesaj 2: {m2}")

    # Eyni açar
    key = generate_random_key(len(m1))  # Açar uzunluğu m1-ə bərabər

    c1 = otp_encrypt_decrypt(m1, key)
    c2 = otp_encrypt_decrypt(m2, key)

    print(f"\n🔒 Şifrəli mətn 1: {c1.hex()}")
    print(f"🔒 Şifrəli mətn 2: {c2.hex()}")

    # Hücumçu c1 və c2-ni ələ keçirir
    print("\n" + "-" * 80)
    print("🕵️ HÜCUMÇU (Eve) c1 və c2-ni ələ keçirir")
    print("-" * 80)

    # c1 ⊕ c2 hesablanır
    xor_c1_c2 = bytes([a ^ b for a, b in zip(c1, c2)])
    print(f"\n🔢 c1 ⊕ c2: {xor_c1_c2.hex()}")

    # m1 ⊕ m2 hesablanır (əgər hücumçu m1-i bilirsə)
    xor_m1_m2 = bytes([a ^ b for a, b in zip(m1, m2)])
    print(f"🔢 m1 ⊕ m2: {xor_m1_m2.hex()}")

    # Onlar bərabərdir?
    print(f"\n✅ c1 ⊕ c2 = m1 ⊕ m2? {xor_c1_c2 == xor_m1_m2}")

    # Tutaq ki, hücumçu m1-i bilir
    print("\n" + "-" * 80)
    print("📌 FƏRZ EDƏK Kİ, HÜCUMÇU m1-i BİLİR")
    print("-" * 80)

    # O zaman m2-ni tapa bilər: m2 = (c1 ⊕ c2) ⊕ m1
    recovered_m2 = bytes([a ^ b for a, b in zip(xor_c1_c2, m1)])
    print(f"\n🔓 Bərpa edilmiş m2: {recovered_m2}")
    print(f"📨 Orijinal m2: {m2}")
    print(f"✅ Bərpa uğurlu? {recovered_m2 == m2}")

    # Hətta m1 bilinməsə belə, statistik analiz mümkündür
    print("\n" + "-" * 80)
    print("📊 STATİSTİK ANALİZ")
    print("-" * 80)

    # xor_c1_c2 əslində m1 ⊕ m2-dir
    # İngilis dilində mətnlər üçün bu, tezlik analizinə imkan verir
    print("c1 ⊕ c2 = m1 ⊕ m2 olduğundan, hücumçu statistik analiz apara bilər.")
    print("Məsələn, ingilis dilində ən çox işlənən hərf 'space' (32) və 'e' (101)-dir.")

    # XOR nəticəsində təkrarlanan baytları aşkarlayaq
    byte_counts = Counter(xor_c1_c2)
    print("\n📈 XOR nəticəsində tez-tez təkrarlanan baytlar:")
    for byte, count in sorted(byte_counts.items(), key=lambda x: x[1], reverse=True)[:5]:
        if count > 1:
            print(f"   0x{byte:02x}: {count} dəfə")

    print("\n" + "=" * 80)
    print("🎯 NƏTİCƏ: Açar təkrarı OTP-nin mükəmməl məxfiliyini tamamilə məhv edir!")
    print("=" * 80)

key_reuse_attack_demo()
````

### 8.2 Tarixi nümunə: Venona layihəsi

<div style="background-color: #fff3e0; padding: 15px; border-radius: 10px; border-left: 5px solid #ff9800;">
<h4>📜 Venona layihəsi (1943-1980)</h4>
<p>ABŞ-ın <b>Venona layihəsi</b> SSRİ-nin OTP istifadəsinə qarşı kəşfiyyat əməliyyatı idi. SSRİ bəzi açarları təkrar istifadə etdiyinə görə, ABŞ bu zəiflikdən istifadə edərək bir çox gizli mesajları deşifrələyə bildi.</p>
<p>Bu, açar təkrarı probleminin real dünyada ən məşhur nümunəsidir. Venona layihəsi sayəsində ABŞ sovet kəşfiyyat şəbəkəsi haqqında mühüm məlumatlar əldə etdi.</p>
</div>

````python
def analyze_xor_of_two_messages():
    """
    İki mesajın XOR-unu analiz edir - statistik metodlar
    """
    # İngilis dilində iki cümlə
    m1 = b"This is a secret message that contains sensitive information"
    m2 = b"The enemy will attack at dawn with full force"

    # Eyni açarla şifrlənmiş kimi düşünək
    xor_result = bytes([a ^ b for a, b in zip(m1, m2)])

    print("\n" + "=" * 80)
    print("🔬 İKİ MESAJIN XOR ANALİZİ")
    print("=" * 80)
    print(f"📨 m1: {m1}")
    print(f"📨 m2: {m2}")
    print(f"\n🔢 m1 ⊕ m2: {xor_result.hex()}")

    # XOR nəticəsində hər bir baytın paylanması
    byte_counts = Counter(xor_result)
    print("\n📊 Bayt paylanması:")
    for byte, count in sorted(byte_counts.items(), key=lambda x: x[1], reverse=True)[:10]:
        percentage = count / len(xor_result) * 100
        bar = '█' * int(percentage * 2)
        print(f"   0x{byte:02x}: {count:3d} dəfə ({percentage:.1f}%) {bar}")

    # XOR nəticəsində 0 olan yerlər - eyni simvollar
    same_positions = []
    for i, b in enumerate(xor_result):
        if b == 0:
            same_positions.append((i, m1[i], m2[i]))

    if same_positions:
        print("\n🔍 Eyni simvollar (XOR=0 olan yerlər):")
        for pos, c1, c2 in same_positions[:10]:
            print(f"   mövqe {pos:2d}: '{chr(c1)}' ({c1}) = '{chr(c2)}' ({c2})")
        if len(same_positions) > 10:
            print(f"   ... və daha {len(same_positions)-10} yer")

    print("\n" + "-" * 80)
    print("📌 STATİSTİK NƏTİCƏ:")
    print("XOR analizi mesajlar haqqında əhəmiyyətli məlumat verir!")
    print("Təkrarlanan baytlar eyni simvolları, tezlik paylanması isə")
    print("mesajların strukturunu aşkar etməyə kömək edir.")

analyze_xor_of_two_messages()
````

````python
def entropy_lab_menu():
    """
    Entropiya laboratoriyası interaktiv menyu
    """
    while True:
        print("\n" + "=" * 60)
        print("🧮 ENTROPİYA LABORATORİYASI - Məşğələ 5")
        print("=" * 60)
        print("1. 📊 Entropiya hesabla (ehtimallardan)")
        print("2. 📝 Entropiya hesabla (mətndən)")
        print("3. 🔗 Birgə/şərti entropiya hesabla")
        print("4. 🔐 OTP şifrləmə nümayişi")
        print("5. ⚠️ Açar təkrarı hücumu simulyasiyası")
        print("6. 📈 Müxtəlif paylanmaları müqayisə et")
        print("7. 📉 İkili entropiya qrafiki")
        print("8. 🔬 İki mesajın XOR analizi")
        print("0. ❌ Çıxış")
        print("=" * 60)

        choice = input("📌 Seçiminiz: ")

        if choice == '1':
            print("\nEhtimalları daxil edin (vergüllə ayırın, cəmi 1 olmalıdır):")
            probs_str = input("Məsələn: 0.5,0.5: ")
            try:
                probs = [float(p.strip()) for p in probs_str.split(',')]
                if abs(sum(probs) - 1.0) > 0.001:
                    print("❌ Ehtimalların cəmi 1 olmalıdır!")
                else:
                    h = entropy(probs)
                    print(f"✅ Entropiya: {h:.4f} bit")
            except:
                print("❌ Yanlış format!")

        elif choice == '2':
            text = input("📨 Mətn daxil edin: ")
            h = entropy_from_text(text)
            print(f"✅ Mətn entropiyası: {h:.4f} bit")

        elif choice == '3':
            print("\n2x2 birgə paylanma üçün ehtimalları daxil edin:")
            try:
                p00 = float(input("P(X=0,Y=0): "))
                p01 = float(input("P(X=0,Y=1): "))
                p10 = float(input("P(X=1,Y=0): "))
                p11 = float(input("P(X=1,Y=1): "))

                joint = [[p00, p01], [p10, p11]]
                if abs(sum([p00, p01, p10, p11]) - 1.0) > 0.001:
                    print("❌ Ehtimalların cəmi 1 olmalıdır!")
                else:
                    mi, hx, hy, hjoint = mutual_information(joint)
                    print(f"\n📊 NƏTİCƏLƏR:")
                    print(f"   H(X): {hx:.4f} bit")
                    print(f"   H(Y): {hy:.4f} bit")
                    print(f"   H(X,Y): {hjoint:.4f} bit")
                    print(f"   I(X;Y): {mi:.4f} bit")
            except:
                print("❌ Yanlış format!")

        elif choice == '4':
            message = input("📨 Mesaj daxil edin: ")
            msg_bytes = message.encode('utf-8')
            key = generate_random_key(len(msg_bytes))
            cipher = otp_encrypt_decrypt(msg_bytes, key)
            decrypted = otp_encrypt_decrypt(cipher, key)

            print(f"\n🔑 Açar (hex): {key.hex()}")
            print(f"🔒 Şifrəli mətn (hex): {cipher.hex()}")
            print(f"🔓 Deşifrələnmiş: {decrypted.decode('utf-8')}")

        elif choice == '5':
            key_reuse_attack_demo()

        elif choice == '6':
            compare_distributions()

        elif choice == '7':
            if HAS_MATPLOTLIB:
                plot_binary_entropy()
            else:
                print("⚠️ matplotlib yüklü deyil")

        elif choice == '8':
            analyze_xor_of_two_messages()

        elif choice == '0':
            print("👋 Proqram bitdi. Sağ olun!")
            break

        else:
            print("❌ Yanlış seçim. Yenidən cəhd edin.")

# Proqramı işə sal (istəyə bağlı)
# entropy_lab_menu()
````
