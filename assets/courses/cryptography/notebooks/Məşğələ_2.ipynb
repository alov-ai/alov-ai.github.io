### 4.1 Xatırlatma: Modul əməliyyatları

Mühazirədə öyrəndiyimiz kimi, klassik şifrələr modul arifmetikaya əsaslanır:

<div style="background-color: #f0f8ff; padding: 15px; border-radius: 10px; border-left: 5px solid #4682b4;">

#### 🔑 Modul arifmetika əsasları

- $a \equiv b \pmod{m}$ o deməkdir ki, $a-b$ fərqi $m$-ə bölünür. Eyni zamanda bu o deməkdir ki, a və b ədədlərinin m-ə böləndə qalıqlar eynidir. $a \mod m$  $a$-nı $m$-ə böləndə alınan qalıq deməkdir  
- Toplama: $(a+b) \bmod m$  
- Çıxma: $(a-b) \bmod m = (a-b+m) \bmod m$  *(mənfi nəticələrdən qaçmaq üçün)*  
- Vurma: $(a \cdot b) \bmod m$

</div>

## 🔤 Sezar şifrəsinin implementasiyası (20 dəq)

### 5.1 Əlifbanın rəqəmlərə çevrilməsi

Sezar şifrəsində hər hərf 0-25 arası rəqəmlə təmsil olunur. `ord()` funksiyası simvolun Unicode kod nöqtəsini qaytarır (məsələn, `ord('A') == 65`). A=0, ..., Z=25 almaq üçün `ord(letter.upper()) - ord('A')` hesablayırıq:

````python
def letter_to_num(letter):
    """Hərfi rəqəmə çevirir (A=0, B=1, ..., Z=25)"""
    return ord(letter.upper()) - ord('A')

def num_to_letter(num):
    """Rəqəmi hərfə çevirir (0=A, 1=B, ..., 25=Z)"""
    return chr(num + ord('A'))

# Test
print(f"'S' hərfi -> {letter_to_num('S')}")
print(f"18 rəqəmi -> '{num_to_letter(18)}'")
print(f"'A' hərfi -> {letter_to_num('A')}")
print(f"0 rəqəmi -> '{num_to_letter(0)}'")
````

### 5.2 Sezar şifrləmə funksiyası

````python
import string


def caesar_encrypt(text, shift):
    """
    Sezar şifrəsi ilə mətni şifrləyir
    shift: sürüşdürmə miqdarı (0-25 arası)
    """
    result = []
    for char in text:
        if char in string.ascii_letters:
            # Hərfin böyük/kiçik olduğunu saxla
            if char in string.ascii_uppercase:
                num = ord(char) - ord('A')
                encrypted_num = (num + shift) % 26
                result.append(chr(encrypted_num + ord('A')))
            else:
                num = ord(char) - ord('a')
                encrypted_num = (num + shift) % 26
                result.append(chr(encrypted_num + ord('a')))
        else:
            # Boşluq, durğu işarələri və qeyri-ASCII hərfləri olduğu kimi saxla
            result.append(char)
    return ''.join(result)


def caesar_decrypt(ciphertext, shift):
    """
    Sezar şifrəsi ilə şifrələnmiş mətni deşifrləyir
    Deşifrləmə üçün əks sürüşdürmə tətbiq edilir
    """
    return caesar_encrypt(ciphertext, -shift)
````

### 5.3 Sezar şifrəsinin test edilməsi

````python
# Test 1: İngilis dilində
message = "HELLO WORLD"
shift = 3
encrypted = caesar_encrypt(message, shift)
decrypted = caesar_decrypt(encrypted, shift)

print("📨 SEZAR ŞİFRƏ TESTİ")
print("=" * 40)
print(f"Orijinal: {message}")
print(f"Şifrələnmiş (shift={shift}): {encrypted}")
print(f"Deşifrələnmiş: {decrypted}")

# Test 2: Boşluq və durğu işarələri ilə
message2 = "Hello, World! How are you?"
encrypted2 = caesar_encrypt(message2, 5)
decrypted2 = caesar_decrypt(encrypted2, 5)

print("\n📨 İKİNCİ TEST")
print("=" * 40)
print(f"Orijinal: {message2}")
print(f"Şifrələnmiş (shift=5): {encrypted2}")
print(f"Deşifrələnmiş: {decrypted2}")
````

### ✍️ Çalışma 2: Sezar şifrəsi (2 bal)

Aşağıdakı tapşırıqları yerinə yetirin:

1. **Şifrləmə:** "KRIPTOQRAFIYA" sözünü $k=5$ açarı ilə şifrləyin
2. **Deşifrləmə:** "MABL" şifrəli mətnini $k=19$ açarı ilə deşifrələyin
3. **Əlavə:** `caesar_decrypt` funksiyasını `caesar_encrypt`-i çağırmadan, birbaşa deşifrləmə düsturu ilə yazın

````python
# Çalışma 2 - Cavablarınızı buraya yazın

print("📝 ÇALIŞMA 2 CAVABLARI")
print("-" * 30)

# 1. KRIPTOQRAFIYA sözünü k=5 ilə şifrləyin
message = "KRIPTOQRAFIYA"
shift = 5
encrypted = caesar_encrypt(message, shift)
print(f"1. {message} (k=5) -> {encrypted}")

# 2. MABL şifrəli mətnini k=19 ilə deşifrləyin
cipher = "MABL"
shift = 19
decrypted = caesar_decrypt(cipher, shift)
print(f"2. {cipher} (k=19) -> {decrypted}")

# 3. Birbaşa deşifrləmə funksiyası
def caesar_decrypt_direct(ciphertext, shift):
    """Birbaşa deşifrləmə düsturu ilə"""
    result = []
    for char in ciphertext:
        if char.isalpha():
            if char.isupper():
                num = ord(char) - ord('A')
                # Deşifrləmə düsturu: (c - k) mod 26
                decrypted_num = (num - shift) % 26
                result.append(chr(decrypted_num + ord('A')))
            else:
                num = ord(char) - ord('a')
                decrypted_num = (num - shift) % 26
                result.append(chr(decrypted_num + ord('a')))
        else:
            result.append(char)
    return ''.join(result)

# Test
test = "KHOOR"  # HELLO-nun şifrələnmiş forması (k=3)
print(f"\n3. Birbaşa deşifrləmə testi: {test} (k=3) -> {caesar_decrypt_direct(test, 3)}")
````

## 🔍 Sezar şifrəsinə tam axtarış hücumu (15 dəq)

Sezar şifrəsinin ən böyük zəifliyi cəmi 26 mümkün açarın olmasıdır. Bu o deməkdir ki, bütün variantları sınaqdan keçirərək şifrəni asanlıqla sındıra bilərik.

````python
def caesar_brute_force(ciphertext):
    """
    Sezar şifrəsinə tam axtarış hücumu
    Bütün 26 mümkün açarı sınaqdan keçirir
    """
    print("=" * 60)
    print("🔓 SEZAR ŞİFRƏSİNƏ TAM AXTARIŞ HÜCUMU")
    print("=" * 60)

    for shift in range(26):
        decrypted = []
        for char in ciphertext:
            if char.isalpha():
                # Hərfin böyük/kiçik olduğunu saxla
                if char.isupper():
                    num = ord(char) - ord('A')
                    decrypted_num = (num - shift) % 26
                    decrypted.append(chr(decrypted_num + ord('A')))
                else:
                    num = ord(char) - ord('a')
                    decrypted_num = (num - shift) % 26
                    decrypted.append(chr(decrypted_num + ord('a')))
            else:
                decrypted.append(char)

        print(f"Shift {shift:2d}: {''.join(decrypted)}")

    print("=" * 60)
    print("📌 İngilis dilində mənalı görünən sətir düzgün açardır")

# Test
ciphertext = "Wklv lv d vhfuhw phvvdjh"
caesar_brute_force(ciphertext)
````

## 🔢 Affin şifrəsinin implementasiyası (25 dəq)

Affin şifrəsi Sezar şifrəsinin ümumiləşdirilmiş formasıdır:  
$E(x) = (a \cdot x + b) \mod 26$  
$D(y) = a^{-1} \cdot (y - b) \mod 26$

### 7.1 GCD və qarşılıqlı sadəlik yoxlanışı

Affin şifrəsində $a$ əmsalı 26 ilə qarşılıqlı sadə olmalıdır. Əks halda şifrələmə tərsinə çevrilə bilməz.

````python
def gcd(a, b):
    """Evklid alqoritmi ilə ƏBOB hesablayır"""
    while b:
        a, b = b, a % b
    return a

def is_coprime(a, m=26):
    """a və m-in qarşılıqlı sadə olub-olmadığını yoxlayır"""
    return gcd(a, m) == 1

# Test
valid_a = [a for a in range(1, 26) if is_coprime(a)]
print(f"✅ Qarşılıqlı sadə a dəyərləri: {valid_a}")
print(f"📊 Sayı: {len(valid_a)} (12 olmalıdır)")

# Nümunə
print(f"\nGCD(7, 26) = {gcd(7, 26)} -> {'qarşılıqlı sadə' if is_coprime(7) else 'deyil'}")
print(f"GCD(13, 26) = {gcd(13, 26)} -> {'qarşılıqlı sadə' if is_coprime(13) else 'deyil'}")
````

### 7.2 Modul tərsin tapılması

Affin şifrəsində deşifrləmə üçün $a^{-1} \mod 26$ tələb olunur. Bu elə bir $x$ ədədidir ki, $(a \cdot x) \mod 26 = 1$ olsun.

````python
def mod_inverse(a, m=26):
    """
    a-nın mod m-ə görə tərsini tapır (əgər varsa)
    a^{-1} mod m = x elə ki, (a * x) % m = 1
    """
    for x in range(1, m):
        if (a * x) % m == 1:
            return x
    return None  # tərs yoxdursa

# Test
print("🔢 MODUL TƏRS CƏDVƏLİ")
print("-" * 40)
for a in valid_a:
    inv = mod_inverse(a)
    print(f"{a}^-1 mod 26 = {inv:2d} | Yoxlama: {a} * {inv} % 26 = {(a*inv)%26}")
````

### 7.3 Affin şifrləmə funksiyası

````python
def affine_encrypt(text, a, b):
    """
    Affin şifrəsi ilə mətni şifrləyir
    E(p) = (a * p + b) mod 26
    a və 26 qarşılıqlı sadə olmalıdır
    """
    if not is_coprime(a):
        raise ValueError(f"❌ Xəta: a={a} və 26 qarşılıqlı sadə deyil. Şifrləmə mümkün deyil.")

    result = []
    for char in text:
        if char.isalpha():
            # Hərfi rəqəmə çevir
            if char.isupper():
                num = ord(char) - ord('A')
                encrypted_num = (a * num + b) % 26
                result.append(chr(encrypted_num + ord('A')))
            else:
                num = ord(char) - ord('a')
                encrypted_num = (a * num + b) % 26
                result.append(chr(encrypted_num + ord('a')))
        else:
            result.append(char)

    return ''.join(result)

def affine_decrypt(ciphertext, a, b):
    """
    Affin şifrəsi ilə şifrələnmiş mətni deşifrləyir
    D(c) = a^{-1} * (c - b) mod 26
    """
    if not is_coprime(a):
        raise ValueError(f"❌ Xəta: a={a} və 26 qarşılıqlı sadə deyil. Deşifrləmə mümkün deyil.")

    a_inv = mod_inverse(a)
    result = []

    for char in ciphertext:
        if char.isalpha():
            if char.isupper():
                num = ord(char) - ord('A')
                decrypted_num = (a_inv * (num - b)) % 26
                result.append(chr(decrypted_num + ord('A')))
            else:
                num = ord(char) - ord('a')
                decrypted_num = (a_inv * (num - b)) % 26
                result.append(chr(decrypted_num + ord('a')))
        else:
            result.append(char)

    return ''.join(result)
````

### 7.4 Affin şifrəsinin test edilməsi

````python
# Test 1: Mühazirədəki nümunə
message = "HELLO"
a, b = 7, 3
encrypted = affine_encrypt(message, a, b)
decrypted = affine_decrypt(encrypted, a, b)

print("📨 AFFİN ŞİFRƏ TESTİ")
print("=" * 50)
print(f"Orijinal: {message}")
print(f"Şifrələnmiş (a={a}, b={b}): {encrypted}")
print(f"Deşifrələnmiş: {decrypted}")
print(f"✓ Uğurlu? {message == decrypted}")

# Test 2: Başqa nümunə
message2 = "CRYPTO"
a2, b2 = 5, 8
encrypted2 = affine_encrypt(message2, a2, b2)
decrypted2 = affine_decrypt(encrypted2, a2, b2)

print(f"\n📨 İKİNCİ TEST")
print("-" * 50)
print(f"Orijinal: {message2}")
print(f"Şifrələnmiş (a={a2}, b={b2}): {encrypted2}")
print(f"Deşifrələnmiş: {decrypted2}")
print(f"✓ Uğurlu? {message2 == decrypted2}")
````

````python
# Çalışma 4 - Cavablar

print("📝 ÇALIŞMA 4 CAVABLARI")
print("=" * 50)

# 1. KRIPTO sözünü şifrləyin
message1 = "KRIPTO"
a1, b1 = 5, 8
encrypted1 = affine_encrypt(message1, a1, b1)
print(f"\n1. {message1} (a={a1}, b={b1}) -> {encrypted1}")

# 2. XH mətnini deşifrələyin
cipher2 = "XH"
a2, b2 = 7, 11
decrypted2 = affine_decrypt(cipher2, a2, b2)
print(f"\n2. {cipher2} (a={a2}, b={b2}) -> {decrypted2}")

# 3. Xəta idarəsi testi
print("\n3. Xəta idarəsi testi:")
try:
    affine_encrypt("TEST", 13, 5)
except ValueError as e:
    print(f"   {e}")

# 4. Tam axtarış funksiyası
def affine_brute_force(ciphertext, show_all=False):
    """Bütün mümkün a və b kombinasiyalarını sınaqdan keçirir.

    show_all=False olduqda nəticələr yalnız sadə evristik filtrdən keçən
    namizədlərlə məhdudlaşdırılır. Bütün namizədləri görmək üçün
    show_all=True istifadə edin.
    """
    valid_a = [a for a in range(1, 26) if is_coprime(a)]
    filter_note = "bütün namizədlər" if show_all else "evristik filtrli namizədlər"
    print(f"\n4. Tam axtarış ({filter_note}; cəmi {len(valid_a)} * 26 = {len(valid_a)*26} kombinasiya):")

    for a in valid_a:
        for b in range(26):
            try:
                decrypted = affine_decrypt(ciphertext, a, b)
                # Evristik filtr: İngilis sözləri adətən boşluqlarla ayrılır.
                # Bu filtr bir sözlü və ya durğu işarəli düzgün cavabları gizlədə bilər.
                passes_filter = ' ' in decrypted and all(c.isalpha() or c.isspace() for c in decrypted)
                if show_all or passes_filter:
                    print(f"   a={a:2d}, b={b:2d}: {decrypted}")
            except:
                pass

# 5. DWWDFN DW GDZQ mətnini deşifrələyin
cipher5 = "DWWDFN DW GDZQ"
a5, b5 = 9, 4
decrypted5 = affine_decrypt(cipher5, a5, b5)
print(f"\n5. {cipher5} (a={a5}, b={b5}) -> {decrypted5}")

# Tam axtarışı test edək
# Bütün 312 namizədi görmək üçün: affine_brute_force(cipher5, show_all=True)
affine_brute_force(cipher5, show_all=False)
````

## 📊 Tezlik analizi (25 dəq)

Tezlik analizi monoalfabetik şifrələri sındırmaq üçün ən effektiv üsullardan biridir. Bu üsul hərflərin mətndə görünmə tezliyinə əsaslanır.

### 8.1 Hərf tezliklərinin hesablanması

````python
def letter_frequencies(text):
    """
    Mətndə hərflərin tezliyini hesablayır
    """
    # Yalnız hərfləri götür
    letters = [char.upper() for char in text if char.isalpha()]
    total = len(letters)

    # Tezlikləri hesabla
    freq = Counter(letters)

    # Faizlərlə birlikdə sırala
    freq_percent = {}
    for letter, count in freq.items():
        freq_percent[letter] = (count, count / total * 100 if total > 0 else 0)

    # Tezliyə görə sırala (çoxdan aza)
    sorted_freq = sorted(freq_percent.items(),
                        key=lambda x: x[1][1],
                        reverse=True)

    return sorted_freq, total

def print_frequencies(freq_data, total):
    """
    Tezlik analizi nəticələrini çap edir
    """
    print(f"📊 Ümumi hərf sayı: {total}")
    print("-" * 60)
    print("Hərf | Say | Tezlik (%) | Qrafik")
    print("-" * 60)

    for letter, (count, percent) in freq_data:
        bar = '█' * int(percent)
        print(f"  {letter}  | {count:3d} | {percent:6.2f}%   | {bar}")

    print("-" * 60)

def plot_frequencies(freq_data, title="Hərf Tezlikləri"):
    """
    Tezlik qrafikini çəkir (matplotlib tələb olunur)
    """
    if not HAS_MATPLOTLIB:
        print("⚠️ matplotlib yüklü deyil, qrafik çəkilə bilməz")
        return

    letters = [item[0] for item in freq_data]
    percents = [item[1][1] for item in freq_data]

    plt.figure(figsize=(12, 6))
    plt.bar(letters, percents, color='skyblue', edgecolor='navy')
    plt.title(title, fontsize=16)
    plt.xlabel('Hərflər', fontsize=12)
    plt.ylabel('Tezlik (%)', fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.show()
````

### 8.2 İngilis dili tezlik cədvəli

İngilis dilində hərflər eyni tezlikdə işlənmir. Ən çox işlənən hərf 'E', ən az işlənən hərf isə 'Z'-dir.

````python
# İngilis dilində hərflərin standart tezlikləri (%)
ENGLISH_FREQ = {
    'E': 12.7, 'T': 9.1, 'A': 8.2, 'O': 7.5, 'I': 7.0,
    'N': 6.7, 'S': 6.3, 'H': 6.1, 'R': 6.0, 'D': 4.3,
    'L': 4.0, 'C': 2.8, 'U': 2.8, 'M': 2.4, 'W': 2.4,
    'F': 2.2, 'G': 2.0, 'Y': 2.0, 'P': 1.9, 'B': 1.5,
    'V': 1.0, 'K': 0.8, 'J': 0.2, 'X': 0.2, 'Q': 0.1,
    'Z': 0.1
}

# İngilis dili tezlik sırası (çoxdan aza)
ENGLISH_ORDER = ['E', 'T', 'A', 'O', 'I', 'N', 'S', 'H', 'R', 'D',
                 'L', 'C', 'U', 'M', 'W', 'F', 'G', 'Y', 'P', 'B',
                 'V', 'K', 'J', 'X', 'Q', 'Z']

def print_english_frequencies():
    """İngilis dili standart tezliklərini çap edir"""
    print("📊 İNGİLİS DİLİ STANDART HƏRF TEZLİKLƏRİ (%):")
    print("-" * 50)

    sorted_freq = sorted(ENGLISH_FREQ.items(),
                        key=lambda x: x[1],
                        reverse=True)

    for letter, freq in sorted_freq:
        bar = '█' * int(freq)
        print(f"  {letter}  | {freq:5.1f}%   | {bar}")

print_english_frequencies()
````

### 8.3 Monoalfabetik şifrənin sındırılması

Tezlik analizi əsasında şifrəni sındırmaq üçün aşağıdakı addımları izləyirik:
1. Şifrəli mətndə hərflərin tezliyini hesabla
2. Ən çox işlənən hərfləri müəyyən et
3. İngilis dilində ən çox işlənən hərflərlə müqayisə et
4. Əvəzetmə cədvəli yarat
5. Mətni deşifrələ

````python
def frequency_attack(ciphertext):
    """
    Tezlik analizi əsasında monoalfabetik şifrəni sındırmağa çalışır
    """
    print("=" * 70)
    print("🔓 TEZLİK ANALİZİ İLƏ ŞİFRƏ SINDIRILMASI")
    print("=" * 70)

    # Şifrəli mətnin tezliklərini hesabla
    cipher_freq, total = letter_frequencies(ciphertext)

    print("\n📊 ŞİFRƏLİ MƏTNİN HƏRF TEZLİKLƏRİ:")
    print_frequencies(cipher_freq, total)

    print("\n📊 İNGİLİS DİLİ STANDART TEZLİKLƏRİ:")
    print_english_frequencies()

    # Qrafik çək (əgər matplotlib varsa)
    if HAS_MATPLOTLIB:
        plot_frequencies(cipher_freq, "Şifrəli Mətnin Hərf Tezlikləri")

    # Ən çox işlənən hərfləri müqayisə et
    print("\n" + "=" * 70)
    print("🔍 ANALİZ NƏTİCƏLƏRİ:")
    print("=" * 70)

    if cipher_freq:
        most_common_cipher = cipher_freq[0][0]
        print(f"\n1. Şifrəli mətndə ən çox işlənən hərf: '{most_common_cipher}'")
        print(f"   İngilis dilində ən çox işlənən hərf: 'E'")
        print(f"   ➡️ Ehtimal: {most_common_cipher} → E")

    if len(cipher_freq) > 1:
        second_common = cipher_freq[1][0]
        print(f"\n2. İkinci ən çox işlənən hərf: '{second_common}'")
        print(f"   İngilis dilində ikinci ən çox işlənən hərf: 'T'")
        print(f"   ➡️ Ehtimal: {second_common} → T")

    if len(cipher_freq) > 2:
        third_common = cipher_freq[2][0]
        print(f"\n3. Üçüncü ən çox işlənən hərf: '{third_common}'")
        print(f"   İngilis dilində üçüncü ən çox işlənən hərf: 'A'")
        print(f"   ➡️ Ehtimal: {third_common} → A")

    print("\n" + "=" * 70)
    print("📝 Əvəzetmə cədvəli yaratmaq üçün bu ehtimallardan istifadə edin.")
    print("Sonra əvəzetmə tətbiq edərək mətni deşifrələyin.")
````

### 8.4 Əvəzetmə şifrəsinin deşifrələnməsi

````python
def substitution_decrypt(ciphertext, mapping):
    """
    Əvəzetmə şifrəsini deşifrləyir
    mapping: şifrəli hərf -> açıq hərf (məsələn, {'G':'E', 'U':'T', ...})
    """
    result = []
    for char in ciphertext:
        if char.isalpha():
            upper_char = char.upper()
            if upper_char in mapping:
                # Əvəzetməni tətbiq et
                decrypted = mapping[upper_char]
                # Orijinal hərfin böyük/kiçik olduğunu saxla
                if char.isupper():
                    result.append(decrypted)
                else:
                    result.append(decrypted.lower())
            else:
                result.append(char)
        else:
            result.append(char)
    return ''.join(result)

def suggest_substitution(cipher_freq):
    """
    Tezlik analizi əsasında əvəzetmə cədvəli təklif edir
    """
    mapping = {}

    # Şifrəli hərfləri tezlik sırasına görə düz
    cipher_letters = [item[0] for item in cipher_freq]

    # Ən çox işlənən hərfləri eşələ
    for i in range(min(len(cipher_letters), len(ENGLISH_ORDER))):
        mapping[cipher_letters[i]] = ENGLISH_ORDER[i]

    return mapping
````

### ✍️ Çalışma 5: Tezlik analizi (2 bal)

Aşağıdakı şifrəli mətn monoalfabetik şifrə ilə şifrlənib. Tezlik analizi apararaq orijinal mətni tapın:

**Şifrəli mətn:** `"GUR FRPERG ZRFFNTR VF UVQQRA VA GUVF GRKG"`

**Addımlar:**
1. Şifrəli mətndə hərflərin tezliyini hesablayın
2. Ən çox işlənən hərfləri müəyyən edin
3. İngilis dilində ən çox işlənən hərflərlə müqayisə edin
4. Əvəzetmə cədvəli yaradın
5. Mətni deşifrələyin

> 💡 **İpucu:** Bu mətndə ən çox işlənən hərf R-dir. İngilis dilində ən çox işlənən hərf E-dir. Deməli, R → E.

````python
# Çalışma 5 - Tezlik analizi ilə şifrə sındırma

ciphertext = "GUR FRPERG ZRFFNTR VF UVQQRA VA GUVF GRKG"

print("📝 ÇALIŞMA 5: TEZLİK ANALİZİ")
print("=" * 60)
print(f"Şifrəli mətn: {ciphertext}")
print("-" * 60)

# 1. Tezlik analizi aparaq
freq, total = letter_frequencies(ciphertext)
print_frequencies(freq, total)

# 2. Əvəzetmə cədvəli təklif edək
mapping = suggest_substitution(freq)
print("\n📋 Təklif edilən əvəzetmə cədvəli:")
for cipher_char, plain_char in mapping.items():
    print(f"   {cipher_char} → {plain_char}")

# 3. Deşifrələyək
decrypted = substitution_decrypt(ciphertext, mapping)
print(f"\n🔓 Deşifrələnmiş mətn: {decrypted}")

# 4. Əl ilə düzəliş (əgər lazımdırsa)
print("\n🔄 Əl ilə düzəliş (əgər təklif olunan cədvəl tam dəqiq deyilsə):")
print("   Məsələn, G → T, U → H, R → E, F → S, P → C, Z → M, N → A")

# Daha dəqiq cədvəl
better_mapping = {
    'G': 'T', 'U': 'H', 'R': 'E', 'F': 'S', 'P': 'C',
    'Z': 'M', 'N': 'A', 'T': 'G', 'V': 'I', 'Q': 'D',
    'A': 'N', 'E': 'R', 'K': 'X'
}
better_decrypted = substitution_decrypt(ciphertext, better_mapping)
print(f"\n🔓 Daha dəqiq deşifrələmə: {better_decrypted}")
````
